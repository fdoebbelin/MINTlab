## Teil 1: Modelle für den 3D-Druck exportieren

### Schritt 1: Ein Objekt vorbereiten

Bevor Sie exportieren, stellen Sie sicher, dass Ihr Modell **druckfertig** ist:
- Das Objekt ist ein **geschlossener Körper** (keine offenen Flächen)
- Wandstärke mindestens **2–3 mm** (je nach Material)
- Keine **schwebenden Teile** ohne Stützstruktur

### Schritt 2: STL-Export in FreeCAD

**Navigieren Sie im Menü:**

```
Menü: Datei → Export…
```

**Im Dateiformat-Dialog:**
- Wählen Sie Ihr Modell im **Modellbaum** (linke Seite)
- Klicken Sie auf **Datei → Export**
- Wählen Sie unter **Dateityp**: `STL Mesh (*.stl)`
- Geben Sie einen **Dateinamen** ein (z. B. `MyPart.stl`)
- Klicken Sie auf **Speichern**

**Im Export-Dialog:**

```
Menü: Speichern im Format STL
→ ASCII (für Lesbarkeit) oder Binary (für Speichergröße)
→ Auflösung: Schieber auf „Fein" für bessere Qualität
```

Die Datei ist nun als STL-Mesh gespeichert – ein universelles Format für 3D-Drucker.

### Schritt 3: Alternative – STEP-Export für Datenaustausch

Wenn Sie das Modell mit anderen CAD-Systemen teilen möchten:

```
Menü: Datei → Export…
Dateityp: STEP (*.step) oder IGES (*.iges)
```

STEP bewahrt **parametrische Informationen** – andere können das Modell später noch bearbeiten.

## Teil 2: Grundlagen des 3D-Druck-Prozesses

### Der Workflow: Von STL zum gedruckten Objekt

```
┌─────────────┐
│  FreeCAD    │
│  3D-Modell  │ → STL-Export
└─────────────┘
      ↓
┌─────────────┐
│   Slicer    │ → Slicing in Schichten (0,1–0,3 mm Höhe)
│  (z.B. Cura)│ → G-Code generieren
└─────────────┘
      ↓
┌─────────────┐
│  3D-Drucker │ → Schicht für Schicht drucken
│             │ → 3–24 Stunden (abhängig von Größe)
└─────────────┘
      ↓
┌─────────────┐
│Fertiges Teil│ → Stützen entfernen → Oberfläche glätten
│             │
└─────────────┘
```

### Was ist der Slicer?

Der **Slicer** (z. B. **Cura**, **PrusaSlicer**, **SuperSlicer**) konvertiert Ihr STL-Modell:
1. Schneidet das Modell in hunderte Querschnitte
2. Plant den Druckweg (X-Y-Bewegung pro Schicht)
3. Generiert **G-Code** – die Maschinensprache des Druckers

**Wichtige Slicer-Einstellungen:**
- **Layer Height** (Schichthöhe): 0,1–0,3 mm (feiner = längerer Druck)
- **Fill Density** (Füllung): 15–100 % (weniger = Gewichtsersparnis, mehr = Stabilität)
- **Support**: Stützmaterial für Überhänge
- **Nozzle Temperature** & **Bed Temperature**: Material-spezifisch (PLA: 200°C/60°C)

## Teil 3: Technische Zeichnungen in FreeCAD

### Schritt 1: TechDraw-Arbeitsbereich aktivieren

Wechseln Sie zum **TechDraw-Arbeitsbereich**:

```
Oben rechts im Fenster:
Klick auf Dropdown-Menü (Standard)
→ TechDraw wählen
```

Nach dem Wechsel sehen Sie neue Werkzeuge in der Werkzeugleiste.

### Schritt 2: Neues Zeichnungsblatt erstellen

```
Menü: TechDraw → Neues Blatt
oder
Menü: TechDraw → Ansichten → Neues Blatt
```

**Im Dialog:**
- Template: `A4_Landscape` (Standard-Zeichnungsblatt)
- Klick auf **OK**

Ein neues Zeichnungsblatt (Blatt) erscheint im **Modellbaum** unter dem Namen `Page`.

### Schritt 3: 3D-Modell auf die Zeichnung projizieren

Um Ansichten (Vorder-, Seiten-, Draufsicht) zu erstellen:

```
Menü: TechDraw → Ansichten → Ansicht einfügen
oder
Werkzeugleiste: „Insert View" Icon anklicken
```

**Dialog Ansicht einfügen:**
- **Quelle**: Ihr 3D-Modell (z. B. `Body`)
- **Standardansicht**: `Front` (Vorderansicht) wählen
- **Scale** (Maßstab): z. B. 1:1 oder 1:2 für größere Teile
- Klick auf **OK**

Die **Vorderansicht** wird auf dem Zeichnungsblatt eingefügt.

### Schritt 4: Zusätzliche Ansichten hinzufügen

Um Seiten- und Draufsicht zu ergänzen:

```
Menü: TechDraw → Ansichten → Ansicht einfügen
→ Standardansicht: Top (Draufsicht)
→ Position: Rechts neben der Vorderansicht
→ OK
```

Wiederholen Sie dies für:
- **Right** (Seitenansicht) – rechts von Vorderansicht
- **Top** (Draufsicht) – oben

Die Ansichten sind jetzt nach technischen Normen angeordnet.

### Schritt 5: Bemaßung hinzufügen

Zum Einfügen von Maßlinien:

```
Menü: TechDraw → Bemaßungen und Hinweise → Horizontale Länge
oder
Werkzeugleiste: „Horizontal Distance" Icon
```

**Bemaßen durchführen:**
1. Klicken Sie auf zwei **Kanten oder Punkte** in der Ansicht
2. Klicken Sie im Zeichnungsblatt, wo die **Maßlinie** platziert werden soll
3. Die **Maßzahl** wird automatisch eingefügt

**Weitere Bemaßungswerkzeuge:**
```
TechDraw → Bemaßungen und Hinweise:
  - Vertikale Länge (Höhen)
  - Winkel (für Schrägflächen)
  - Radius/Durchmesser (für Kreise)
```

### Schritt 6: Text und Beschriftungen

Um Beschriftungen (z. B. Material, Toleranzen) hinzuzufügen:

```
Menü: TechDraw → Bemaßungen und Hinweise → Text
oder
Werkzeugleiste: „Insert Text" Icon
```

Klicken Sie auf das Zeichnungsblatt und geben Sie Text ein:
```
Beispiel:
Werkstoff: Aluminum 6061
Toleranz: ±0,1 mm
Gewicht: ~150 g
```

### Schritt 7: Ausgabe – Zeichnung ausdrucken oder als PDF speichern

```
Menü: Datei → Drucken
oder
Menü: Datei → Exportieren → PDF
```

**Als Bild speichern** (für Präsentationen):
```
Menü: Datei → Exportieren → PNG/SVG
```

## Praktisches Beispiel: Einfache Platte mit Bohrung

### 3D-Modell (kurz wiederholt)

Sie haben in **Modul 5** ein Modell erstellt:
- Grundform: **Quader** (100 × 50 × 10 mm)
- Feature: **Vier Bohrungen** (Ø8 mm) in regelmäßigem Muster

### Jetzt exportieren wir und zeichnen es:

#### Schritt A: STL-Export
```
1. Modellbaum: Body (oder Ihr Objekt-Name) wählen
2. Menü: Datei → Export
3. Format: STL Mesh
4. Name: Plate.stl
5. Speichern
```

#### Schritt B: Technische Zeichnung erstellen
```
1. Arbeitsbereich: TechDraw
2. TechDraw → Neues Blatt (A4)
3. TechDraw → Ansichten → Ansicht einfügen
   - Quelle: Body
   - Standardansicht: Front
   - Maßstab: 1:1
4. Ansicht einfügen (Top)
5. Bemaßungen hinzufügen:
   - Länge 100 mm
   - Breite 50 mm
   - Höhe 10 mm
   - Bohrung-Abstände
6. Text hinzufügen: „Montageplatte"
7. Als PDF exportieren
```

## Checkliste: Druckfehlervermeidung

Bevor Sie zum 3D-Druck exportieren, prüfen Sie:

- [ ] **Geschlossenes Objekt?** → Menü: Überprüfung → Überprüfe Geometrie
- [ ] **Wandstärke ≥ 2 mm?** → Messen Sie im 3D-Modell nach
- [ ] **Keine schwebenden Teile?** → Optisch in der 3D-Ansicht prüfen
- [ ] **STL-Datei < 50 MB?** → Zu große Dateien → Drucker-Probleme
- [ ] **Druckerkompatibilität?** → Größe des Bauvolumens prüfen
