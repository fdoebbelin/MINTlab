### üîß Schritte und Erkl√§rungen

#### Schritt 3: Die drei Rotationswerte verstehen

In FreeCAD gibt es drei Rotationswerte (Euler-Winkel):

| Wert | Name | Drehachse | Effekt |
|------|------|-----------|--------|
| Yaw | Gier | Z-Achse (blau) | Oben drehen (wie Kompass) |
| Pitch | Neigung | Y-Achse (gr√ºn) | Vorne/hinten kippen |
| Roll | Rolle | X-Achse (rot) | Links/rechts rollen |

#### Schritt 4: Yaw = 45¬∞

```freecad
# Zylinder mit Yaw = 45¬∞
cylinder.Placement.Rotation = FreeCAD.Rotation(0, 0, 45)
# Die Parameter sind: (Pitch, Yaw, Roll) ‚Äì ACHTUNG: Reihenfolge!
```

**Beobachtung**: Der Zylinder **dreht sich um die vertikale Achse** (die Spitze zeigt jetzt in eine andere Himmelsrichtung, aber der Zylinder steht immer noch aufrecht).

_Erkl√§rung_: Yaw ist die **Kompass-Rotation** ‚Äì oben drehen. Wenn du einen Kreisel mit der Spitze nach oben drehst, ist das Yaw.

#### Schritt 5: Roll = 90¬∞ (mit Yaw = 45¬∞)

```freecad
cylinder.Placement.Rotation = FreeCAD.Rotation(0, 45, 90)
# Yaw=45¬∞, Roll=90¬∞
```

**Beobachtung**: Der Zylinder **kippt zur Seite**. Er liegt jetzt nicht mehr aufrecht.

_Warum √§ndert sich die Richtung?_ Weil die Rotationen **nacheinander** angewendet werden. Erst Yaw (45¬∞), dann Roll (90¬∞). Die Reihenfolge beeinflusst das Endergebnis!

**Wichtig**: Dies ist ein wichtiges Konzept: **Rotationsreihenfolge z√§hlt!**

#### Schritt 7: Pitch = 90¬∞ (mit Yaw = 0¬∞, Roll = 0¬∞)

```freecad
cylinder.Placement.Rotation = FreeCAD.Rotation(90, 0, 0)
# Pitch=90¬∞, Yaw=0¬∞, Roll=0¬∞
```

**Beobachtung**: Der Zylinder **kippt nach vorne** und liegt auf der Seite. Die Spitze zeigt jetzt nach links oder rechts (abh√§ngig vom Blickwinkel).

_Erkl√§rung_: Pitch ist die **Vorw√§rts/R√ºckw√§rts-Neigung** ‚Äì genau wie wenn du deinen Kopf nach vorne neigst.

### üí° Antworten auf Reflexionsfragen

**Frage 1:** Wenn du einen Zylinder **90¬∞** um **Pitch** drehst, liegt er dann wirklich komplett auf der Seite?

**Antwort**: **Ja**, wenn du nur Pitch √§ndert und die anderen Werte 0 l√§sst, liegt der Zylinder perfekt auf der Seite.

_Visualisierung_:
```
Vorher (aufrecht):        Nach Pitch=90¬∞ (auf der Seite):
        |                           ___
        |                        /       \
        |                       |         |
      \ | /                     |_________|
```

**Frage 2:** Was passiert, wenn du alle drei Werte auf **45¬∞** stellst?

**Antwort**: Der Zylinder dreht sich **gleichzeitig** um alle drei Achsen. Das Ergebnis ist eine **komplexe 3D-Drehung**. Der Zylinder liegt schr√§g im Raum ‚Äì nicht aufrecht, nicht auf der Seite, sondern irgendwo dazwischen.

_Mathematik dahinter_: Dies funktioniert durch **Rotationsmatrizen** ‚Äì aber das ist f√ºr diesen Kurs zu tief. Wichtig: Experimentiere einfach!

- Durchschnittmengen bilden (Schnittmenge)

Damit kannst du noch komplexere 3D-Formen erschaffen!
