## Lösung Aufgabe 1: Temperaturskalen und thermisches Gleichgewicht

Berechnung der Mischungstemperatur:

Formel:
$$
m_1 \cdot (T_m - T_1) + m_2 \cdot (T_m - T_2) = 0
$$

Einsetzen der Werte:
- $m_1 = 0{,}5 \ \text{kg}$, $T_1 = 20^\circ \,\text{C}$
- $m_2 = 0{,}3 \ \text{kg}$, $T_2 = 80^\circ \,\text{C}$

Umformen:
$$
T_m = \frac{m_1 T_1 + m_2 T_2}{m_1 + m_2} = \frac{0{,}5 \cdot 20 + 0{,}3 \cdot 80}{0{,}8}
$$
$$
= \frac{10 + 24}{0{,}8} = \frac{34}{0{,}8} = 42{,}5^\circ  \text{C}
$$

**Python-Code:**

```python
m1 = 0.5
T1 = 20
m2 = 0.3
T2 = 80
T_mix = (m1 * T1 + m2 * T2)/(m1 + m2)
print(T_mix) # 42.5 °C
```

**Diskussion:**
Kein Wärmeverlust, spezifische Wärmekapazität kürzt sich heraus.

## Lösung Aufgabe 2: Ideales Gasgesetz und Zustandsänderungen

### Teil A:
$$
p \cdot V = n \cdot R \cdot T$$

Gegeben:
- $n = 2 \ \text{mol}$, $R = 8{,}314$, $T = 300 \ \text{K}$, $p = 1 \ \text{bar} = 1 \times 10^5 \ \text{Pa}$
$$
V = \frac{nRT}{p} = \frac{2 \cdot 8{,}314 \cdot 300}{100{,}000} = 0{,}0499 \ \text{m}^3 = 49{,}94 \ \text{L}
$$

### Teil B:

$$
p_1 V_1 = p_2 V_2$$

Neuer Druck: $p_2 = 0{,}5 \ \text{bar} = 50{,}000 \ \text{Pa}$
$$
V_2 = \frac{p_1 V_1}{p_2} = \frac{1 \cdot 0{,}0499}{0{,}5} = 0{,}0998 \ \text{m}^3
$$

### Teil C:

$$
W = nRT \ln\left(\frac{V_2}{V_1}\right) = nRT \ln(2) = 2 \cdot 8{,}314 \cdot 300 \cdot 0{,}693 = 3460 \ \text{J}
$$

## Lösung Aufgabe 6: Hauptsätze kombiniert – Kreisprozess

**Schritt A → B (isotherm):**
$$
W = nRT \ln\left(\frac{p_A}{p_B}\right)
$$
**Schritt B → C (isobar):**
$$
W = nR(T_C-T_B), \quad Q = nc_p(T_C-T_B)
$$
**Schritt C → A (isochor):**
$$
W = 0, \quad Q = nc_v(T_A-T_C)
$$

**Gesamtprozess:**
Die Summe der Arbeit und Wärme sollte über einen vollständigen Kreisprozess nahe 0 sein. Kleine numerische Abweichungen entstehen durch Rundung der Werte.

## Lösung Aufgabe 7: Spezifische Wärmekapazität bestimmen

Energiebilanz:
$$
c_\text{Metall} = - \frac{m_\text{Wasser} \cdot c_\text{Wasser} \cdot \Delta T_\text{Wasser}}{m_\text{Metall} \cdot \Delta T_\text{Metall}}
$$
Einsetzen:
- $m_\text{Metall}=0{,}5$, $T_\text{Metall initial}=100$, $T_\text{final}=25$
- $m_\text{Wasser}=0{,}2$, $T_\text{Wasser initial}=20$
- $c_\text{Wasser}=4180$

$$
\Delta T_\text{Metall} = 25-100 = -75 \\
\Delta T_\text{Wasser} = 25-20 = 5
$$

$$
c_\text{Metall} = -\frac{0{,}2 \cdot 4180 \cdot 5}{0{,}5 \cdot -75} = \frac{4180}{75} = 446.6 \ \text{J/(kg·K)}
$$
