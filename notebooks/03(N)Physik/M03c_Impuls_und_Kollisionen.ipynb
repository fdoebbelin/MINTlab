## Lösung zu Aufgabe 1: Impuls berechnen

### Vollständige Lösung

````python
# Aufgabe 1: Impuls berechnen

# Objekt 1: Auto
masse_auto = 1500  # kg
geschwindigkeit_auto = 20  # m/s
impuls_auto = masse_auto * geschwindigkeit_auto

# Objekt 2: Tennisball
masse_ball = 0.058  # kg
geschwindigkeit_ball = 50  # m/s
impuls_ball = masse_ball * geschwindigkeit_ball

# Objekt 3: Mensch
masse_mensch = 70  # kg
geschwindigkeit_mensch = 2  # m/s
impuls_mensch = masse_mensch * geschwindigkeit_mensch

print(f"Impuls Auto: {impuls_auto} kg·m/s")
print(f"Impuls Tennisball: {impuls_ball} kg·m/s")
print(f"Impuls Mensch: {impuls_mensch} kg·m/s")

# Zusatzfrage: Welcher Impuls ist am größten?
impulse = [
    ("Auto", impuls_auto),
    ("Tennisball", impuls_ball),
    ("Mensch", impuls_mensch)
]

max_impuls = max(impulse, key=lambda x: x[1])
print(f"\nGrößter Impuls: {max_impuls[0]} mit {max_impuls[1]} kg·m/s")
````

### Ausgabe

```
Impuls Auto: 30000 kg·m/s
Impuls Tennisball: 2.9 kg·m/s
Impuls Mensch: 140 kg·m/s

Größter Impuls: Auto mit 30000 kg·m/s
```

### Erklärung & Verständnis

**Warum hat das Auto einen größeren Impuls als der Tennisball, obwohl der Ball schneller ist?**

Der Impuls hängt von **zwei Faktoren** ab: Masse und Geschwindigkeit. Das Auto ist etwa 26.000-mal schwerer als der Tennisball. Obwohl der Ball 2,5× schneller ist, überwiegt die große Massendifferenz.

Mathematisch:
$$
\text{Impuls Auto} = 1500 \times 20 = 30.000 \text{ kg·m/s}
$$
$$
\text{Impuls Ball} = 0,058 \times 50 = 2,9 \text{ kg·m/s}
$$

Das zeigt: **Massive, langsame Objekte können den gleichen oder größeren Impuls haben wie leichte, schnelle Objekte.**

## Lösung zu Aufgabe 2: Inelastischer Stoß

### Vollständige Lösung

````python
# Aufgabe 2: Inelastischer Stoß

# Vorher-Zustand
m_A = 5  # kg
v_A_vor = 10  # m/s
m_B = 15  # kg
v_B_vor = 0  # m/s

# Schritt 1: Gesamtimpuls vorher berechnen
p_gesamt = m_A * v_A_vor + m_B * v_B_vor

# Schritt 2: Gesamtmasse berechnen
m_gesamt = m_A + m_B

# Schritt 3: Geschwindigkeit nach dem Stoß berechnen
v_nachher = p_gesamt / m_gesamt

print(f"Gesamtimpuls: {p_gesamt} kg·m/s")
print(f"Gesamtmasse: {m_gesamt} kg")
print(f"Geschwindigkeit nach Stoß: {v_nachher} m/s")
print(f"Geschwindigkeit nach Stoß (gerundet): {v_nachher:.2f} m/s")
````

### Ausgabe

```
Gesamtimpuls: 50 kg·m/s
Gesamtmasse: 20 kg
Geschwindigkeit nach Stoß: 2.5 m/s
Geschwindigkeit nach Stoß (gerundet): 2.50 m/s
```

### Zusatz-Lösung: Energievergleich

````python
# Kinetische Energie vorher (beide Objekte separat)
KE_vorher = 0.5 * m_A * v_A_vor**2 + 0.5 * m_B * v_B_vor**2

# Kinetische Energie nachher (kombiniert)
KE_nachher = 0.5 * m_gesamt * v_nachher**2

# Differenz
energie_verlust = KE_vorher - KE_nachher

print(f"Energie vorher: {KE_vorher} J")
print(f"Energie nachher: {KE_nachher} J")
print(f"Energie verloren: {energie_verlust} J")
print(f"Prozentuale Energie verloren: {100 * energie_verlust / KE_vorher:.1f}%")
````

### Ausgabe

```
Energie vorher: 250.0 J
Energie nachher: 62.5 J
Energie verloren: 187.5 J
Prozentuale Energie verloren: 75.0%
```

### Analyse & Vertiefung

**Was passiert hier?**

1. **Impulserhaltung funktioniert:** Vorher 50 kg·m/s, nachher auch 50 kg·m/s ✓
2. **Energieverlust:** Von 250 J auf 62,5 J → **75% der Energie verschwindet!**

**Wo geht die Energie hin?**
- Verformung der Objekte
- Reibung
- Wärme
- Schall

Dies ist charakteristisch für **inelastische Stöße**: Nur Impuls bleibt erhalten, nicht die kinetische Energie.

## Lösung zu Aufgabe 3: Elastischer Stoß

### Vollständige Lösung

````python
# Aufgabe 3: Elastischer Stoß

m1 = 0.17  # kg
v1_vor = 4  # m/s
m2 = 0.17  # kg
v2_vor = 0  # m/s

# Bei gleicher Masse und elastischem Stoß: Geschwindigkeiten tauschen
v1_nach = v2_vor  # Kugel 1 bekommt die Geschwindigkeit von Kugel 2
v2_nach = v1_vor  # Kugel 2 bekommt die Geschwindigkeit von Kugel 1

# Impulserhaltung prüfen
p_vorher = m1 * v1_vor + m2 * v2_vor
p_nachher = m1 * v1_nach + m2 * v2_nach

print(f"Kugel 1 - vorher: {v1_vor} m/s, nachher: {v1_nach} m/s")
print(f"Kugel 2 - vorher: {v2_vor} m/s, nachher: {v2_nach} m/s")
print(f"\nImpuls vorher: {p_vorher} kg·m/s")
print(f"Impuls nachher: {p_nachher} kg·m/s")
print(f"Impuls erhalten? {p_vorher == p_nachher}")

# Energieprüfung
KE_vorher = 0.5 * m1 * v1_vor**2 + 0.5 * m2 * v2_vor**2
KE_nachher = 0.5 * m1 * v1_nach**2 + 0.5 * m2 * v2_nach**2

print(f"\nEnergie vorher: {KE_vorher} J")
print(f"Energie nachher: {KE_nachher} J")
print(f"Energie erhalten? {KE_vorher == KE_nachher}")
````

```
Kugel 2 - vorher: 0 m/s, nachher: 4 m/s

Impuls vorher: 0.68 kg·m/s
Impuls nachher: 0.68 kg·m/s
Impuls erhalten? True

Energie vorher: 1.36 J
Energie nachher: 1.36 J
Energie erhalten? True
```

### Tiefergehendes Verständnis

**Warum funktioniert der Geschwindigkeitstausch?**

Bei einem elastischen Stoß zwischen zwei gleich schweren Objekten (eines in Ruhe) gibt es eine mathematische Symmetrie:

$$
m_1 v_1 = m_1 v_1' + m_2 v_2' \quad \text{(Impuls)}
$$
$$
\frac{1}{2}m_1 v_1^2 = \frac{1}{2}m_1 v_1'^2 + \frac{1}{2}m_2 v_2'^2 \quad \text{(Energie)}
$$

Wenn $m_1 = m_2$ und $v_2 = 0$:
- $v_1 = v_1' + v_2'$
- $v_1^2 = v_1'^2 + v_2'^2$

Die einzige Lösung ist: $v_1' = 0$ und $v_2' = v_1$ (oder $v_1' = v_1, v_2' = 0$, aber das ist der triviale Fall vor dem Stoß).

## Lösung zu Aufgabe 4: Parameter-Variation

### Vollständige Lösung

````python
# Aufgabe 4: Parameter-Variation mit verschiedenen Massen

m_A = 10  # kg (bleibt gleich)
v_A = 5   # m/s (bleibt gleich)

szenarien = [
    (5, "m_B ist leicht (5 kg)"),
    (10, "m_B ist gleich schwer (10 kg)"),
    (20, "m_B ist schwer (20 kg)"),
]

print("Effekt der Masse von B auf die Endgeschwindigkeit:")
print("-" * 60)

for m_B_wert, beschreibung in szenarien:
    m_B = m_B_wert
    v_B = 0  # Ruht
    
    # Impulserhaltung anwenden
    p_gesamt = m_A * v_A + m_B * v_B
    m_gesamt = m_A + m_B
    v_nachher = p_gesamt / m_gesamt
    
    print(f"{beschreibung}:")
    print(f"  v_nachher = {v_nachher:.3f} m/s")
````

### Ausgabe

```
Effekt der Masse von B auf die Endgeschwindigkeit:
------------------------------------------------------------
m_B ist leicht (5 kg):
  v_nachher = 3.333 m/s
m_B ist gleich schwer (10 kg):
  v_nachher = 2.500 m/s
m_B ist schwer (20 kg):
  v_nachher = 1.667 m/s
```

### Visualisierung und Interpretation

````python
import matplotlib.pyplot as plt
import numpy as np

# Parametervariationsreihe
massen_B = np.linspace(1, 30, 100)
geschwindigkeiten = []

m_A = 10
v_A = 5

for m_B in massen_B:
    v_B = 0
    p_gesamt = m_A * v_A + m_B * v_B
    m_gesamt = m_A + m_B
    v_nach = p_gesamt / m_gesamt
    geschwindigkeiten.append(v_nach)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(massen_B, geschwindigkeiten, 'b-', linewidth=2, label='Endgeschwindigkeit')
plt.scatter([5, 10, 20], [3.333, 2.5, 1.667], color='red', s=100, zorder=5, label='Beispiel-Szenarien')

plt.xlabel('Masse von Objekt B (kg)', fontsize=12)
plt.ylabel('Endgeschwindigkeit (m/s)', fontsize=12)
plt.title('Parameter-Variation: Einfluss von m_B auf die Endgeschwindigkeit', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.ylim(0, 6)

plt.tight_layout()
plt.show()
````

### Beobachtungen & Physikalische Bedeutung

**Beobachtung:**
- Mit m_B = 5 kg → v_nachher = 3,33 m/s
- Mit m_B = 10 kg → v_nachher = 2,50 m/s
- Mit m_B = 20 kg → v_nachher = 1,67 m/s

**Antwort auf Fragen:**

1. **Wird die Endgeschwindigkeit größer oder kleiner, je schwerer m_B ist?**
   → Die Endgeschwindigkeit wird **kleiner**

2. **Warum ist das physikalisch sinnvoll?**
   → Ein schwereres Objekt "bremst" das System mehr. Der Impuls wird auf eine größere Gesamtmasse verteilt, daher wird die durchschnittliche Geschwindigkeit kleiner.

**Mathematisch:**
$$
v_{nachher} = \frac{m_A \cdot v_A}{m_A + m_B}
$$

Je größer $m_B$, desto kleiner wird der Bruch (Nenner wächst).

## Lösung zu Aufgabe 5: 2D-Kollision mit Vektoren

### Vollständige Lösung

````python
import numpy as np

# Aufgabe 5: 2D-Kollision

m_A = 5  # kg
v_A = np.array([6, 0])  # m/s (nach rechts)

m_B = 5  # kg
v_B = np.array([0, 4])  # m/s (nach oben)

# Berechne die Impulsvektoren
p_A = m_A * v_A
p_B = m_B * v_B

# Berechne den Gesamtimpuls (Vektoraddition)
p_gesamt = p_A + p_B

# Berechne die Gesamtmasse
m_gesamt = m_A + m_B

# Berechne die neue Geschwindigkeit
v_nachher = p_gesamt / m_gesamt

print(f"Impuls A: {p_A}")
print(f"Impuls B: {p_B}")
print(f"Gesamtimpuls: {p_gesamt}")
print(f"Geschwindigkeit nach Stoß: {v_nachher}")
print(f"Betrag der Geschwindigkeit: {np.linalg.norm(v_nachher):.2f} m/s")

# Winkel berechnen
winkel_rad = np.arctan2(v_nachher[1], v_nachher[0])
winkel_grad = np.degrees(winkel_rad)
print(f"Richtung: {winkel_grad:.1f}° von der x-Achse")
````

### Ausgabe

```
Impuls A: [30  0]
Impuls B: [ 0 20]
Gesamtimpuls: [30 20]
Geschwindigkeit nach Stoß: [3. 2.]
Betrag der Geschwindigkeit: 3.61 m/s
Richtung: 33.7° von der x-Achse
```

### Visualisierung mit Matplotlib

````python
import matplotlib.pyplot as plt

# Zeichne die Vektoren
fig, ax = plt.subplots(figsize=(10, 8))

# Skalierung für bessere Sichtbarkeit
scale = 1

# Pfeile für die Geschwindigkeitsvektoren
ax.arrow(0, 0, v_A[0]*scale, v_A[1]*scale, 
         head_width=0.3, head_length=0.2, fc='red', ec='red', 
         linewidth=2, label='v_A = (6, 0) m/s', alpha=0.7)

ax.arrow(0, 0, v_B[0]*scale, v_B[1]*scale, 
         head_width=0.3, head_length=0.2, fc='blue', ec='blue', 
         linewidth=2, label='v_B = (0, 4) m/s', alpha=0.7)

ax.arrow(0, 0, v_nachher[0]*scale, v_nachher[1]*scale, 
         head_width=0.3, head_length=0.2, fc='green', ec='green', 
         linewidth=3, label=f'v_nachher = ({v_nachher[0]:.1f}, {v_nachher[1]:.1f}) m/s', 
         zorder=5)

# Konstruktionslinie (Parallelogramm)
ax.plot([v_A[0], p_gesamt[0]/m_gesamt], [0, v_B[1]], 'k--', alpha=0.3, linewidth=1)
ax.plot([0, v_A[0]], [v_B[1], v_B[1]], 'k--', alpha=0.3, linewidth=1)

ax.set_xlim(-1, 8)
ax.set_ylim(-1, 6)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=11, loc='upper left')
ax.set_xlabel('x-Komponente (m/s)', fontsize=12)
ax.set_ylabel('y-Komponente (m/s)', fontsize=12)
ax.set_title('2D-Kollision: Vektorielle Impulserhaltung', fontsize=14)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)

plt.tight_layout()
plt.show()
````

### Tiefergehendes Verständnis

**Wie funktioniert die Vektorrechnung?**

Impuls ist ein **Vektor**:
$$
\vec{p} = m \cdot \vec{v}
$$

Für die x-Komponente:
$$
p_{x,gesamt} = 5 \times 6 + 5 \times 0 = 30 \text{ kg·m/s}
$$

Für die y-Komponente:
$$
p_{y,gesamt} = 5 \times 0 + 5 \times 4 = 20 \text{ kg·m/s}
$$

Die neue Geschwindigkeit hat Komponenten:
$$
v_{x,nachher} = \frac{30}{10} = 3 \text{ m/s}
$$
$$
v_{y,nachher} = \frac{20}{10} = 2 \text{ m/s}
$$

Der **Betrag** ist:
$$
|\vec{v}_{nachher}| = \sqrt{3^2 + 2^2} = \sqrt{13} \approx 3,61 \text{ m/s}
$$

## Lösung zu Aufgabe 6: Reflexion und Vertiefung

### Frage 1: Warum ist Impulserhaltung wichtig?

**Antwort:**

Das Prinzip der Impulserhaltung ist fundamental, weil es:

1. **Universell gültig** ist – in isolierten Systemen ohne externe Kräfte
2. **Vorhersagen ermöglicht** – wir können Endgeschwindigkeiten berechnen, ohne die Kräfte während der Kollision zu kennen
3. **Energieeffizienz analysiert** – in Kombination mit Energieerhaltung können wir Art und Grad von Energieverlusten bestimmen
4. **Technologische Anwendungen** hat – Raketen, Airbags, Sicherheitsysteme

**Alltags-Beispiel:**
Stell dir einen Billard-Tisch vor. Du kennst die Startgeschwindigkeit der weißen Kugel, nicht aber die genauen Kräfte beim Stoß. Mit dem Impulssatz kannst du trotzdem vorhersagen, wie sich die farbigen Kugeln nach dem Stoß bewegen.

---

### Frage 2: Wohin geht die Energie bei inelastischen Stößen?

**Antwort – Mindestens 3 Möglichkeiten:**

1. **Verformung:** Wenn zwei Autos frontal kollidieren, verformen sie sich. Die kinetische Energie wird in Verformungsenergie umgewandelt und bleibt als Dellen erhalten.

2. **Wärmeerzeugung:** Bei der Reibung zwischen Oberflächen und der inneren Reibung im Material entsteht Wärme.

3. **Schall:** Das laute Krachen bei einer Kollision ist eine Form von Schallenergie, die sich als Wellen ausbreitet.

4. **Innere Energie des Materials:** Moleküle im Material werden stärker in Schwingung versetzt, was die Temperatur erhöht.

5. **Plastische Verformung:** Atombindungen werden gebrochen und neu geformt.

## Erweiterte Konzepte (Bonus-Material)

### Elastischer Stoß – Allgemeine Formel

Für zwei Objekte verschiedener Massen mit elastischem Stoß:

$$
v_1' = \frac{(m_1 - m_2)v_1 + 2m_2 v_2}{m_1 + m_2}
$$

$$
v_2' = \frac{(m_2 - m_1)v_2 + 2m_1 v_1}{m_1 + m_2}
$$

**Spezialfall: $m_1 = m_2$ und $v_2 = 0$:**
$$
v_1' = \frac{(m - m)v_1}{2m} = 0
$$
$$
v_2' = \frac{2m v_1}{2m} = v_1
$$

Dies bestätigt den Geschwindigkeitstausch aus Aufgabe 3!

---

### Coefficient of Restitution (Rücksprungkoeffizient)

Der **Restitutionskoeffizient** $e$ misst, wie elastisch ein Stoß ist:

$$
e = \frac{|v_2' - v_1'|}{|v_1 - v_2|}
$$

- $e = 1$: Perfekt elastisch (Energie bleibt erhalten)
- $e = 0$: Perfekt inelastisch (Objekte bleiben zusammen)
- $0 < e < 1$: Teilweise elastisch (reale Fälle)

**Beispiel: Gummiball auf Beton**
- Ein Gummiball hat $e \approx 0,8$ (bounced gut)
- Tonerde hat $e \approx 0,05$ (absorbiert fast alles)

````python
# Restitutionskoeffizient berechnen
v1_vor = 4.0
v2_vor = 0.0
v1_nach = 0.0
v2_nach = 4.0

e = abs(v2_nach - v1_nach) / abs(v1_vor - v2_vor)
print(f"e = {e} (perfekt elastisch!)")
````

## Zusammenfassung der Lösungsstrategien

### Strategie 1: Impulsberechnung
```
p = m × v
```

### Strategie 2: Impulserhaltung (eindimensional)
```
m₁v₁ + m₂v₂ = m₁v₁' + m₂v₂'
```

### Strategie 3: Inelastischer Stoß (Objekte bleiben zusammen)
```
(m₁ + m₂) × v_nachher = m₁v₁ + m₂v₂
```

### Strategie 4: Vektorielle Impulserhaltung (2D/3D)
```
Komponentenweise rechnen:
p_x_gesamt = m₁v₁ₓ + m₂v₂ₓ
p_y_gesamt = m₁v₁ᵧ + m₂v₂ᵧ

v_nachher = p_gesamt / m_gesamt
```

## Häufige Fehler und wie man sie vermeidet

| Fehler | Beispiel | Korrektur |
|--------|----------|-----------|
| Verwechslung Impuls ↔ Energie | Denken, Impuls = 0,5mv² | Impuls = m·v, Energie = 0,5m·v² |
| Richtungen ignorieren | In 2D alle Komponenten addieren | Komponentenweise rechnen |
| Äußere Kräfte übersehen | System als isoliert annehmen | Prüfen, ob Reibung/Luft wirken |
| Geschwindigkeit negieren | Vorzeichen ignorieren | Richtung als Vorzeichen nutzen |
