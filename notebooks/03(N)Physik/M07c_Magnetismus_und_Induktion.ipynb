## LÃ¶sung zu Aufgabe 1: Rechte-Hand-Regel und Lorentzkraft

### Theoretischer Hintergrund

Die **Lorentzkraft** wird durch das Kreuzprodukt berechnet:
$$\vec{F} = q(\vec{v} \times \vec{B})$$

**Wichtig:** Die Rechte-Hand-Regel funktioniert fÃ¼r **positive Ladungen**. Bei **negativen Ladungen** (wie Elektronen) zeigt die Kraft in die **entgegengesetzte Richtung**.

### Schritt-fÃ¼r-Schritt LÃ¶sung

1. **Geschwindigkeit** nach rechts: $\vec{v} = v \cdot \vec{e}_x$
2. **Magnetfeld** aus der Ebene: $\vec{B} = B \cdot \vec{e}_z$
3. **Rechte-Hand-Regel** (fÃ¼r positive Ladung):
   - Zeigefinger: +x (rechts)
   - Mittelfinger: +z (aus Ebene)
   - Daumen: +y (oben)
   - â†’ $\vec{v} \times \vec{B} = v \cdot B \cdot \vec{e}_y$ (nach oben)

4. **FÃ¼r Elektron** (q < 0):
   - $\vec{F} = q \cdot (v \times B) = (-e) \cdot (v \cdot B \cdot \vec{e}_y)$
   - $\vec{F} = -v \cdot B \cdot \vec{e}_y$ (nach **unten**, also in -y-Richtung)

### VollstÃ¤ndiger Python-Code

````python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Parameter
q_electron = -1.602e-19  # Ladung eines Elektrons (negativ!)
v_magnitude = 1e6        # Geschwindigkeit (m/s)
B_magnitude = 0.5        # Magnetische Flussdichte (Tesla)

# Einheitsvektoren und Skalierung
v = v_magnitude * np.array([1, 0, 0])  # Geschwindigkeit nach rechts (+x)
B = B_magnitude * np.array([0, 0, 1])  # Magnetfeld aus der Ebene (+z)

# Kreuzprodukt v Ã— B
cross_product = np.cross(v, B)
print(f"v Ã— B = {cross_product}")  # Sollte in +y-Richtung zeigen

# Lorentzkraft F = q * (v Ã— B)
F = q_electron * cross_product
print(f"FÃ¼r negative Ladung: F = qÂ·(v Ã— B) = {F}")
print(f"Kraft zeigt in -{' y'}-Richtung (nach unten)")

# Betrag der Kraft
F_betrag = np.linalg.norm(F)
print(f"\n|F| = qÂ·vÂ·BÂ·sin(Î¸) = {q_electron:.2e} * {v_magnitude:.2e} * {B_magnitude} * sin(90Â°)")
print(f"|F| = {F_betrag:.2e} N")

# 3D-Visualisierung
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# Ursprung (Ausgangspunkt der Vektoren)
origin = np.array([0, 0, 0])

# Pfeile zeichnen
scale = 0.5  # Skalierungsfaktor fÃ¼r Visualisierung
ax.quiver(0, 0, 0, v[0]*scale/v_magnitude, v[1]*scale/v_magnitude, v[2]*scale/v_magnitude, 
          color='red', arrow_length_ratio=0.1, linewidth=3, label='Geschwindigkeit v (+x)')

ax.quiver(0, 0, 0, B[0]*scale/B_magnitude, B[1]*scale/B_magnitude, B[2]*scale/B_magnitude, 
          color='green', arrow_length_ratio=0.1, linewidth=3, label='Magnetfeld B (+z)')

# Kraft normalisieren fÃ¼r Visualisierung
F_normalized = F / np.linalg.norm(F)
ax.quiver(0, 0, 0, F_normalized[0]*scale, F_normalized[1]*scale, F_normalized[2]*scale, 
          color='blue', arrow_length_ratio=0.1, linewidth=3, label='Lorentzkraft F (-y)')

# Hilfslinie: Rechte-Hand-Regel
ax.plot([0, 1], [0, 0], [0, 0], 'r--', alpha=0.3, linewidth=1)
ax.plot([0, 0], [0, 1], [0, 0], 'g--', alpha=0.3, linewidth=1)
ax.plot([0, 0], [0, 0], [0, 1], 'b--', alpha=0.3, linewidth=1)

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_zlim(-1.5, 1.5)
ax.set_xlabel('X (Rechts)', fontsize=11, fontweight='bold')
ax.set_ylabel('Y (Oben/Unten)', fontsize=11, fontweight='bold')
ax.set_zlabel('Z (Aus Ebene)', fontsize=11, fontweight='bold')
ax.set_title('Lorentzkraft auf Elektron: F = q(v Ã— B)\nRoter Pfeil: v, GrÃ¼ner: B, Blauer: F (Elektron â†’ Kraft zeigt nach unten!)', 
             fontsize=12, fontweight='bold')
ax.legend(fontsize=10, loc='upper right')

plt.show()

print("\n" + "="*60)
print("ZUSAMMENFASSUNG")
print("="*60)
print("FÃ¼r ein Elektron (negative Ladung):")
print("â€¢ Geschwindigkeit: +x (rechts)")
print("â€¢ Magnetfeld:      +z (aus Ebene)")
print("â€¢ Kraft:           -y (nach UNTEN)")
print("â€¢ Grund: q < 0 kehrt das Vorzeichen um!")
print("="*60)
````

## LÃ¶sung zu Aufgabe 2: Zyklotronbewegung â€“ KreisfÃ¶rmige Bahn

### Theoretischer Hintergrund

Die Lorentzkraft wirkt **zentripetal** (zum Mittelpunkt hin), wodurch das Elektron eine **Kreisbahn** beschreibt.

**Zentripetalkraft gleich Lorentzkraft:**
$$F_z = \frac{m v^2}{r} = q v B$$

Daraus folgt:
$$r = \frac{m v}{q B}$$

Die **Winkelgeschwindigkeit** (Zyklotron-Frequenz):
$$\omega = \frac{v}{r} = \frac{q B}{m}$$

Die **Umlaufperiode**:
$$T = \frac{2\pi}{\omega} = \frac{2\pi m}{q B}$$

### VollstÃ¤ndiger Python-Code

````python
import numpy as np
import matplotlib.pyplot as plt

# Physikalische Konstanten
q_electron = 1.602e-19   # Elementarladung Betrag (Coulomb)
m_electron = 9.109e-31   # Elektronenmasse (kg)

# Gegebene Parameter
B = 0.1          # Magnetische Flussdichte (Tesla)
v = 1e6          # Geschwindigkeit (m/s)

# === BERECHNUNGEN ===

# 1. Radius der Kreisbahn
r = (m_electron * v) / (q_electron * B)

# 2. Winkelgeschwindigkeit (Zyklotron-Frequenz)
omega = (q_electron * B) / m_electron

# 3. Umlaufperiode
T = 2 * np.pi / omega

# 4. Frequenz
f = 1 / T

print("="*60)
print("ZYKLOTRONBEWEGUNG â€“ KREISFÃ–RMIGE BAHN")
print("="*60)
print(f"Gegeben:")
print(f"  Magnetische Flussdichte: B = {B} T")
print(f"  Geschwindigkeit:         v = {v:.2e} m/s")
print(f"\nBerechnete GrÃ¶ÃŸen:")
print(f"  Radius der Kreisbahn:    r = {r:.2e} m = {r*1e6:.4f} Î¼m")
print(f"  Zyklotron-Frequenz:      Ï‰ = {omega:.2e} rad/s")
print(f"  Umlaufperiode:           T = {T:.2e} s = {T*1e9:.2f} ns")
print(f"  Frequenz:                f = {f:.2e} Hz = {f/1e9:.2f} GHz")
print("="*60)

# Formel-Verifikation
print(f"\nFormel-Verifikation:")
print(f"  r = mÂ·v/(qÂ·B) = {m_electron:.2e} Â· {v:.2e} / ({q_electron:.2e} Â· {B})")
print(f"    = {r:.2e} m âœ“")
print(f"\n  Ï‰ = qÂ·B/m = {q_electron:.2e} Â· {B} / {m_electron:.2e}")
print(f"    = {omega:.2e} rad/s âœ“")
print(f"\n  T = 2Ï€/Ï‰ = 2Ï€ / {omega:.2e}")
print(f"    = {T:.2e} s âœ“")

# === VISUALISIERUNG ===

# Zeitarray fÃ¼r fÃ¼nf UmlÃ¤ufe
t = np.linspace(0, 5*T, 1000)

# Kreisbahn (parametrische Darstellung)
x = r * np.cos(omega * t)
y = r * np.sin(omega * t)

# Visualisierung
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Plot 1: Kreisbahn
ax = axes[0]
ax.plot(x*1e6, y*1e6, 'b-', linewidth=2.5, label='Elektronenbahn')
ax.plot(x[0]*1e6, y[0]*1e6, 'go', markersize=12, label='Start', zorder=5)
ax.plot(x[-1]*1e6, y[-1]*1e6, 'ro', markersize=12, label='Ende', zorder=5)
ax.plot(0, 0, 'k+', markersize=20, markeredgewidth=2.5, label='Mittelpunkt', zorder=5)

# Radius einzeichnen
ax.plot([0, x[0]*1e6], [0, y[0]*1e6], 'k--', alpha=0.5, linewidth=1.5, label=f'Radius = {r*1e6:.4f} Î¼m')

# Pfeile fÃ¼r Richtung
indices = [250, 500, 750]
for i in indices:
    dx = x[i+10] - x[i]
    dy = y[i+10] - y[i]
    ax.arrow(x[i]*1e6, y[i]*1e6, dx*1e6*5, dy*1e6*5, 
             head_width=0.05, head_length=0.03, fc='gray', ec='gray', alpha=0.6)

ax.set_xlabel('x (Î¼m)', fontsize=12, fontweight='bold')
ax.set_ylabel('y (Î¼m)', fontsize=12, fontweight='bold')
ax.set_title(f'Elektronenbahn im Magnetfeld\n(B = {B} T, v = {v:.1e} m/s)', 
             fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.axis('equal')
ax.legend(fontsize=10, loc='upper right')

# Plot 2: Position und Geschwindigkeit Ã¼ber Zeit
ax = axes[1]
ax_twin = ax.twinx()

# Position (nur x-Komponente)
ax.plot(t*1e9, x*1e6, 'b-', linewidth=2, label='x(t) = rÂ·cos(Ï‰t)')

# Geschwindigkeit (numerisch)
vx = np.gradient(x, t)
vy = np.gradient(y, t)
v_betrag = np.sqrt(vx**2 + vy**2)
ax_twin.plot(t*1e9, v_betrag/1e6, 'r-', linewidth=2, label='|v(t)| (konstant)')
ax_twin.axhline(y=v/1e6, color='r', linestyle='--', alpha=0.5)

ax.set_xlabel('Zeit (ns)', fontsize=12, fontweight='bold')
ax.set_ylabel('Position x (Î¼m)', fontsize=11, color='blue', fontweight='bold')
ax_twin.set_ylabel('Geschwindigkeit (10â¶ m/s)', fontsize=11, color='red', fontweight='bold')
ax.tick_params(axis='y', labelcolor='blue')
ax_twin.tick_params(axis='y', labelcolor='red')
ax.set_title('Position und Geschwindigkeit Ã¼ber Zeit\n(Geschwindigkeit bleibt konstant!)', 
             fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(loc='upper left', fontsize=10)
ax_twin.legend(loc='upper right', fontsize=10)

plt.tight_layout()
plt.show()

print("\n" + "="*60)
print("WICHTIGE ERKENNTNISSE")
print("="*60)
print("1. Der Radius der Kreisbahn:")
print(f"   â€¢ Ist proportional zu v: r âˆ v")
print(f"   â€¢ Ist umgekehrt proportional zu B: r âˆ 1/B")
print(f"   â€¢ HÃ¤ngt von m und q ab")
print("\n2. Die Umlaufperiode T:")
print(f"   â€¢ Ist UNABHÃ„NGIG von der Geschwindigkeit!")
print(f"   â€¢ T = 2Ï€Â·m/(qÂ·B) â‰ˆ {T*1e9:.2f} ns fÃ¼r alle Elektronen")
print(f"   â€¢ Dies ist die Grundlage fÃ¼r Zyklotron-Beschleuniger")
print("\n3. Die Geschwindigkeit:")
print(f"   â€¢ Bleibt in der Magnitude konstant: |v| = {v:.1e} m/s")
print(f"   â€¢ Ã„ndert aber stÃ¤ndig die Richtung (zentripetale Beschleunigung)")
print("="*60)
````

## LÃ¶sung zu Aufgabe 3: Magnetfeld um einen Stromleiter

### Theoretischer Hintergrund

**Biot-Savart-Gesetz** fÃ¼r einen geraden Stromleiter:
$$B(r) = \frac{\mu_0 I}{2\pi r}$$

Die **Feldrichtung** folgt der **Rechten-Hand-Regel**:
- **Daumen**: Stromrichtung
- **Finger**: Feldlinien verlaufen tangential in Richtung der Finger

### VollstÃ¤ndiger Python-Code

````python
import numpy as np
import matplotlib.pyplot as plt

# Parameter
mu_0 = 4 * np.pi * 1e-7  # PermeabilitÃ¤t des Vakuums (H/m)
I = 5                     # StromstÃ¤rke (Ampere)

# === KOORDINATENGITTER ===
x = np.linspace(-4, 4, 150)
y = np.linspace(-4, 4, 150)
X, Y = np.meshgrid(x, y)

# Abstand vom Stromleiter (im Ursprung)
R = np.sqrt(X**2 + Y**2)
R[R < 0.3] = 0.3  # SingularitÃ¤t vermeiden

# === MAGNETFELD-BERECHNUNG ===

# FeldstÃ¤rke nach Biot-Savart
B_magnitude = (mu_0 * I) / (2 * np.pi * R)

# Feldkomponenten (Richtung ist tangential, zirkulÃ¤r)
# B_x = -B * (y/r), B_y = B * (x/r)
Bx = -B_magnitude * (Y / R)
By = B_magnitude * (X / R)

# === NUMERISCHE ÃœBERPRÃœFUNG ===
print("="*60)
print("MAGNETFELD UM EINEN STROMLEITER")
print("="*60)
print(f"StromstÃ¤rke: I = {I} A")
print(f"\nFormel: B(r) = Î¼â‚€Â·I / (2Ï€Â·r)")
print(f"\nBeispiele:")
for r_test in [0.5, 1, 2, 4]:
    B_test = (mu_0 * I) / (2 * np.pi * r_test)
    print(f"  r = {r_test} m:  B = {B_test*1e6:.2f} Î¼T = {B_test*1e9:.2f} nT")
print("="*60)

# === VISUALISIERUNG ===
fig, ax = plt.subplots(figsize=(11, 11))

# Quiver-Plot (Vektorfeld mit Schrittweite 3)
step = 3
quiv = ax.quiver(X[::step, ::step], Y[::step, ::step], 
                  Bx[::step, ::step], By[::step, ::step],
                  np.sqrt(Bx[::step, ::step]**2 + By[::step, ::step]**2),
                  cmap='hot', alpha=0.8, scale=50, scale_units='inches')

# Feldlinien (konzentrische Kreise)
theta = np.linspace(0, 2*np.pi, 100)
radii = [0.5, 1, 2, 3, 4]
for radius in radii:
    x_circle = radius * np.cos(theta)
    y_circle = radius * np.sin(theta)
    ax.plot(x_circle, y_circle, 'k--', alpha=0.2, linewidth=1)
    # Etikett fÃ¼r Radius
    if radius <= 3:
        ax.text(radius, -0.3, f'{radius}m', fontsize=9, ha='center', alpha=0.5)

# Stromleiter (Ursprung)
ax.plot(0, 0, 'ro', markersize=15, label='Stromleiter', zorder=5)
ax.text(0.2, 0.2, 'âŠ™\nStrom heraus\n(+z-Richtung)', fontsize=10, 
        bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.7))

# Rechte-Hand-Regel Annotation
ax.annotate('Daumen: Strom\nFinger: B-Feld (zirkulÃ¤r)', 
            xy=(1.5, 3), xytext=(2.5, 3.5),
            fontsize=10,
            bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.7),
            arrowprops=dict(arrowstyle='->', color='blue', lw=1.5))

# Richtungspfeile entlang eines Kreises
circle_radius = 2
angles = [0, np.pi/2, np.pi, 3*np.pi/2]
for angle in angles:
    x_pos = circle_radius * np.cos(angle)
    y_pos = circle_radius * np.sin(angle)
    # Tangentialrichtung
    dx = -np.sin(angle) * 0.3
    dy = np.cos(angle) * 0.3
    ax.arrow(x_pos, y_pos, dx, dy, head_width=0.2, head_length=0.1, 
             fc='red', ec='red', alpha=0.6, linewidth=2)

# Colorbar
cbar = plt.colorbar(quiv, ax=ax, label='Magnetische Flussdichte (T)')

ax.set_xlabel('x (m)', fontsize=12, fontweight='bold')
ax.set_ylabel('y (m)', fontsize=12, fontweight='bold')
ax.set_title(f'Magnetfeld um einen geraden Stromleiter (I = {I} A)\nRechte-Hand-Regel: Daumen in Stromrichtung â†’ Feldlinien zirkulÃ¤r', 
             fontsize=13, fontweight='bold')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.grid(True, alpha=0.2)
ax.axis('equal')
ax.legend(fontsize=11, loc='upper left')

plt.tight_layout()
plt.show()

# === ZUSATZ: 3D-PLOT ===
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# 3D-Plot der FeldstÃ¤rke
surf = ax.plot_surface(X, Y, B_magnitude*1e9, cmap='hot', alpha=0.8)

ax.set_xlabel('x (m)', fontsize=11)
ax.set_ylabel('y (m)', fontsize=11)
ax.set_zlabel('B-Feld (nT)', fontsize=11)
ax.set_title('Magnetische Flussdichte B(r) um einen Stromleiter\nB(r) = Î¼â‚€Â·I/(2Ï€Â·r)', 
             fontsize=12, fontweight='bold')

cbar = plt.colorbar(surf, ax=ax, label='B (nT)', pad=0.1)

plt.show()
````

## LÃ¶sung zu Aufgabe 4: Faradaysches Induktionsgesetz

### Theoretischer Hintergrund

**Faradaysches Induktionsgesetz:**
$$\mathcal{E} = -\frac{d\Phi}{dt}$$

FÃ¼r eine rotierende Spule mit konstanter Winkelgeschwindigkeit:
$$\Phi(t) = N \cdot B \cdot A \cdot \cos(\omega t)$$

Daraus folgt die induzierte Spannung:
$$\mathcal{E}(t) = -\frac{d\Phi}{dt} = N \cdot B \cdot A \cdot \omega \cdot \sin(\omega t)$$

**Wichtige GrÃ¶ÃŸen:**
- **Maximale Spannung:** $\mathcal{E}_{max} = N \cdot B \cdot A \cdot \omega$
- **RMS-Spannung** (Effektivwert): $\mathcal{E}_{rms} = \frac{\mathcal{E}_{max}}{\sqrt{2}}$

### VollstÃ¤ndiger Python-Code

````python
import numpy as np
import matplotlib.pyplot as plt

# === PARAMETER ===
B = 1.0          # Magnetische Flussdichte (Tesla)
A = 0.05 * 0.05  # SpulenflÃ¤che (0.05m Ã— 0.05m)
N = 200          # Anzahl der Windungen
f = 50           # Frequenz (Hz) ~ 50 Hz Netzfrequenz in Europa
omega = 2 * np.pi * f  # Winkelgeschwindigkeit (rad/s)

# Periodendauer
T_periode = 1 / f

# === BERECHNUNG ===

# Zeit Ã¼ber zwei Perioden
t = np.linspace(0, 2*T_periode, 1000)

# 1. Rotationswinkel
theta = omega * t

# 2. Magnetischer Fluss
Phi = N * B * A * np.cos(theta)

# 3. Induzierte Spannung (analytisch)
epsilon = N * B * A * omega * np.sin(theta)

# 4. Numerische ÃœberprÃ¼fung (Ableitung)
# epsilon_numeric = -np.gradient(Phi, t)  # Sollte -epsilon entsprechen

# 5. Maximale und RMS-Spannung
epsilon_max = N * B * A * omega
epsilon_rms = epsilon_max / np.sqrt(2)

# === AUSGABE ===
print("="*60)
print("FARADAYSCHES INDUKTIONSGESETZ")
print("="*60)
print(f"Spulenparameter:")
print(f"  Anzahl Windungen:          N = {N}")
print(f"  Magnetische Flussdichte:   B = {B} T")
print(f"  SpulenflÃ¤che:              A = {A} mÂ² = {A*1e4} cmÂ²")
print(f"  Frequenz:                  f = {f} Hz")
print(f"  Winkelgeschwindigkeit:     Ï‰ = {omega:.2f} rad/s = 2Ï€Â·f")
print(f"\nMagnetischer Fluss:")
print(f"  Maximum:                   Î¦_max = NÂ·BÂ·A = {N*B*A:.4f} Wb")
print(f"  Minimum:                   Î¦_min = 0 Wb (bei Î¸ = 90Â°)")
print(f"\nInduzierte Spannung:")
print(f"  Maximum (Peak):            Îµ_max = NÂ·BÂ·AÂ·Ï‰ = {epsilon_max:.2f} V")
print(f"  Effektivwert (RMS):        Îµ_rms = Îµ_max/âˆš2 = {epsilon_rms:.2f} V")
print(f"\nVergleich mit Haushaltsanlage:")
print(f"  Typische Haushaltsspannung in Europa: 230 V (RMS)")
print(f"  Diese Spule wÃ¼rde erzeugen: {epsilon_rms:.2f} V (RMS)")
print("="*60)

# Formel-Verifikation
print(f"\nFormel-Verifizierung:")
print(f"  Îµ_max = NÂ·BÂ·AÂ·Ï‰")
print(f"        = {N} Ã— {B} Ã— {A} Ã— {omega:.2f}")
print(f"        = {epsilon_max:.2f} V âœ“")
print(f"\n  Îµ_rms = Îµ_max / âˆš2")
print(f"        = {epsilon_max:.2f} / âˆš2")
print(f"        = {epsilon_rms:.2f} V âœ“")

# === VISUALISIERUNG ===
fig, axes = plt.subplots(3, 1, figsize=(13, 11))

# Plot 1: Rotationswinkel und Feldorientierung
ax = axes[0]
theta_degrees = np.degrees(theta)
ax.plot(t*1000, theta_degrees, 'purple', linewidth=2.5)
ax.axhline(y=0, color='k', linestyle='-', alpha=0.2)
ax.axhline(y=90, color='k', linestyle='--', alpha=0.3, label='Max. FeldÃ¤nderung (dÎ¦/dt max)')
ax.axhline(y=180, color='k', linestyle='-', alpha=0.2)
ax.fill_between(t*1000, 0, theta_degrees, where=(theta_degrees <= 90), alpha=0.2, color='orange')
ax.set_ylabel('Rotationswinkel Î¸ (Â°)', fontsize=12, fontweight='bold')
ax.set_title('Rotation der Spule: Î¸(t) = Ï‰Â·t', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=10)

# Plot 2: Magnetischer Fluss
ax = axes[1]
ax.plot(t*1000, Phi, 'g-', linewidth=2.5, label='Î¦(t) = NÂ·BÂ·AÂ·cos(Î¸)')
ax.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax.axhline(y=N*B*A, color='green', linestyle='--', alpha=0.5, label=f'Î¦_max = {N*B*A:.4f} Wb')
ax.axhline(y=-N*B*A, color='green', linestyle='--', alpha=0.5)
ax.fill_between(t*1000, 0, Phi, where=(Phi >= 0), alpha=0.3, color='green', label='Fluss steigt')
ax.fill_between(t*1000, 0, Phi, where=(Phi < 0), alpha=0.3, color='red', label='Fluss fÃ¤llt')
ax.set_ylabel('Magnetischer Fluss (Wb)', fontsize=12, fontweight='bold')
ax.set_title('Fluss durch Spule: Î¦(t) = NÂ·BÂ·AÂ·cos(Ï‰t)\n(Maximum, wenn Spule senkrecht zum Feld, Null wenn parallel)', 
             fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=10, loc='upper right')

# Plot 3: Induzierte Spannung
ax = axes[2]
ax.plot(t*1000, epsilon, 'r-', linewidth=2.5, label='Îµ(t) = NÂ·BÂ·AÂ·Ï‰Â·sin(Î¸)')
ax.axhline(y=0, color='k', linestyle='-', alpha=0.3)
ax.axhline(y=epsilon_max, color='red', linestyle='--', alpha=0.5, label=f'Îµ_max = {epsilon_max:.2f} V')
ax.axhline(y=-epsilon_max, color='red', linestyle='--', alpha=0.5)
ax.axhline(y=epsilon_rms, color='blue', linestyle='--', alpha=0.5, linewidth=2, label=f'Îµ_rms = {epsilon_rms:.2f} V (Effektivwert)')
ax.axhline(y=-epsilon_rms, color='blue', linestyle='--', alpha=0.5, linewidth=2)
ax.fill_between(t*1000, 0, epsilon, where=(epsilon >= 0), alpha=0.3, color='red', label='Positive Spannung')
ax.fill_between(t*1000, 0, epsilon, where=(epsilon < 0), alpha=0.3, color='blue', label='Negative Spannung')
ax.set_xlabel('Zeit (ms)', fontsize=12, fontweight='bold')
ax.set_ylabel('Induzierte Spannung (V)', fontsize=12, fontweight='bold')
ax.set_title('Induzierte Spannung: Îµ(t) = -dÎ¦/dt = NÂ·BÂ·AÂ·Ï‰Â·sin(Ï‰t)', fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3)
ax.legend(fontsize=10, loc='upper right')

plt.tight_layout()
plt.show()

# === ZUSATZ: Beziehung zwischen Î¦ und Îµ ===
print("\n" + "="*60)
print("ZUSAMMENHANG ZWISCHEN FLUSS UND SPANNUNG")
print("="*60)
print("â€¢ Wenn Î¦ maximal ist (cos(Î¸) = 1):   dÎ¦/dt = 0  â†’  Îµ = 0")
print("â€¢ Wenn Î¦ minimal ist (cos(Î¸) = -1):  dÎ¦/dt = 0  â†’  Îµ = 0")
print("â€¢ Wenn Î¦ = 0 (cos(Î¸) = 0):           dÎ¦/dt = max  â†’  Îµ = Â±max")
print("\nDies ist das Grundprinzip der Wechselstrom-Erzeugung:")
print("â€¢ Stromgenerator (Dynamo): Spule rotiert im Magnetfeld")
print("â€¢ Transformator: Wechselndes Magnetfeld induziert Spannung")
print("â€¢ Drahtlose EnergieÃ¼bertragung (induktives Laden)")
print("="*60)

# === BONUS: Vergleich mit anderen Frequenzen ===
print("\nVerkleinerter Vergleich (bei anderen Frequenzen):")
frequencies = [50, 60, 100, 150]
for f_test in frequencies:
    omega_test = 2 * np.pi * f_test
    epsilon_max_test = N * B * A * omega_test
    epsilon_rms_test = epsilon_max_test / np.sqrt(2)
    print(f"  f = {f_test} Hz:  Îµ_max = {epsilon_max_test:.2f} V,  Îµ_rms = {epsilon_rms_test:.2f} V")
````
