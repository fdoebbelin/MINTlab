## üî¨ L√∂sung zu Aufgabe 1: Reflexionsgesetz ‚Äì Interaktive Parametervariationen

### Vollst√§ndige L√∂sung

```python
import matplotlib.pyplot as plt
import numpy as np

# Einfallswinkel (in Grad) ‚Äì kann beliebig ge√§ndert werden
theta_einfallswinkel = 35  # Beispiel: Versuche auch 45¬∞, 60¬∞, etc.

# Figure erstellen
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

# ============ SPIEGEL zeichnen ============
spiegel_x = [-3, 3]
spiegel_y = [0, 0]
ax.plot(spiegel_x, spiegel_y, 'b-', linewidth=3, label='Spiegel')

# ============ NORMALE (Lotrechte) zeichnen ============
# Die Normale steht senkrecht auf dem Spiegel
ax.plot([0, 0], [-0.5, 2.5], 'gray', linewidth=1.5, linestyle='--', label='Normale')

# ============ EINFALLENDEN STRAHL berechnen und zeichnen ============
# Der einfallende Strahl kommt von links oben zur Spiegelfl√§che
# F√ºr einen Winkel theta zur Normalen:
#   x-Komponente: -sin(theta) (negativ = nach links)
#   y-Komponente: cos(theta) (positiv = nach oben)

# Berechne die Anfangsposition (von wo kommt der Strahl?)
# Wir wollen, dass der Strahl 2 Einheiten vor dem Spiegel startet
strahl_einfall_start_x = -2 * np.sin(np.radians(theta_einfallswinkel))
strahl_einfall_start_y = 2 * np.cos(np.radians(theta_einfallswinkel))

# Der Strahl endet am Spiegelpunkt (0, 0)
strahl_einfall_end_x = 0
strahl_einfall_end_y = 0

# Zeichne den einfallenden Strahl (ROT)
ax.plot([strahl_einfall_start_x, strahl_einfall_end_x], 
        [strahl_einfall_start_y, strahl_einfall_end_y], 
        'r-', linewidth=2.5, label='Einfallender Strahl')

# ============ REFLEKTIERTEN STRAHL berechnen und zeichnen ============
# Das Reflexionsgesetz besagt: Œ∏_einfallswinkel = Œ∏_reflexionswinkel
# Der reflektierte Strahl hat also den gleichen Winkel, aber auf der anderen Seite der Normalen

# Der reflektierte Strahl startet am Spiegelpunkt (0, 0)
# und geht 2 Einheiten weit nach rechts oben
strahl_reflektiert_start_x = 0
strahl_reflektiert_start_y = 0

strahl_reflektiert_end_x = 2 * np.sin(np.radians(theta_einfallswinkel))
strahl_reflektiert_end_y = 2 * np.cos(np.radians(theta_einfallswinkel))

# Zeichne den reflektierten Strahl (GR√úN)
ax.plot([strahl_reflektiert_start_x, strahl_reflektiert_end_x], 
        [strahl_reflektiert_start_y, strahl_reflektiert_end_y], 
        'g-', linewidth=2.5, label='Reflektierter Strahl')

# ============ WINKEL beschriften ============
# Einfallswinkel
ax.text(-0.8, 1.2, f'Œ∏‚ÇÅ = {theta_einfallswinkel}¬∞', fontsize=12, color='red', fontweight='bold')

# Reflexionswinkel
ax.text(0.5, 1.2, f'Œ∏‚ÇÇ = {theta_einfallswinkel}¬∞', fontsize=12, color='green', fontweight='bold')

# ============ ACHSEN-EINSTELLUNGEN ============
ax.set_xlim(-3.5, 3.5)
ax.set_ylim(-1, 3)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.set_xlabel('Position x (beliebige Einheit)', fontsize=11)
ax.set_ylabel('Position y (beliebige Einheit)', fontsize=11)
ax.set_title('Reflexionsgesetz: Einfallswinkel = Reflexionswinkel', fontsize=13, fontweight='bold')
ax.legend(loc='upper right', fontsize=10)

plt.tight_layout()
plt.show()

# ============ √úBERPR√úFUNG DES REFLEXIONSGESETZES ============
print("=" * 50)
print("√úBERPR√úFUNG DES REFLEXIONSGESETZES")
print("=" * 50)
print(f"Einfallswinkel Œ∏‚ÇÅ: {theta_einfallswinkel}¬∞")
print(f"Reflexionswinkel Œ∏‚ÇÇ: {theta_einfallswinkel}¬∞")
print(f"Sind sie gleich? {theta_einfallswinkel == theta_einfallswinkel}")
print("‚úì Reflexionsgesetz erf√ºllt!")
print("=" * 50)
```

### Erkl√§rungen zum Code

**Warum funktioniert das?**
1. Die Normale zeigt senkrecht vom Spiegel nach oben (in y-Richtung)
2. Der Einfallswinkel wird **relativ zur Normalen** gemessen
3. Der einfallende Strahl kommt von links oben (x < 0, y > 0)
4. Der reflektierte Strahl geht nach rechts oben (x > 0, y > 0)
5. Beide bilden den gleichen Winkel mit der Normalen

**Variationen zum Experimentieren:**

```python
# Probiere verschiedene Winkel:
theta_einfallswinkel = 0    # Strahl f√§llt senkrecht ein
theta_einfallswinkel = 45   # 45¬∞ Winkel
theta_einfallswinkel = 80   # Sehr flacher Winkel (fast parallel zum Spiegel)
```

## üî¨ L√∂sung zu Aufgabe 2: Brechungsgesetz ‚Äì Materialwechsel

### Vollst√§ndige L√∂sung

```python
import matplotlib.pyplot as plt
import numpy as np

# ============ MATERIAL-PARAMETER ============
# Luft hat Brechungsindex n ‚âà 1,0
# Wasser hat Brechungsindex n ‚âà 1,33
# Glas hat Brechungsindex n ‚âà 1,5

n1 = 1.0   # Luft (oben)
n2 = 1.33  # Wasser (unten)

# Einfallswinkel in Luft
theta1_grad = 45  # Grad

# ============ BRECHUNGSGESETZ ANWENDEN ============
# Snellius' Gesetz: n‚ÇÅ sin(Œ∏‚ÇÅ) = n‚ÇÇ sin(Œ∏‚ÇÇ)
# Umgestellt nach Œ∏‚ÇÇ: Œ∏‚ÇÇ = arcsin((n‚ÇÅ/n‚ÇÇ) sin(Œ∏‚ÇÅ))

# 1. Konvertiere Einfallswinkel zu Radianten
theta1_rad = np.radians(theta1_grad)

# 2. Berechne sin(Œ∏‚ÇÅ)
sin_theta1 = np.sin(theta1_rad)

# 3. Berechne sin(Œ∏‚ÇÇ) mit Snellius' Gesetz
sin_theta2 = (n1 / n2) * sin_theta1

# 4. Berechne Œ∏‚ÇÇ mit arcsin
theta2_rad = np.arcsin(sin_theta2)

# 5. Konvertiere zur√ºck zu Grad
theta2_grad = np.degrees(theta2_rad)

# ============ FIGUR ERSTELLEN ============
fig, ax = plt.subplots(1, 1, figsize=(10, 8))

# ============ MEDIEN mit Farben kennzeichnen ============
# Oben: Luft (gelb)
ax.fill_between([-4, 4], 0, 3, alpha=0.2, color='yellow', label='Medium 1: Luft (n = 1.0)')

# Unten: Wasser (blau)
ax.fill_between([-4, 4], -3, 0, alpha=0.2, color='blue', label='Medium 2: Wasser (n = 1.33)')

# ============ GRENZFL√ÑCHE zeichnen ============
grenzflaeche_x = [-4, 4]
grenzflaeche_y = [0, 0]
ax.plot(grenzflaeche_x, grenzflaeche_y, 'k-', linewidth=3, label='Grenzfl√§che')

# ============ NORMALE zeichnen ============
ax.plot([0, 0], [-3, 3], 'gray', linewidth=1.5, linestyle='--', label='Normale')
ax.text(0.2, -2.8, 'Normale', fontsize=10, color='gray', fontweight='bold')

# ============ EINFALLENDEN STRAHL zeichnen ============
# Der Strahl kommt von links oben und trifft die Grenzfl√§che bei (0, 0)
strahl_einfall_start_x = -3 * np.sin(np.radians(theta1_grad))
strahl_einfall_start_y = 3 * np.cos(np.radians(theta1_grad))

strahl_einfall_end_x = 0
strahl_einfall_end_y = 0

ax.plot([strahl_einfall_start_x, strahl_einfall_end_x], 
        [strahl_einfall_start_y, strahl_einfall_end_y], 
        'r-', linewidth=2.5, label=f'Einfallender Strahl (Œ∏‚ÇÅ = {theta1_grad}¬∞)')

# ============ GEBROCHENEN STRAHL zeichnen ============
# Der Strahl wird nach der Grenzfl√§che gebrochen und geht in Wasser weiter
strahl_gebrochen_start_x = 0
strahl_gebrochen_start_y = 0

strahl_gebrochen_end_x = 3 * np.sin(np.radians(theta2_grad))
strahl_gebrochen_end_y = -3 * np.cos(np.radians(theta2_grad))

ax.plot([strahl_gebrochen_start_x, strahl_gebrochen_end_x], 
        [strahl_gebrochen_start_y, strahl_gebrochen_end_y], 
        'orange', linewidth=2.5, label=f'Gebrochener Strahl (Œ∏‚ÇÇ = {theta2_grad:.1f}¬∞)')

# ============ WINKEL beschriften ============
ax.text(0.5, 1.5, f'Œ∏‚ÇÅ = {theta1_grad}¬∞', fontsize=12, color='red', fontweight='bold')
ax.text(0.5, -1.2, f'Œ∏‚ÇÇ = {theta2_grad:.1f}¬∞', fontsize=12, color='orange', fontweight='bold')

# ============ ACHSEN-EINSTELLUNGEN ============
ax.set_xlim(-4, 4)
ax.set_ylim(-3, 3)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
ax.set_xlabel('Position x (beliebige Einheit)', fontsize=11)
ax.set_ylabel('Position y (beliebige Einheit)', fontsize=11)
ax.set_title("Brechungsgesetz (Snellius' Gesetz): n‚ÇÅ¬∑sin(Œ∏‚ÇÅ) = n‚ÇÇ¬∑sin(Œ∏‚ÇÇ)", 
             fontsize=13, fontweight='bold')
ax.legend(loc='upper left', fontsize=10)

plt.tight_layout()
plt.show()

# ============ √úBERPR√úFUNG VON SNELLIUS' GESETZ ============
print("=" * 60)
print("√úBERPR√úFUNG VON SNELLIUS' GESETZ")
print("=" * 60)
print(f"Einfallswinkel Œ∏‚ÇÅ (in Luft):  {theta1_grad}¬∞")
print(f"Brechungswinkel Œ∏‚ÇÇ (in Wasser): {theta2_grad:.2f}¬∞")
print(f"\nBrechungsindizes:")
print(f"  n‚ÇÅ (Luft):   {n1}")
print(f"  n‚ÇÇ (Wasser): {n2}")
print(f"\nLinke Seite:  n‚ÇÅ ¬∑ sin(Œ∏‚ÇÅ) = {n1} √ó sin({theta1_grad}¬∞) = {n1 * sin_theta1:.4f}")
print(f"Rechte Seite: n‚ÇÇ ¬∑ sin(Œ∏‚ÇÇ) = {n2} √ó sin({theta2_grad:.2f}¬∞) = {n2 * np.sin(theta2_rad):.4f}")
print(f"Differenz: {abs(n1 * sin_theta1 - n2 * np.sin(theta2_rad)):.6f}")
print(f"‚úì Snellius' Gesetz erf√ºllt (Differenz ‚âà 0)!")
print("=" * 60)

# ============ BEOBACHTUNGEN ============
print("\nBEOBACHTUNGEN:")
print(f"‚Ä¢ Das Licht wird ZUM LOT HIN gebrochen")
print(f"‚Ä¢ Einfallswinkel: {theta1_grad}¬∞ > Brechungswinkel: {theta2_grad:.2f}¬∞")
print(f"‚Ä¢ Das liegt daran, dass Wasser optisch dichter ist als Luft")
print(f"‚Ä¢ Allgemein: Bei dichterem Medium = st√§rkere Brechung = kleinerer Winkel")
```

### Erkl√§rungen zum Code

**Warum wird Licht gebrochen?**
- Licht bewegt sich in verschiedenen Medien mit unterschiedlichen Geschwindigkeiten
- In Wasser ist die Lichtgeschwindigkeit langsamer als in Luft
- Diese Geschwindigkeits√§nderung f√ºhrt zur Richtungs√§nderung

**Die Faustregel:**
- Wenn n‚ÇÇ > n‚ÇÅ (optisch dichter): Brechung zum Lot hin (Winkel wird kleiner)
- Wenn n‚ÇÇ < n‚ÇÅ (optisch d√ºnner): Brechung vom Lot weg (Winkel wird gr√∂√üer)

**Variationen zum Experimentieren:**

```python
# Probiere andere Materialien:
n2 = 1.5   # Glas (noch st√§rkere Brechung)
n2 = 2.4   # Diamant (extreme Brechung!)

# Probiere andere Winkel:
theta1_grad = 30   # Flacherer Einfallswinkel
theta1_grad = 60   # Steilerer Einfallswinkel
theta1_grad = 85   # Fast parallel zur Grenzfl√§che
```

## üî¨ L√∂sung zu Aufgabe 3: Linsenschleiferformel ‚Äì Bildberechnung

### Vollst√§ndige L√∂sung

```python
import matplotlib.pyplot as plt
import numpy as np

# ============ LINSENPARAMETER ============
f = 2.0  # Brennweite der Sammellinse (positiv = Sammellinse, negativ = Zerstreuungslinse)
g = 5.0  # Gegenstandsweite (Abstand vom Gegenstand zur Linse)

# ============ LINSENSCHLEIFERFORMEL anwenden ============
# Formel: 1/f = 1/g + 1/b
# Umgestellt nach b: b = 1 / (1/f - 1/g)

# Berechne die Bildweite
b = 1 / (1/f - 1/g)

# Berechne die Vergr√∂√üerung
# V = b/g (negativ = invertiertes Bild, positiv = aufrechtes Bild)
V = b / g

# ============ FIGUR ERSTELLEN ============
fig, ax = plt.subplots(1, 1, figsize=(12, 6))

# ============ LINSE zeichnen ============
# Linse ist eine vertikale Linie in der Mitte
linse_hoehe = 2.5
linse_y = np.linspace(-linse_hoehe, linse_hoehe, 100)
linse_x = [0] * len(linse_y)
ax.plot(linse_x, linse_y, 'b-', linewidth=3, label='Sammellinse (konvex)')

# ============ BRENNPUNKTE markieren ============
# Die beiden Brennpunkte sind bei x = f und x = -f
ax.plot([f, -f], [0, 0], 'ro', markersize=8, label=f'Brennpunkte (f = {f})')

# Brennpunkt-Labels
ax.text(f, -0.4, f"F' (f = {f})", fontsize=10, ha='center', color='red')
ax.text(-f, -0.4, f"F (f = {-f})", fontsize=10, ha='center', color='red')

# ============ OPTISCHE ACHSE ============
ax.axhline(y=0, color='k', linewidth=0.5, linestyle='-', alpha=0.3)

# ============ GEGENSTAND zeichnen ============
# Gegenstand ist ein gr√ºner Pfeil bei x = -g
gegenstand_hoehe = 1.5
ax.arrow(-g, 0, 0, gegenstand_hoehe, head_width=0.2, head_length=0.15, 
         fc='green', ec='green', linewidth=2, label='Gegenstand')
ax.plot([-g], [0], 'go', markersize=8)
ax.text(-g, gegenstand_hoehe + 0.3, f'Gegenstand\n(g = {g})', fontsize=10, 
        ha='center', color='green', fontweight='bold')

# ============ BILD zeichnen ============
# Bild ist ein blauer Pfeil bei x = b
# Die H√∂he des Bildes ist: H√∂he_Gegenstand √ó V
bild_hoehe = gegenstand_hoehe * V
# Wenn V < 0, ist das Bild invertiert (Pfeil zeigt nach unten)
ax.arrow(b, 0, 0, -bild_hoehe, head_width=0.2, head_length=0.15, 
         fc='purple', ec='purple', linewidth=2, label='Bild')
ax.plot([b], [0], 'mo', markersize=8)
ax.text(b, -bild_hoehe - 0.5, f'Bild\n(b = {b:.2f})\nV = {V:.2f}', fontsize=10, 
        ha='center', color='purple', fontweight='bold')

# ============ KONSTRUKTIONSSTRAHLEN zeichnen ============
# Diese Strahlen zeigen, wie das Bild entsteht

# Strahl 1: Parallel zur Achse ‚Üí durch Brennpunkt
ax.arrow(-g, gegenstand_hoehe, g-0.3, 0, head_width=0.1, head_length=0.1, 
         fc='red', ec='red', alpha=0.5, linestyle='--', linewidth=1.5)
# Nach der Linse durch Brennpunkt
if f != 0:
    steigung_strahl1 = -gegenstand_hoehe / f
    y_end_strahl1 = steigung_strahl1 * (b - 0)
    ax.arrow(0.05, gegenstand_hoehe, b-0.1, y_end_strahl1 - gegenstand_hoehe, 
             head_width=0.1, head_length=0.1, fc='red', ec='red', alpha=0.7, linewidth=1.5)

# Strahl 2: Durch Linsenmitte (geht ungebrochen weiter)
ax.arrow(-g, gegenstand_hoehe, b+g-0.2, -bild_hoehe-gegenstand_hoehe, 
         head_width=0.1, head_length=0.1, fc='green', ec='green', alpha=0.5, 
         linestyle='--', linewidth=1.5)

# ============ ABSTANDSMARKIERUNGEN ============
ax.plot([0, 0], [-2.8, -2.5], 'k-', linewidth=1)
ax.plot([-g, -g], [-2.8, -2.5], 'k-', linewidth=1)
ax.plot([-g, 0], [-2.65, -2.65], 'k-', linewidth=1)
ax.text(-g/2, -2.9, f'g = {g}', fontsize=9, ha='center', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

ax.plot([0, 0], [2.5, 2.8], 'k-', linewidth=1)
ax.plot([b, b], [2.5, 2.8], 'k-', linewidth=1)
ax.plot([0, b], [2.65, 2.65], 'k-', linewidth=1)
ax.text(b/2, 2.9, f'b = {b:.2f}', fontsize=9, ha='center', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

# ============ ACHSEN-EINSTELLUNGEN ============
ax.set_xlim(-7, 7)
ax.set_ylim(-3.5, 3.5)
ax.set_aspect('equal')
ax.grid(True, alpha=0.2)
ax.set_xlabel('Position (beliebige Einheit)', fontsize=11)
ax.set_ylabel('H√∂he (beliebige Einheit)', fontsize=11)
ax.set_title(f'Linsenschleiferformel: 1/f = 1/g + 1/b | Vergr√∂√üerung V = {V:.2f}', 
             fontsize=13, fontweight='bold')
ax.legend(loc='upper right', fontsize=9)

plt.tight_layout()
plt.show()

# ============ √úBERPR√úFUNG DER LINSENSCHLEIFERFORMEL ============
print("=" * 60)
print("√úBERPR√úFUNG DER LINSENSCHLEIFERFORMEL")
print("=" * 60)
print(f"Parameter:")
print(f"  Brennweite f: {f}")
print(f"  Gegenstandsweite g: {g}")
print(f"  Berechnete Bildweite b: {b:.2f}")
print(f"  Vergr√∂√üerung V: {V:.2f}")

print(f"\nLinsenschleiferformel-Test:")
print(f"  1/f = {1/f:.4f}")
print(f"  1/g + 1/b = {1/g:.4f} + {1/b:.4f} = {1/g + 1/b:.4f}")
print(f"  Differenz: {abs(1/f - (1/g + 1/b)):.8f}")
if abs(1/f - (1/g + 1/b)) < 0.0001:
    print("  ‚úì Linsenschleiferformel erf√ºllt!")
else:
    print("  ‚úó Fehler in der Berechnung!")

print(f"\nBildtyp:")
if b > 0:
    print(f"  ‚Ä¢ Bildweite b = {b:.2f} > 0 ‚Üí REELLES BILD")
    print(f"  ‚Ä¢ Kann auf einem Schirm projiziert werden")
else:
    print(f"  ‚Ä¢ Bildweite b = {b:.2f} < 0 ‚Üí VIRTUELLES BILD")
    print(f"  ‚Ä¢ Nur durch die Linse sichtbar (z.B. Lupe)")

print(f"\nVergr√∂√üerung:")
print(f"  ‚Ä¢ V = {V:.2f}")
if abs(V) > 1:
    print(f"  ‚Ä¢ Bild ist {abs(V):.2f}√ó VERGR√ñSSERT")
elif abs(V) < 1:
    print(f"  ‚Ä¢ Bild ist {1/abs(V):.2f}√ó VERKLEINERT")
else:
    print(f"  ‚Ä¢ Bild hat gleiche Gr√∂√üe wie Gegenstand")

if V < 0:
    print(f"  ‚Ä¢ Bild ist INVERTIERT (umgekehrt)")
else:
    print(f"  ‚Ä¢ Bild ist AUFRECHT")

print("=" * 60)
```

### Erkl√§rungen zum Code

**Die Linsenschleiferformel:**
$$\frac{1}{f} = \frac{1}{g} + \frac{1}{b}$$

Diese Formel verkn√ºpft die drei wichtigsten Gr√∂√üen einer Linse:
- **f** = Brennweite (Eigenschaft der Linse)
- **g** = Gegenstandsweite
- **b** = Bildweite

**Interpretation der Ergebnisse:**
- **b > 0:** Reelles Bild (Projektion m√∂glich)
- **b < 0:** Virtuelles Bild (nur durch die Linse sichtbar)
- **|V| > 1:** Vergr√∂√üerung
- **|V| < 1:** Verkleinerung
- **V < 0:** Invertiertes Bild

**Variationen zum Experimentieren:**

```python
# Sehr nah an der Linse (Lupe-Effekt)
g = 1.5  # Sehr nah!

# Weit weg von der Linse (Kamera-Effekt)
g = 100

# Unterschiedliche Brennweiten
f = 0.5  # Sehr starke Linse
f = 10   # Schwache Linse
f = -2   # Zerstreuungslinse (negativ)
```

## üî¨ L√∂sung zu Aufgabe 4: Ray-Tracing ‚Äì Mehrere Strahlen

### Vollst√§ndige L√∂sung

```python
import matplotlib.pyplot as plt
import numpy as np

# ============ PARAMETER ============
f = 2.0  # Brennweite
g = 4.0  # Gegenstandsweite
num_strahlen = 7  # Anzahl der Lichtstrahlen

# ============ BILDWEITE BERECHNEN ============
b = 1 / (1/f - 1/g)

# ============ FIGUR ERSTELLEN ============
fig, ax = plt.subplots(1, 1, figsize=(12, 8))

# ============ LINSE zeichnen ============
linse_hoehe = 2.5
linse_y = np.linspace(-linse_hoehe, linse_hoehe, 100)
linse_x = [0] * len(linse_y)
ax.plot(linse_x, linse_y, 'b-', linewidth=3, label='Sammellinse')

# ============ BRENNPUNKTE ============
ax.plot([f, -f], [0, 0], 'ro', markersize=8, label=f'Brennpunkte (f = {f})')

# ============ OPTISCHE ACHSE ============
ax.axhline(y=0, color='k', linewidth=0.5, linestyle='-', alpha=0.3)

# ============ RAY-TRACING: MEHRERE STRAHLEN ============
# Wir zeichnen num_strahlen Strahlen in verschiedenen H√∂hen

# H√∂hen der Strahlen: von -linse_hoehe bis +linse_hoehe
strahlhoehen = np.linspace(-linse_hoehe, linse_hoehe, num_strahlen)

# Farben f√ºr die Strahlen (von rot zu blau Farbverlauf)
farben = plt.cm.rainbow(np.linspace(0, 1, num_strahlen))

print(f"Ray-Tracing mit {num_strahlen} Strahlen:")
print(f"Brennweite f = {f}")
print(f"Gegenstandsweite g = {g}")
print(f"Bildweite b = {b:.2f}\n")

# Schleife √ºber alle Strahlen
for i, h in enumerate(strahlhoehen):
    
    # ============ VOR DER LINSE: Strahl parallel zur Achse ============
    # Der Strahl kommt von links, parallel zur optischen Achse
    # Von x = -g zur Linse bei x = 0
    strahl_vor_x = [-g, 0]
    strahl_vor_y = [h, h]  # Parallel = konstante y-Koordinate
    
    ax.plot(strahl_vor_x, strahl_vor_y, color=farben[i], alpha=0.5, linestyle='--', linewidth=1.5)
    
    # ============ NACH DER LINSE: Strahl durch Brennpunkt ============
    # Das Wichtigste Prinzip: Ein Strahl, der parallel einf√§llt,
    # wird nach der Linse durch den Brennpunkt F' bei x = f gelenkt
    
    # Aber wir m√ºssen aufpassen: Wenn h = 0 (auf der Achse),
    # geht der Strahl einfach geradeaus durch die Linse
    
    if abs(h) < 0.01:  # h ‚âà 0: Strahl auf der optischen Achse
        # Strahl geht durch den Brennpunkt und dann weiter
        strahl_nach_x = [0, b]
        strahl_nach_y = [0, 0]
    else:
        # Der Strahl geht durch den Brennpunkt bei (f, 0)
        # Berechne die Steigung: m = (0 - h) / (f - 0) = -h/f
        steigung = -h / f
        
        # Endposition: Der Strahl geht bis x ‚âà b
        x_end = b
        y_end = steigung * (x_end - 0)  # y = m * (x - x_brennpunkt)
        
        strahl_nach_x = [0, x_end]
        strahl_nach_y = [h, y_end]
    
    ax.plot(strahl_nach_x, strahl_nach_y, color=farben[i], alpha=0.8, linewidth=2)
    
    if i == 0 or i == num_strahlen - 1:
        print(f"  Strahl {i+1}: H√∂he h = {h:6.2f} ‚Üí Bildpunkt bei x = {b:.2f}")

# ============ BILDPUNKT MARKIEREN ============
ax.plot([b], [0], 'g*', markersize=15, label='Bildpunkt (Brennpunkt der Strahlen)', zorder=5)

# ============ GEGENSTAND und BILD MARKIEREN ============
ax.plot([-g], [0], 'ks', markersize=10, label=f'Linsen-Mittelpunkt')

# ============ ACHSEN-EINSTELLUNGEN ============
ax.set_xlim(-6, 6)
ax.set_ylim(-3.5, 3.5)
ax.set_aspect('equal')
ax.grid(True, alpha=0.2)
ax.set_xlabel('Position (beliebige Einheit)', fontsize=11)
ax.set_ylabel('H√∂he (beliebige Einheit)', fontsize=11)
ax.set_title(f'Ray-Tracing: {num_strahlen} Lichtstrahlen durch eine Sammellinse', 
             fontsize=13, fontweight='bold')
ax.legend(loc='upper right', fontsize=10)

plt.tight_layout()
plt.show()

print(f"\n‚úì Alle {num_strahlen} Strahlen konvergieren im Bildpunkt!")
print(f"‚úì Dies ist die Grundlage des Ray-Tracings!")
```

### Erkl√§rungen zum Code

**Das Ray-Tracing-Prinzip:**
1. Ein **paralleler Strahl** f√§llt auf die Linse
2. Nach der Linse wird der Strahl **gebrochen**
3. Der Strahl l√§uft durch den **Brennpunkt F'**
4. Alle Strahlen konvergieren in einem **Bildpunkt**

**Mathematik dahinter:**
- Ein Strahl in der H√∂he h wird mit der Steigung $$m = -\frac{h}{f}$$ gebrochen
- Die neue H√∂he nach Durchgang durch die Linse: $$y = m \cdot x = -\frac{h}{f} \cdot x$$

## üî¨ L√∂sung zu Aufgabe 5: Materialabh√§ngigkeit ‚Äì Vergleich von 3 Medien

### Vollst√§ndige L√∂sung

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ============ MATERIAL-DATEN ============
materialien = {
    "Luft‚ÜíLuft": {"n1": 1.0, "n2": 1.0},
    "Luft‚ÜíWasser": {"n1": 1.0, "n2": 1.33},
    "Luft‚ÜíGlas": {"n1": 1.0, "n2": 1.5}
}

einfallswinkel_grad = 45

# ============ BERECHNUNGEN F√úR ALLE MATERIALIEN ============
ergebnisse = []

for material_name, indices in materialien.items():
    n1 = indices["n1"]
    n2 = indices["n2"]
    
    # Berechne den Brechungswinkel mit Snellius' Gesetz
    theta1_rad = np.radians(einfallswinkel_grad)
    sin_theta2 = (n1 / n2) * np.sin(theta1_rad)
    
    # √úberpr√ºfe, ob Totalreflexion m√∂glich ist
    if sin_theta2 > 1:
        brechungswinkel = None  # Totalreflexion!
        differenz = None
    else:
        theta2_rad = np.arcsin(sin_theta2)
        brechungswinkel = np.degrees(theta2_rad)
        differenz = einfallswinkel_grad - brechungswinkel
    
    ergebnisse.append({
        "Material": material_name,
        "n‚ÇÅ": n1,
        "n‚ÇÇ": n2,
        "Œ∏‚ÇÅ (¬∞)": einfallswinkel_grad,
        "Œ∏‚ÇÇ (¬∞)": brechungswinkel,
        "Differenz (¬∞)": differenz
    })

# ============ FIGUR MIT 3 SUBPLOTS ============
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Farben f√ºr die drei Szenen
farben_medium = ["yellow", "lightblue", "lightgray"]
material_names = list(materialien.keys())

# ============ SCHLEIFE √úBER DIE 3 SZENEN ============
for idx, (material_name, ax) in enumerate(zip(material_names, axes)):
    
    n1 = materialien[material_name]["n1"]
    n2 = materialien[material_name]["n2"]
    
    # Berechne die Winkel
    theta1_rad = np.radians(einfallswinkel_grad)
    sin_theta2 = (n1 / n2) * np.sin(theta1_rad)
    
    if sin_theta2 > 1:
        brechungswinkel = None
    else:
        theta2_rad = np.arcsin(sin_theta2)
        brechungswinkel = np.degrees(theta2_rad)
    
    # ============ MEDIEN zeichnen ============
    ax.fill_between([-3, 3], 0, 2.5, alpha=0.2, color=farben_medium[idx])
    ax.fill_between([-3, 3], -2.5, 0, alpha=0.3, color=farben_medium[idx])
    
    # ============ GRENZFL√ÑCHE ============
    ax.plot([-3, 3], [0, 0], 'k-', linewidth=2.5)
    
    # ============ NORMALE ============
    ax.plot([0, 0], [-2.5, 2.5], 'gray', linewidth=1, linestyle='--', alpha=0.7)
    
    # ============ STRAHLEN zeichnen ============
    if brechungswinkel is not None:
        # Einfallender Strahl
        x1_end = -2.5 * np.sin(np.radians(einfallswinkel_grad))
        y1_end = 2.5 * np.cos(np.radians(einfallswinkel_grad))
        ax.plot([x1_end, 0], [y1_end, 0], 'r-', linewidth=2.5, label='Einfallend')
        
        # Gebrochener Strahl
        x2_end = 2.5 * np.sin(np.radians(brechungswinkel))
        y2_end = -2.5 * np.cos(np.radians(brechungswinkel))
        ax.plot([0, x2_end], [0, y2_end], 'orange', linewidth=2.5, label='Gebrochen')
        
        # Winkel beschriften
        ax.text(-0.6, 1.5, f'{einfallswinkel_grad}¬∞', fontsize=11, color='red', fontweight='bold')
        ax.text(0.4, -1.0, f'{brechungswinkel:.1f}¬∞', fontsize=11, color='orange', fontweight='bold')
    else:
        # Totalreflexion
        ax.text(0, 0.5, "TOTALREFLEXION!", fontsize=12, ha='center', color='red', 
                fontweight='bold', bbox=dict(boxstyle='round', facecolor='yellow', alpha=0.8))
    
    # ============ ACHSEN-EINSTELLUNGEN ============
    ax.set_xlim(-3, 3)
    ax.set_ylim(-3, 3)
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.2)
    ax.set_xlabel('x', fontsize=10)
    ax.set_ylabel('y', fontsize=10)
    
    # Titel mit Material-Info
    titel_text = f"{material_name}\n(n‚ÇÅ={n1}, n‚ÇÇ={n2})"
    ax.set_title(titel_text, fontsize=11, fontweight='bold')
    
    if idx == 0:
        ax.legend(loc='upper right', fontsize=9)

plt.suptitle(f"Brechung bei verschiedenen Materialien (Einfallswinkel = {einfallswinkel_grad}¬∞)", 
             fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# ============ TABELLE MIT ERGEBNISSEN ============
print("=" * 80)
print("VERGLEICH DER BRECHUNG IN VERSCHIEDENEN MATERIALIEN")
print("=" * 80)

df = pd.DataFrame(ergebnisse)
print(df.to_string(index=False))

print("\n" + "=" * 80)
print("ANALYSE:")
print("=" * 80)

for i, ergebnis in enumerate(ergebnisse):
    print(f"\n{ergebnis['Material']}:")
    print(f"  ‚Ä¢ Brechungsindizes: n‚ÇÅ = {ergebnis['n‚ÇÅ']}, n‚ÇÇ = {ergebnis['n‚ÇÇ']}")
    print(f"  ‚Ä¢ Einfallswinkel: {ergebnis['Œ∏‚ÇÅ (¬∞)']}¬∞")
    
    if ergebnis['Œ∏‚ÇÇ (¬∞)'] is not None:
        print(f"  ‚Ä¢ Brechungswinkel: {ergebnis['Œ∏‚ÇÇ (¬∞)']:.2f}¬∞")
        print(f"  ‚Ä¢ Differenz: {ergebnis['Differenz (¬∞)']:.2f}¬∞")
        
        if ergebnis['Differenz (¬∞)'] == 0:
            print(f"  ‚Ä¢ ‚ÑπÔ∏è Keine Brechung (gleiche Medien)!")
        elif ergebnis['Differenz (¬∞)'] > 0:
            print(f"  ‚Ä¢ ‚úì Brechung zum Lot hin (optisch dichter)")
    else:
        print(f"  ‚Ä¢ ‚ö†Ô∏è TOTALREFLEXION m√∂glich!")

print("\n" + "=" * 80)
print("REGEL:")
print("=" * 80)
print("Je gr√∂√üer der Unterschied zwischen n‚ÇÅ und n‚ÇÇ, desto st√§rker die Brechung!")
print("Luft‚ÜíLuft: Keine Brechung (Œîn = 0)")
print("Luft‚ÜíWasser: Starke Brechung (Œîn = 0,33)")
print("Luft‚ÜíGlas: Sehr starke Brechung (Œîn = 0,5)")
print("=" * 80)
```

## üîç H√§ufige Fehler und wie man sie vermeidet

### Fehler 1: Grad vs. Radianten

```python
# ‚ùå FALSCH:
angle_rad = np.sin(45)  # Vergibt, 45 ist in Grad, nicht Radianten!

# ‚úÖ RICHTIG:
angle_rad = np.sin(np.radians(45))  # Zuerst zu Radianten konvertieren
```

### Fehler 2: Linsenschleiferformel-Berechnung

```python
# ‚ùå FALSCH:
b = 1/f - 1/g  # Falsche Reihenfolge!

# ‚úÖ RICHTIG:
b = 1 / (1/f - 1/g)  # Erst addieren/subtrahieren, dann invertieren
```

### Fehler 3: Ray-Tracing-Steigung

```python
# ‚ùå FALSCH:
steigung = h / f  # Falsch: ignoriert die Konvention

# ‚úÖ RICHTIG:
steigung = -h / f  # Richtig: Die Steigung ist negativ (Strahl wird nach unten gebogen)
```

## üí° Weitere Vertiefungsm√∂glichkeiten

### Bonus 1: Totalreflexion
Bei bestimmten Winkeln tritt **Totalreflexion** auf ‚Äì das Licht wird komplett reflektiert, nicht gebrochen.

Der kritische Winkel:
$$\theta_c = \arcsin\left(\frac{n_2}{n_1}\right)$$

### Bonus 2: Mehrere Linsen
Kombiniere zwei Linsen hintereinander ‚Äì das Bild der ersten Linse wird zum Gegenstand der zweiten!

### Bonus 3: Dispersion
Unterschiedliche Farben haben unterschiedliche Brechungsindizes ‚Üí **Farbzerlegung** (Prismen, Regenbogen)

### Bonus 4: Linsen-Hersteller-Formel
$$P = (n-1) \left(\frac{1}{R_1} - \frac{1}{R_2}\right)$$

Diese Formel beschreibt, wie die **Kr√ºmmungsradien** R‚ÇÅ und R‚ÇÇ die Brennweite beeinflussen.
