## Lösung Aufgabe 1: Ort und Verschiebung berechnen

### Aufgabenstellung (Wiederholung)

Zugfahrer-Positionen: A (12 km), B (45 km), C (52 km), D (38 km)

### Lösungsweg

#### Verschiebung A zu B:
$$
\Delta s_{AB} = s_B - s_A = 45 - 12 = 33 \text{ km}
$$

#### Verschiebung B zu C:
$$
\Delta s_{BC} = s_C - s_B = 52 - 45 = 7 \text{ km}
$$

#### Verschiebung C zu D:
$$
\Delta s_{CD} = s_D - s_C = 38 - 52 = -14 \text{ km}
$$
(Negativ bedeutet: der Zug fährt rückwärts!)

#### Gesamtverschiebung A zu D:
$$
\Delta s_{AD} = s_D - s_A = 38 - 12 = 26 \text{ km}
$$

**Alternative Berechnung:** Alle Verschiebungen addieren:
$$
\Delta s_{AD} = 33 + 7 + (-14) = 26 \text{ km}
$$

#### Gesamtweg (Summe aller Streckenabschnitte):
$$
\text{Gesamtweg} = |33| + |7| + |-14| = 33 + 7 + 14 = 54 \text{ km}
$$

### Vollständiger Python-Code

````
# Positionen der Haltestellen (in km)
pos_A = 12
pos_B = 45
pos_C = 52
pos_D = 38

# Verschiebung A zu B berechnen
verschiebung_A_zu_B = pos_B - pos_A  # 45 - 12 = 33 km
print(f"Verschiebung A zu B: {verschiebung_A_zu_B} km")

# Verschiebung B zu C berechnen
verschiebung_B_zu_C = pos_C - pos_B  # 52 - 45 = 7 km
print(f"Verschiebung B zu C: {verschiebung_B_zu_C} km")

# Verschiebung C zu D berechnen
verschiebung_C_zu_D = pos_D - pos_C  # 38 - 52 = -14 km
print(f"Verschiebung C zu D: {verschiebung_C_zu_D} km")

# Gesamtverschiebung A zu D berechnen
gesamtverschiebung = pos_D - pos_A  # 38 - 12 = 26 km
print(f"Gesamtverschiebung: {gesamtverschiebung} km")

# Gesamtweg berechnen (Summe der Beträge!)
gesamtweg = abs(verschiebung_A_zu_B) + abs(verschiebung_B_zu_C) + abs(verschiebung_C_zu_D)
# = |33| + |7| + |-14| = 33 + 7 + 14 = 54 km
print(f"Gesamtweg: {gesamtweg} km")

print("\n=== Interpretation ===")
print(f"Der Zug ist netto um {gesamtverschiebung} km vorangekommen,")
print(f"aber hat insgesamt {gesamtweg} km fahren müssen.")
````

**Ausgabe:**
```
Verschiebung A zu B: 33 km
Verschiebung B zu C: 7 km
Verschiebung C zu D: -14 km
Gesamtverschiebung: 26 km
Gesamtweg: 54 km

=== Interpretation ===
Der Zug ist netto um 26 km vorangekommen,
aber hat insgesamt 54 km fahren müssen.
```

### Antwort auf Reflexionsfragen

**Frage 1:** Warum ist der Gesamtweg größer als die Gesamtverschiebung?

**Antwort:** Weil der Zug von C nach D **zurückfährt** (-14 km). Der Gesamtweg zählt alle Bewegungen, ohne die Richtung zu berücksichtigen. Die Gesamtverschiebung berücksichtigt nur Start und Endpunkt.

**Frage 2:** Wann wäre der Gesamtweg gleich der Gesamtverschiebung?

**Antwort:** Wenn der Zug immer in die gleiche Richtung fährt (nie rückwärts). In unserem Beispiel wäre das der Fall, wenn statt D = 38 km die Endposition z.B. 60 km gewesen wäre.

## Lösung Aufgabe 2: Durchschnittsgeschwindigkeit berechnen

### Aufgabenstellung (Wiederholung)

Radfahrer: Phase 1 (6 km in 20 min), Phase 2 (8 km in 40 min)

### Lösungsweg

#### Phase 1: 10:00 bis 10:20 Uhr

Zeitumrechnung:
$$
t_1 = 20 \text{ min} = 20 \times 60 = 1200 \text{ s}
$$

Streckenumrechnung:
$$
s_1 = 6 \text{ km} = 6 \times 1000 = 6000 \text{ m}
$$

Geschwindigkeit:
$$
v_{\text{avg}, 1} = \frac{s_1}{t_1} = \frac{6000}{1200} = 5 \text{ m/s}
$$

Umrechnung in km/h:
$$
v_{\text{avg}, 1} = 5 \times 3,6 = 18 \text{ km/h}
$$

#### Phase 2: 10:20 bis 11:00 Uhr

Zeitumrechnung:
$$
t_2 = 40 \text{ min} = 40 \times 60 = 2400 \text{ s}
$$

Streckenumrechnung:
$$
s_2 = 8 \text{ km} = 8 \times 1000 = 8000 \text{ m}
$$

Geschwindigkeit:
$$
v_{\text{avg}, 2} = \frac{s_2}{t_2} = \frac{8000}{2400} \approx 3,33 \text{ m/s}
$$

Umrechnung in km/h:
$$
v_{\text{avg}, 2} \approx 3,33 \times 3,6 \approx 12 \text{ km/h}
$$

#### Gesamtdurchschnittsgeschwindigkeit

**Wichtig:** Die Gesamtgeschwindigkeit ist **nicht** der Durchschnitt der beiden Geschwindigkeiten!

Gesamtstrecke:
$$
s_{\text{ges}} = 6 + 8 = 14 \text{ km} = 14000 \text{ m}
$$

Gesamtzeit:
$$
t_{\text{ges}} = 20 + 40 = 60 \text{ min} = 3600 \text{ s}
$$

Gesamtgeschwindigkeit:
$$
v_{\text{avg}, ges} = \frac{s_{\text{ges}}}{t_{\text{ges}}} = \frac{14000}{3600} \approx 3,89 \text{ m/s}
$$

Umrechnung in km/h:
$$
v_{\text{avg}, ges} \approx 3,89 \times 3,6 = 14 \text{ km/h}
$$

### Vollständiger Python-Code

````
# Phase 1: 10:00 bis 10:20 Uhr (20 Minuten)
strecke_1 = 6      # km
zeit_1 = 20        # Minuten

# Zeit in Sekunden umrechnen
zeit_1_s = zeit_1 * 60  # 20 * 60 = 1200 s

# Strecke in Meter umrechnen
strecke_1_m = strecke_1 * 1000  # 6 * 1000 = 6000 m

# Durchschnittsgeschwindigkeit berechnen (in m/s)
v_avg_1_ms = strecke_1_m / zeit_1_s  # 6000 / 1200 = 5 m/s

# Umrechnung in km/h
v_avg_1_kmh = v_avg_1_ms * 3.6  # 5 * 3.6 = 18 km/h

print(f"Phase 1: {v_avg_1_ms:.2f} m/s oder {v_avg_1_kmh:.2f} km/h")

# Phase 2: 10:20 bis 11:00 Uhr (40 Minuten)
strecke_2 = 8      # km
zeit_2 = 40        # Minuten

# Zeit in Sekunden umrechnen
zeit_2_s = zeit_2 * 60  # 40 * 60 = 2400 s

# Strecke in Meter umrechnen
strecke_2_m = strecke_2 * 1000  # 8 * 1000 = 8000 m

# Durchschnittsgeschwindigkeit berechnen (in m/s)
v_avg_2_ms = strecke_2_m / zeit_2_s  # 8000 / 2400 = 3.33... m/s

# Umrechnung in km/h
v_avg_2_kmh = v_avg_2_ms * 3.6  # 3.33... * 3.6 = 12 km/h

print(f"Phase 2: {v_avg_2_ms:.2f} m/s oder {v_avg_2_kmh:.2f} km/h")

# Gesamtdurchschnittsgeschwindigkeit
gesamtstrecke_km = strecke_1 + strecke_2  # 6 + 8 = 14 km
gesamtzeit_min = zeit_1 + zeit_2          # 20 + 40 = 60 min

# Umrechnung
gesamtstrecke_m = gesamtstrecke_km * 1000  # 14 * 1000 = 14000 m
gesamtzeit_s = gesamtzeit_min * 60          # 60 * 60 = 3600 s

# Gesamtdurchschnittsgeschwindigkeit
v_avg_gesamt_ms = gesamtstrecke_m / gesamtzeit_s  # 14000 / 3600 = 3.889 m/s
v_avg_gesamt_kmh = v_avg_gesamt_ms * 3.6         # 3.889 * 3.6 = 14 km/h

print(f"\nGesamtdurchschnittsgeschwindigkeit: {v_avg_gesamt_ms:.2f} m/s oder {v_avg_gesamt_kmh:.2f} km/h")

# Warnung: Das ist NICHT der Durchschnitt der Durchschnitte!
print(f"\n⚠️  WICHTIG:")
print(f"Durchschnitt der Geschwindigkeiten: {(v_avg_1_ms + v_avg_2_ms) / 2:.2f} m/s")
print(f"Das ist FALSCH! Richtig: {v_avg_gesamt_ms:.2f} m/s")
````

**Ausgabe:**
```
Phase 1: 5.00 m/s oder 18.00 km/h
Phase 2: 3.33 m/s oder 12.00 km/h

Gesamtdurchschnittsgeschwindigkeit: 3.89 m/s oder 14.00 km/h

⚠️  WICHTIG:
Durchschnitt der Geschwindigkeiten: 4.17 m/s
Das ist FALSCH! Richtig: 3.89 m/s
```

### Warum ist die Gesamtgeschwindigkeit nicht das Durchschnitt der Phasen-Geschwindigkeiten?

Der Radfahrer fährt in Phase 2 **langsamer als in Phase 1**, aber **länger**. Darum wird die Gesamtgeschwindigkeit mehr von Phase 2 beeinflusst.

Allgemein: Gesamtgeschwindigkeit = Gesamtstrecke / Gesamtzeit (nicht Durchschnitt der Einzelgeschwindigkeiten!)

## Lösung Aufgabe 3: Beschleunigung aus Geschwindigkeitsdaten

### Aufgabenstellung (Wiederholung)

Speedtest: Nach 2 s → 10 m/s, nach 5 s → 25 m/s, nach 8 s → 22 m/s

### Lösungsweg

#### Beschleunigung von 2–5 Sekunden

$$
a_1 = \frac{\Delta v}{\Delta t} = \frac{v_2 - v_1}{t_2 - t_1} = \frac{25 - 10}{5 - 2} = \frac{15}{3} = 5 \text{ m/s}^2
$$

Das Auto **beschleunigt** mit 5 m/s².

#### Beschleunigung von 5–8 Sekunden

$$
a_2 = \frac{\Delta v}{\Delta t} = \frac{v_3 - v_2}{t_3 - t_2} = \frac{22 - 25}{8 - 5} = \frac{-3}{3} = -1 \text{ m/s}^2
$$

Das Auto **verzögert** (negative Beschleunigung) mit -1 m/s².

### Vollständiger Python-Code

````
# Geschwindigkeitsmessungen
t_1, v_1 = 2, 10      # Zeitpunkt 1: 2 s, Geschwindigkeit 10 m/s
t_2, v_2 = 5, 25      # Zeitpunkt 2: 5 s, Geschwindigkeit 25 m/s
t_3, v_3 = 8, 22      # Zeitpunkt 3: 8 s, Geschwindigkeit 22 m/s

# Beschleunigung zwischen t_1 und t_2 berechnen
delta_v_1 = v_2 - v_1  # 25 - 10 = 15 m/s
delta_t_1 = t_2 - t_1  # 5 - 2 = 3 s
a_1 = delta_v_1 / delta_t_1  # 15 / 3 = 5 m/s²

print(f"Beschleunigung (2–5 s): {a_1:.2f} m/s²")
print(f"  → Das Auto wird schneller (positive Beschleunigung)\n")

# Beschleunigung zwischen t_2 und t_3 berechnen
delta_v_2 = v_3 - v_2  # 22 - 25 = -3 m/s
delta_t_2 = t_3 - t_2  # 8 - 5 = 3 s
a_2 = delta_v_2 / delta_t_2  # -3 / 3 = -1 m/s²

print(f"Beschleunigung (5–8 s): {a_2:.2f} m/s²")
print(f"  → Das Auto wird langsamer (negative Beschleunigung / Verzögerung)\n")

# Interpretation
print("=== Interpretation ===")
if a_1 > 0:
    print(f"Phase 1: Positive Beschleunigung → Fahrer drückt aufs Gas")
if a_2 < 0:
    print(f"Phase 2: Negative Beschleunigung → Fahrer tritt auf die Bremse")
    print(f"        oder nimmt den Fuß vom Gas")
````

**Ausgabe:**
```
Beschleunigung (2–5 s): 5.00 m/s²
  → Das Auto wird schneller (positive Beschleunigung)

Beschleunigung (5–8 s): -1.00 m/s²
  → Das Auto wird langsamer (negative Beschleunigung / Verzögerung)

=== Interpretation ===
Phase 1: Positive Beschleunigung → Fahrer drückt aufs Gas
Phase 2: Negative Beschleunigung → Fahrer tritt auf die Bremse
        oder nimmt den Fuß vom Gas
```

### Was bedeuten die Ergebnisse?

- **Positive a:** Die Geschwindigkeit nimmt zu → Beschleunigung "im allgemeinen Sinne"
- **Negative a:** Die Geschwindigkeit nimmt ab → Verzögerung / Bremsen
- **a = 0:** Konstante Geschwindigkeit (kein Gas, keine Bremse)

## Lösung Aufgabe 4: Position berechnen – Gleichförmige Bewegung

### Aufgabenstellung (Wiederholung)

Zug mit v = 25 m/s, s₀ = 100 m. Berechne Position nach 2, 5, 10 Sekunden und zeichne Diagramm.

### Lösungsweg

**Formel:** $$s(t) = s_0 + v \cdot t = 100 + 25 \cdot t$$

Position nach 2 Sekunden:
$$
s(2) = 100 + 25 \times 2 = 100 + 50 = 150 \text{ m}
$$

Position nach 5 Sekunden:
$$
s(5) = 100 + 25 \times 5 = 100 + 125 = 225 \text{ m}
$$

Position nach 10 Sekunden:
$$
s(10) = 100 + 25 \times 10 = 100 + 250 = 350 \text{ m}
$$

### Vollständiger Python-Code

````
import matplotlib.pyplot as plt
import numpy as np

# Gegebene Werte
s_0 = 100      # Startposition in m
v = 25         # Geschwindigkeit in m/s (konstant)

# Berechne die Position nach 2, 5 und 10 Sekunden
s_nach_2s = s_0 + v * 2    # 100 + 25*2 = 150 m
s_nach_5s = s_0 + v * 5    # 100 + 25*5 = 225 m
s_nach_10s = s_0 + v * 10  # 100 + 25*10 = 350 m

print(f"Position nach 2 s: {s_nach_2s} m")
print(f"Position nach 5 s: {s_nach_5s} m")
print(f"Position nach 10 s: {s_nach_10s} m\n")

# Erstelle ein Zeit-Array von 0 bis 10 Sekunden mit 100 Punkten
t = np.linspace(0, 10, 100)

# Berechne die Positionen für alle Zeitpunkte
s = s_0 + v * t

# Zeichne das Diagramm
plt.figure(figsize=(10, 6))
plt.plot(t, s, linewidth=2.5, color='blue', label='Position s(t)')

# Markiere die berechneten Punkte
plt.plot([2, 5, 10], [s_nach_2s, s_nach_5s, s_nach_10s], 'ro', markersize=8, label='Berechnete Punkte')

# Achsenbeschriftung und Gitter
plt.xlabel('Zeit t (s)', fontsize=12)
plt.ylabel('Position s (m)', fontsize=12)
plt.title('Gleichförmige Bewegung: Zug mit v = 25 m/s', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)

# Zusätzlich: Werte an den Achsen anzeigen
plt.axhline(y=s_0, color='k', linestyle='--', alpha=0.2, linewidth=1)
plt.text(0.1, s_0, f's₀ = {s_0} m', fontsize=10)

plt.tight_layout()
plt.show()

# Zusätzliche Informationen
print("=== Eigenschaften der Kurve ===")
print("Die Kurve ist eine GERADE Linie → gleichförmige Bewegung")
print(f"Steigung = v = {v} m/s")
print(f"y-Achsenabschnitt = s₀ = {s_0} m")
````

**Ausgabe:**
```
Position nach 2 s: 150 m
Position nach 5 s: 225 m
Position nach 10 s: 350 m

=== Eigenschaften der Kurve ===
Die Kurve ist eine GERADE Linie → gleichförmige Bewegung
Steigung = v = 25 m/s
y-Achsenabschnitt = s₀ = 100 m
```

### Interpretation

Die Position-Zeit-Kurve ist eine **Gerade**. Das ist das Erkennungszeichen einer gleichförmigen Bewegung. Die Steigung der Geraden ist genau die Geschwindigkeit (25 m/s).

## Lösung Aufgabe 5: Position berechnen – Beschleunigte Bewegung

### Aufgabenstellung (Wiederholung)

Skateboarder: v₀ = 0, a = 1,5 m/s². Berechne Positionen und Geschwindigkeiten, zeichne Diagramme.

### Lösungsweg

**Formeln:**
$$
s(t) = v_0 \cdot t + \frac{1}{2} a \cdot t^2 = 0 + \frac{1}{2} \cdot 1,5 \cdot t^2 = 0,75 \cdot t^2
$$

$$
v(t) = v_0 + a \cdot t = 0 + 1,5 \cdot t = 1,5 \cdot t
$$

| Zeit (s) | Position (m) | Geschwindigkeit (m/s) |
|----------|-------------|----------------------|
| 1 | 0,75 | 1,5 |
| 2 | 3,00 | 3,0 |
| 3 | 6,75 | 4,5 |
| 4 | 12,00 | 6,0 |
| 5 | 18,75 | 7,5 |

### Vollständiger Python-Code

````
import matplotlib.pyplot as plt
import numpy as np

# Gegebene Werte
s_0 = 0        # Startposition (am Anfang der Rampe)
v_0 = 0        # Anfangsgeschwindigkeit (Stillstand)
a = 1.5        # Beschleunigung in m/s²

# Zeitpunkte (in Sekunden)
zeiten = [1, 2, 3, 4, 5]

print("Berechnung für verschiedene Zeitpunkte:")
print("Zeit (s) | Position (m) | Geschwindigkeit (m/s)")
print("-" * 50)

for t_i in zeiten:
    # Position berechnen
    s_i = s_0 + v_0 * t_i + 0.5 * a * t_i**2  # = 0.75 * t²
    
    # Geschwindigkeit berechnen
    v_i = v_0 + a * t_i  # = 1.5 * t
    
    print(f"{t_i:7} | {s_i:12.2f} | {v_i:19.2f}")

# Feinere Zeit-Auflösung für die Visualisierung
t = np.linspace(0, 5, 100)

# Position und Geschwindigkeit für alle Zeitpunkte berechnen
s = s_0 + v_0 * t + 0.5 * a * t**2  # = 0.75 * t²
v = v_0 + a * t  # = 1.5 * t

# Zwei Diagramme nebeneinander zeichnen
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Diagramm 1: Position über Zeit
ax1.plot(t, s, linewidth=2.5, color='blue', label='s(t) = 0.75·t²')
ax1.scatter([1, 2, 3, 4, 5], [0.75, 3, 6.75, 12, 18.75], color='red', s=60, zorder=5)
ax1.set_xlabel('Zeit t (s)', fontsize=11)
ax1.set_ylabel('Position s (m)', fontsize=11)
ax1.set_title('Beschleunigte Bewegung: Position über Zeit', fontsize=12)
ax1.grid(True, alpha=0.3)
ax1.legend(fontsize=10)

# Diagramm 2: Geschwindigkeit über Zeit
ax2.plot(t, v, linewidth=2.5, color='green', label='v(t) = 1.5·t')
ax2.scatter([1, 2, 3, 4, 5], [1.5, 3.0, 4.5, 6.0, 7.5], color='red', s=60, zorder=5)
ax2.set_xlabel('Zeit t (s)', fontsize=11)
ax2.set_ylabel('Geschwindigkeit v (m/s)', fontsize=11)
ax2.set_title('Beschleunigte Bewegung: Geschwindigkeit über Zeit', fontsize=12)
ax2.grid(True, alpha=0.3)
ax2.legend(fontsize=10)

fig.suptitle('Skateboarder auf der Rampe (a = 1.5 m/s²)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

# Zusätzliche Analyse
print("\n=== Kurvenformen ===")
print("s(t)-Kurve: PARABEL (quadratische Funktion)")
print("v(t)-Kurve: GERADE (lineare Funktion)")
print("Steigung der v(t)-Kurve = a = 1.5 m/s²")
````

**Ausgabe:**
```
Berechnung für verschiedene Zeitpunkte:
Zeit (s) | Position (m) | Geschwindigkeit (m/s)
--------------------------------------------------
      1 |         0.75 |                  1.50
      2 |         3.00 |                  3.00
      3 |         6.75 |                  4.50
      4 |        12.00 |                  6.00
      5 |        18.75 |                  7.50

=== Kurvenformen ===
s(t)-Kurve: PARABEL (quadratische Funktion)
v(t)-Kurve: GERADE (lineare Funktion)
Steigung der v(t)-Kurve = a = 1.5 m/s²
```

### Interpretation

- Die **s(t)-Kurve ist eine Parabel**: Die Position nimmt quadratisch mit der Zeit zu.
- Die **v(t)-Kurve ist eine Gerade**: Die Geschwindigkeit nimmt linear mit der Zeit zu.
- Die **Steigung** der v(t)-Geraden ist genau die Beschleunigung (1,5 m/s²).

## Lösung Aufgabe 6: Mini-Projekt – Freier Fall simulieren

### Aufgabenstellung (Wiederholung)

Ball fällt aus 100 m. Berechne Fallzeit, Aufprallgeschwindigkeit und erstelle drei Diagramme.

### Lösungsweg

#### Fallzeit berechnen

Formel für freien Fall: $$h = h_0 - \frac{1}{2} g t^2$$

Der Ball hat Boden erreicht, wenn h = 0:
$$
0 = h_0 - \frac{1}{2} g t^2
$$

Umformen:
$$
\frac{1}{2} g t^2 = h_0
$$

$$
t^2 = \frac{2 h_0}{g}
$$

$$
t = \sqrt{\frac{2 h_0}{g}} = \sqrt{\frac{2 \times 100}{9.81}} = \sqrt{20.39} \approx 4.52 \text{ s}
$$

#### Aufprallgeschwindigkeit

$$
v_{\text{Aufprall}} = g \cdot t = 9.81 \times 4.52 \approx 44.3 \text{ m/s}
$$

### Vollständiger Python-Code

````
import matplotlib.pyplot as plt
import numpy as np

# Anfangsbedingungen
h_0 = 100      # Anfangshöhe in m
v_0 = 0        # Anfangsgeschwindigkeit (wird losgelassen)
g = 9.81       # Fallbeschleunigung in m/s²
m = 0.5        # Ballmasse in kg

# Berechne die Fallzeit (t = sqrt(2*h_0 / g))
t_fall = np.sqrt(2 * h_0 / g)
print(f"Fallzeit: {t_fall:.2f} Sekunden\n")

# Berechne die Aufprallgeschwindigkeit
v_aufprall = g * t_fall
print(f"Aufprallgeschwindigkeit: {v_aufprall:.2f} m/s\n")

# Zeitarray für die Simulation (0 bis t_fall)
t = np.linspace(0, t_fall, 100)

# Höhe, Geschwindigkeit und kinetische Energie berechnen
h = h_0 - 0.5 * g * t**2
v = g * t
E_kin = 0.5 * m * v**2

# Drei Diagramme nebeneinander zeichnen
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Diagramm 1: Höhe
axes[0].plot(t, h, linewidth=2.5, color='blue')
axes[0].axhline(y=0, color='r', linestyle='--', alpha=0.5, label='Boden')
axes[0].set_xlabel('Zeit t (s)', fontsize=11)
axes[0].set_ylabel('Höhe h (m)', fontsize=11)
axes[0].set_title('Freier Fall: Höhe über Zeit', fontsize=12)
axes[0].grid(True, alpha=0.3)
axes[0].legend()

# Diagramm 2: Geschwindigkeit
axes[1].plot(t, v, linewidth=2.5, color='green')
axes[1].axhline(y=v_aufprall, color='r', linestyle='--', alpha=0.5, label=f'v_Aufprall = {v_aufprall:.1f} m/s')
axes[1].set_xlabel('Zeit t (s)', fontsize=11)
axes[1].set_ylabel('Fallgeschwindigkeit v (m/s)', fontsize=11)
axes[1].set_title('Freier Fall: Geschwindigkeit über Zeit', fontsize=12)
axes[1].grid(True, alpha=0.3)
axes[1].legend()

# Diagramm 3: Kinetische Energie
axes[2].plot(t, E_kin, linewidth=2.5, color='red')
E_kin_max = E_kin[-1]
axes[2].axhline(y=E_kin_max, color='b', linestyle='--', alpha=0.5, label=f'E_kin,max = {E_kin_max:.0f} J')
axes[2].set_xlabel('Zeit t (s)', fontsize=11)
axes[2].set_ylabel('Kinetische Energie E_kin (J)', fontsize=11)
axes[2].set_title('Freier Fall: Kinetische Energie über Zeit', fontsize=12)
axes[2].grid(True, alpha=0.3)
axes[2].legend()

fig.suptitle('Ball fällt aus 100 m Höhe (Freier Fall)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

# Zusätzliche Analyse
print("=== Energieumwandlung ===")
E_pot_initial = m * g * h_0  # Anfangspotenzielle Energie
print(f"Anfängliche potenzielle Energie: {E_pot_initial:.1f} J")
print(f"Kinetische Energie bei Aufprall: {E_kin_max:.1f} J")
print(f"Unterschied (Messfehler): {abs(E_pot_initial - E_kin_max):.1f} J")

print("\n=== Kurvenformen ===")
print("h(t)-Kurve: PARABEL (nach oben konkav)")
print("v(t)-Kurve: GERADE (linear, da a = konstant)")
print("E_kin(t)-Kurve: PARABEL (quadratisch, da E ~ v²)")
````


### Physikalische Interpretation

**Energieerhaltung:** Die potenzielle Energie wird vollständig in kinetische Energie umgewandelt:
$$
E_{\text{pot}} = mgh_0 = 0.5 \times 9.81 \times 100 = 490.5 \text{ J}
$$

$$
E_{\text{kin}} = \frac{1}{2} m v^2 = \frac{1}{2} \times 0.5 \times 44.3^2 \approx 490.5 \text{ J}
$$

## Zusammenfassung der Lösungsstrategien

| Aufgabentyp | Formel | Python-Umsetzung |
|-----------|--------|------------------|
| Verschiebung | Δs = s_end - s_start | `delta_s = s_end - s_start` |
| Gesamtweg | ∑\|Δs_i\| | `gesamtweg = sum([abs(x) for x in delta_s_liste])` |
| Geschwindigkeit | v = s / t | `v = strecke / zeit` |
| Beschleunigung | a = Δv / Δt | `a = (v_end - v_start) / (t_end - t_start)` |
| Position (gleichförmig) | s = s₀ + v·t | `s = s_0 + v * t` |
| Position (beschleunigt) | s = s₀ + v₀·t + ½a·t² | `s = s_0 + v_0 * t + 0.5 * a * t**2` |
| Geschwindigkeit (beschleunigt) | v = v₀ + a·t | `v = v_0 + a * t` |

## Häufige Fehler und wie man sie vermeidet

1. **Falsche Einheiten:** Immer in SI-Einheiten arbeiten (m, s, kg)
2. **Verwechslung Weg vs. Verschiebung:** Weg ist immer ≥ 0, Verschiebung kann negativ sein
3. **Gesamtgeschwindigkeit falsch berechnet:** Nicht Durchschnitt der Einzelgeschwindigkeiten, sondern Gesamtstrecke / Gesamtzeit
4. **Parabel vs. Gerade verwechselt:** Beschleunigte Bewegung → s(t) ist Parabel, v(t) ist Gerade
5. **Fehlende Markierungen:** Immer Achsen, Legende und Titel im Diagramm einfügen
