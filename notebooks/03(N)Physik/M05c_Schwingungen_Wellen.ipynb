## Lösung zu Aufgabe 1: Harmonische Schwingung – Parameter erkennen

**Gegeben:**

$$
x(t) = 0,5 \sin(4\pi t)
$$

- $A = 0,5$ m
- $\omega = 4\pi$ rad/s
- $f = \frac{\omega}{2\pi} = \frac{4\pi}{2\pi} = 2$ Hz
- $T = \frac{1}{f} = 0,5$ s

Berechnung von $x(0)$, $x(T/4)$, $x(T/2)$:

$$
x(0) = 0,5 \cdot \sin(4\pi \cdot 0) = 0\text{ m}
$$
$$
x(T/4) = 0,5 \cdot \sin(4\pi \cdot 0,125) = 0,5 \cdot \sin(\pi/2) = 0,5 \cdot 1 = 0,5\text{ m}
$$
$$
x(T/2) = 0,5 \cdot \sin(4\pi \cdot 0,25) = 0,5 \cdot \sin(\pi) = 0\text{ m}
$$

**Antworten:**
- Amplitude $A = 0,5$ m
- Kreisfrequenz $\omega = 4\pi$ rad/s
- Frequenz $f = 2$ Hz
- Periode $T = 0,5$ s
- $x(0) = 0$ m, $x(T/4) = 0,5$ m, $x(T/2) = 0$ m

**Code:**

```python
import numpy as np

A = 0.5
omega = 4 * np.pi
f = omega / (2 * np.pi)
T = 1 / f

x_0 = A * np.sin(omega * 0)
x_T4 = A * np.sin(omega * T / 4)
x_T2 = A * np.sin(omega * T / 2)

print(f"Amplitude: A = {A} m")
print(f"Kreisfrequenz: ω = {omega:.4f} rad/s")
print(f"Frequenz: f = {f:.4f} Hz")
print(f"Periode: T = {T:.4f} s")
print(f"x(0) = {x_0:.4f} m")
print(f"x(T/4) = {x_T4:.4f} m")
print(f"x(T/2) = {x_T2:.4f} m")
```

**Reflexion:** $x(T/2) = x(0) = 0$ m, Weil der Sinus nach einer halben Periode wieder die Nullstelle erreicht. Die Schwingung ist periodisch.

## Lösung zu Aufgabe 2: Federpendel – Energie und Schwingung

- $m = 0,5$ kg
- $k = 50$ N/m
- $A = 0,1$ m

Kreisfrequenz:
$$
\omega = \sqrt{\frac{k}{m}} = \sqrt{\frac{50}{0,5}} = \sqrt{100} = 10\ \text{rad/s}
$$

Periode und Frequenz:
$$
T = \frac{2\pi}{\omega} = \frac{2\pi}{10} = 0,628\ \text{s}
$$
$$f = \frac{1}{T} \approx 1,592\ \text{Hz}
$$

Energien:

Potenzielle Energie (bei maximaler Auslenkung):
$$E_{pot,max} = \frac{1}{2} k A^2 = 0,5 \cdot 50 \cdot 0,1^2 = 0,25\ \text{J}$$

Kinetische Energie (bei Durchgang durch Gleichgewichtslage):
$$E_{kin,max} = E_{pot,max} = 0,25\ \text{J}$$

Maximalwert der Geschwindigkeit bei $x=0$ (T/4):

**Code:**

```python
m = 0.5
k = 50
A = 0.1
omega = np.sqrt(k / m)
T = 2 * np.pi / omega
f = 1 / T
E_pot_max = 0.5 * k * A**2
E_kin_max = E_pot_max
t_v_max = T / 4

print(f"Kreisfrequenz: ω = {omega:.4f} rad/s")
print(f"Periode: T = {T:.4f} s")
print(f"Frequenz: f = {f:.4f} Hz")
print(f"Max. potenzielle Energie: {E_pot_max:.4f} J")
print(f"Max. kinetische Energie: {E_kin_max:.4f} J")
print(f"Zeitpunkt: t = {t_v_max:.4f} s")
```

**Reflexion:** Die Periode ist 
emassenunabhängig von der Amplitude. Sie hängt nur von Masse und Federkonstante ab.

## Lösung zu Aufgabe 3: Welle – Wellenlänge und Frequenz

$c = 20$ m/s, $f = 5$ Hz

$$
\lambda = \frac{c}{f} = \frac{20}{5} = 4\ \text{m}
$$

$$
T = \frac{1}{f} = 0,2\ \text{s}
$$

$$
k = \frac{2\pi}{\lambda} = \frac{2\pi}{4} = 1,571\ \text{rad/m}
$$

$$
\omega = 2\pi f = 10\pi \approx 31,416\ \text{rad/s}
$$

Wellengleichung:
$$
y(x,t) = A \sin(1,571 \cdot x - 31,416 \cdot t)
$$

**Python:**

```python
c = 20
f = 5
lambda_ = c / f
T = 1 / f
k = 2 * np.pi / lambda_
omega = 2 * np.pi * f
print(f"Wellenlänge: λ = {lambda_:.4f} m")
print(f"Periode: T = {T:.4f} s")
print(f"Wellenzahl: k = {k:.4f} rad/m")
print(f"Kreisfrequenz: ω = {omega:.4f} rad/s")
print(f"Wellengleichung: y(x,t) = A sin({k:.4f}·x - {omega:.4f}·t)")
```

**Reflexion:** Bei Verdopplung der Frequenz halbiert sich bei konstantem $c$ die Wellenlänge $(\lambda = c/f)$.

## Lösung zu Aufgabe 4: Visualisierung – Federpendel simulieren

```python
import numpy as np
import matplotlib.pyplot as plt

m = 1.0
k = 25
A = 0.5
omega = np.sqrt(k / m)
T = 2 * np.pi / omega
f = 1 / T
t = np.linspace(0, 3*T, 1000)
x = A * np.sin(omega * t)
v = A * omega * np.cos(omega * t)
a = -omega**2 * x
fig, axes = plt.subplots(3, 1, figsize=(12, 8))
axes[0].plot(t, x, 'b-')
axes[0].axhline(y=0, color='k', linestyle='--', alpha=0.3)
for i in range(1, 4):
    axes[0].axvline(i*T, color='purple', linestyle='--', alpha=0.5)
axes[0].set_ylabel('x (m)')
axes[0].set_title('Auslenkung')
axes[0].grid(True, alpha=0.3)
axes[1].plot(t, v, 'r-')
axes[1].axhline(y=0, color='k', linestyle='--', alpha=0.3)
axes[1].set_ylabel('v (m/s)')
axes[1].set_title('Geschwindigkeit')
axes[1].grid(True, alpha=0.3)
axes[2].plot(t, a, 'g-')
axes[2].axhline(y=0, color='k', linestyle='--', alpha=0.3)
axes[2].set_ylabel('a (m/s²)')
axes[2].set_xlabel('Zeit (s)')
axes[2].set_title('Beschleunigung')
axes[2].grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

Die Maxima treten phasenverschoben auf: $\max(x)$ → $\max(v)$ nach $T/4$, $\max(a)$ vor $\max(x)$, denn $a(t)=-\omega^2 \cdot x(t)$.

## Zusatzaufgabe: Resonanz

Die Amplitudenkurve hat ihr Maximum bei $\omega_{ext}=\omega_{res}$. Die Dämpfung $\gamma$ reduziert das Maximum.

**Code:**

```python
import numpy as np
import matplotlib.pyplot as plt
m = 1.0
k = 25.0
F0 = 1.0
gamma = 0.5
omega_res = np.sqrt(k / m)
omega_ext = np.linspace(0, 3*omega_res, 500)
A_res = (F0/m) / np.sqrt((omega_res**2 - omega_ext**2)**2 + (gamma*omega_ext)**2)
plt.figure(figsize=(10, 6))
plt.plot(omega_ext, A_res, 'b-')
plt.axvline(omega_res, color='r', linestyle='--')
plt.xlabel('ω_ext (rad/s)')
plt.ylabel('A (m)')
plt.title('Resonanzkurve')
plt.grid(True)
plt.show()
```

**Reflexion:** Wird $\gamma$ kleiner, steigt $A_{max}$ an (peak wird schmaler und höher).
