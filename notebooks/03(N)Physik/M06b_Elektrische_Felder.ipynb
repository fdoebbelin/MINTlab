## üìã Aufgabe 3: Vektorfeld einer Ladung visualisieren

**Ziel:** Verstehe, wie Feldvektoren (Pfeile) die Feldrichtung und St√§rke zeigen.

**Aufgabe:**
Zeichne das Vektorfeld (Pfeile) einer positiven Ladung.

**Hinweise f√ºr deine L√∂sung:**
- Der Feldvektor zeigt von der Ladung radial nach au√üen
- Du brauchst die Komponenten Ex und Ey des Feldes
- Nutze `plt.quiver(X, Y, Ex, Ey)` um Pfeile zu zeichnen

```
import numpy as np
import matplotlib.pyplot as plt

k = 8.99e9
Q = 1e-6  # Eine positive Ladung

# TODO: Erstelle ein gr√∂beres Gitter (weniger Punkte als zuvor, z.B. 10√ó10)
x = np.linspace(-4, 4, 10)
y = np.linspace(-4, 4, 10)
X, Y = np.meshgrid(x, y)

# TODO: Berechne den Abstand r von jedem Punkt zur Ladung
r = np.sqrt(X**2 + Y**2)

# TODO: Berechne die Feldst√§rke (Magnitude)
E_mag = np.zeros_like(r)
mask = r > 0.1
E_mag[mask] = k * Q / (r[mask]**2)

# TODO: Berechne die Feldvektor-Komponenten
# Der Vektor zeigt radial nach au√üen: E = E_mag * (X/r, Y/r)
Ex = np.zeros_like(X)
Ey = np.zeros_like(Y)

for i in range(len(x)):
    for j in range(len(y)):
        if r[j, i] > 0.1:
            # Einheitsvektor in radiale Richtung
            u_r = X[j, i] / r[j, i]
            v_r = Y[j, i] / r[j, i]
            # Feldvektor
            Ex[j, i] = E_mag[j, i] * u_r
            Ey[j, i] = E_mag[j, i] * v_r

# TODO: Zeichne mit plt.quiver()
plt.figure(figsize=(10, 10))
plt.quiver(X, Y, Ex, Ey, ...)  # TODO: Welche Parameter brauchst du?

# TODO: Markiere die Ladung
plt.plot(0, 0, 'r*', markersize=20, label='Positive Ladung')

plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Vektorfeld einer Punktladung')
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.legend()
plt.show()
```

**L√∂sungshinweise:**
- `plt.quiver()` braucht mindestens: `X, Y, Ex, Ey`
- Du kannst die Pfeile einf√§rben: `plt.quiver(X, Y, Ex, Ey, E_mag, cmap='hot')`
- `scale` Parameter steuert die Pfeill√§nge
