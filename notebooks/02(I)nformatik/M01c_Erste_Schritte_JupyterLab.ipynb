## ‚úèÔ∏è Aufgabe 3: Arithmetische Operationen

### Musterl√∂sung

````
a = 17
b = 5
print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
print(f"{a} // {b} = {a // b}")
print(f"{a} % {b} = {a % b}")
print(f"{a} ** {b} = {a ** b}")
````

## ‚úèÔ∏è Aufgabe 6: Berechnung mit Variablen

### Musterl√∂sung

````
zahl1_text = input("Gib die erste Zahl ein: ")
zahl1 = int(zahl1_text)
zahl2_text = input("Gib die zweite Zahl ein: ")
zahl2 = int(zahl2_text)
summe = zahl1 + zahl2
differenz = zahl1 - zahl2
produkt = zahl1 * zahl2
quotient = zahl1 / zahl2
print(f"\n--- Rechenergebnisse ---")
print(f"{zahl1} + {zahl2} = {summe}")
print(f"{zahl1} - {zahl2} = {differenz}")
print(f"{zahl1} * {zahl2} = {produkt}")
print(f"{zahl1} / {zahl2} = {quotient:.2f}")
````

**Erkl√§rung:**
