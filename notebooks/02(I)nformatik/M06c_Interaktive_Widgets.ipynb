## üéØ Aufgabe 1: Einfacher Schieberegler ‚Äì L√∂sung

### Vollst√§ndige L√∂sung

```
import ipywidgets as widgets

# Schieberegler erstellen
slider = widgets.IntSlider(
    value=10,
    min=0,
    max=20,
    step=1,
    description='Zahl:'
)

# Funktion zur Pr√ºfung
def pr√ºfe_gerade_ungerade(n):
    """Pr√ºft, ob eine Zahl gerade oder ungerade ist."""
    if n % 2 == 0:
        print(f"{n} ist eine gerade Zahl")
    else:
        print(f"{n} ist eine ungerade Zahl")

# Observer hinzuf√ºgen
def handle_slider_change(change):
    pr√ºfe_gerade_ungerade(change['new'])

slider.observe(handle_slider_change, names='value')

display(slider)
```

### Erkl√§rung

- **`n % 2 == 0`**: Der Modulo-Operator `%` gibt den Rest nach der Division zur√ºck. Wenn `n % 2` gleich 0 ist, dann ist die Zahl durch 2 teilbar, also gerade.
- **`observe()`**: Diese Methode registriert eine Funktion, die aufgerufen wird, wenn sich der Wert des Widgets √§ndert.
- **`change['new']`**: Das Dictionary `change` enth√§lt Informationen √ºber die √Ñnderung. `change['new']` ist der neue Wert.

### Alternativl√∂sung (mit Lambda)

```
slider.observe(lambda change: pr√ºfe_gerade_ungerade(change['new']), names='value')
```

### Antworten auf Reflexionsfragen

- **Wenn du den Schieberegler auf 15 stellst:** Die Ausgabe ist "15 ist eine ungerade Zahl"
- **Lieblingszahl:** Beliebig, z. B. 7, 13, 19 (ungerade) oder 8, 14, 20 (gerade)

## üéØ Aufgabe 2: Interaktiver Temperatur-Umrechner ‚Äì L√∂sung

### Vollst√§ndige L√∂sung

```
from ipywidgets import interact
import ipywidgets as widgets

def umrechnen(celsius):
    """Konvertiert Celsius in Fahrenheit und Kelvin."""
    fahrenheit = (celsius * 9/5) + 32
    kelvin = celsius + 273.15
    
    print(f"Celsius:    {celsius}¬∞C")
    print(f"Fahrenheit: {fahrenheit:.2f}¬∞F")
    print(f"Kelvin:     {kelvin:.2f} K")
    print()

# interact() mit Schieberegler
interact(umrechnen, celsius=(-50, 50, 5))
```

### Erkl√§rung

- **Fahrenheit-Formel:** \(F = C \times \frac{9}{5} + 32\) ‚Äì Eine lineare Transformation mit Versatz
- **Kelvin-Formel:** \(K = C + 273.15\) ‚Äì Eine einfache Addition eines konstanten Versatzes
- **`:.2f`**: Formatiert die Zahl mit 2 Dezimalstellen
- **Tupel `(-50, 50, 5)`**: Bedeutet min=-50, max=50, step=5

### Erweiterte Version (mit mehr Kontrolle)

```
from ipywidgets import interact, widgets, VBox, Output

def umrechnen_erweitert(celsius):
    """Konvertiert Celsius in Fahrenheit und Kelvin mit zus√§tzlichen Informationen."""
    fahrenheit = (celsius * 9/5) + 32
    kelvin = celsius + 273.15
    
    print(f"{'Celsius:':<15} {celsius}¬∞C")
    print(f"{'Fahrenheit:':<15} {fahrenheit:.2f}¬∞F")
    print(f"{'Kelvin:':<15} {kelvin:.2f} K")
    print()
    
    # Zusatzinformation
    if celsius == 0:
        print("Das ist der Gefrierpunkt von Wasser (Normal).")
    elif celsius == 100:
        print("Das ist der Siedepunkt von Wasser (Normal).")
    elif celsius == -273.15:
        print("Das ist der absolute Nullpunkt!")
    elif celsius < -273.15:
        print("‚ö†Ô∏è Unm√∂gliche Temperatur! (unter absolutem Nullpunkt)")

interact(umrechnen_erweitert, celsius=(-50, 50, 5))
```

### Antworten auf Reflexionsfragen

- **Wann ist Celsius gleich Fahrenheit?** Bei -40¬∞C = -40¬∞F (L√∂sung: \(C = (C \times 9/5) + 32\) ergibt \(C = -40\))
- **Temperaturunterschied Erde:** Von etwa -89¬∞C (Antarktis) bis +54¬∞C (Todestal) = 143¬∞C

## üéØ Aufgabe 3: Farbmischer mit Dropdown ‚Äì L√∂sung

### Vollst√§ndige L√∂sung

```
import ipywidgets as widgets
from ipywidgets import VBox, Output

# Erstelle drei Dropdown-Widgets
farbe1 = widgets.Dropdown(
    options=['Rot', 'Gr√ºn', 'Blau'],
    value='Rot',
    description='Farbe 1:'
)

farbe2 = widgets.Dropdown(
    options=['Rot', 'Gr√ºn', 'Blau'],
    value='Gr√ºn',
    description='Farbe 2:'
)

farbe3 = widgets.Dropdown(
    options=['Rot', 'Gr√ºn', 'Blau'],
    value='Blau',
    description='Farbe 3:'
)

# Farb-Misch-Logik
FARB_MISCHUNGEN = {
    ('Rot', 'Rot', 'Rot'): 'Dunkelrot',
    ('Rot', 'Rot', 'Gr√ºn'): 'Orange-Rot',
    ('Rot', 'Rot', 'Blau'): 'Magenta',
    ('Rot', 'Gr√ºn', 'Blau'): 'Wei√ü',
    ('Rot', 'Gr√ºn', 'Gr√ºn'): 'Gelb-Gr√ºn',
    ('Rot', 'Gr√ºn', 'Rot'): 'Orange-Rot',
    ('Rot', 'Blau', 'Rot'): 'Magenta',
    ('Rot', 'Blau', 'Blau'): 'Violett',
    ('Rot', 'Blau', 'Gr√ºn'): 'Wei√ü',
    ('Gr√ºn', 'Gr√ºn', 'Gr√ºn'): 'Dunkelgr√ºn',
    ('Gr√ºn', 'Rot', 'Blau'): 'Wei√ü',
    ('Gr√ºn', 'Rot', 'Gr√ºn'): 'Gelb-Gr√ºn',
    ('Gr√ºn', 'Rot', 'Rot'): 'Gelb-Rot',
    ('Gr√ºn', 'Gr√ºn', 'Rot'): 'Gelb-Gr√ºn',
    ('Gr√ºn', 'Gr√ºn', 'Blau'): 'Cyan',
    ('Gr√ºn', 'Blau', 'Rot'): 'Wei√ü',
    ('Gr√ºn', 'Blau', 'Gr√ºn'): 'Cyan',
    ('Gr√ºn', 'Blau', 'Blau'): 'Cyan-Blau',
    ('Blau', 'Blau', 'Blau'): 'Dunkelblau',
    ('Blau', 'Rot', 'Gr√ºn'): 'Wei√ü',
    ('Blau', 'Rot', 'Blau'): 'Violett',
    ('Blau', 'Rot', 'Rot'): 'Magenta-Rot',
    ('Blau', 'Gr√ºn', 'Rot'): 'Wei√ü',
    ('Blau', 'Gr√ºn', 'Gr√ºn'): 'Cyan',
    ('Blau', 'Gr√ºn', 'Blau'): 'Cyan',
}

def mische_farben(f1, f2, f3):
    """Mischt drei Farben und zeigt das Ergebnis."""
    # Sortiere die Farben, um unabh√§ngig von Reihenfolge zu sein
    farben_sorted = tuple(sorted([f1, f2, f3]))
    
    # Z√§hle die H√§ufigkeit jeder Farbe
    rot_count = [f1, f2, f3].count('Rot')
    gr√ºn_count = [f1, f2, f3].count('Gr√ºn')
    blau_count = [f1, f2, f3].count('Blau')
    
    # Bestimme die Mischung
    if (f1, f2, f3) in FARB_MISCHUNGEN:
        ergebnis = FARB_MISCHUNGEN[(f1, f2, f3)]
    elif farben_sorted in FARB_MISCHUNGEN:
        ergebnis = FARB_MISCHUNGEN[farben_sorted]
    else:
        # Fallback: Beschreibe die Mischung selbst
        if rot_count == 3:
            ergebnis = 'Reines Rot'
        elif gr√ºn_count == 3:
            ergebnis = 'Reines Gr√ºn'
        elif blau_count == 3:
            ergebnis = 'Reines Blau'
        elif rot_count == 2:
            ergebnis = 'Rot-dominiert'
        elif gr√ºn_count == 2:
            ergebnis = 'Gr√ºn-dominiert'
        elif blau_count == 2:
            ergebnis = 'Blau-dominiert'
        else:
            ergebnis = 'Gemischte Farbe'
    
    print(f"{f1} + {f2} + {f3} = {ergebnis}")
    print(f"RGB-Anteile: Rot={rot_count}, Gr√ºn={gr√ºn_count}, Blau={blau_count}")

# Output-Widget
output = Output()

def aktualisiere_mischung(change):
    with output:
        output.clear_output()
        mische_farben(farbe1.value, farbe2.value, farbe3.value)

# Observer hinzuf√ºgen
farbe1.observe(aktualisiere_mischung, names='value')
farbe2.observe(aktualisiere_mischung, names='value')
farbe3.observe(aktualisiere_mischung, names='value')

# Initiale Anzeige
aktualisiere_mischung(None)

# Alles anzeigen
display(VBox([farbe1, farbe2, farbe3, output]))
```

### Erkl√§rung

- **Dictionary-Lookup:** Mit `(f1, f2, f3) in FARB_MISCHUNGEN` pr√ºfen wir, ob die Kombination im Dictionary existiert
- **Count-Methode:** `liste.count(element)` z√§hlt, wie oft ein Element vorkommt
- **Output-Widget:** Mit `output.clear_output()` l√∂schen wir die alte Ausgabe vor neuer Aktualisierung

### Antworten auf Reflexionsfragen

- **Anzahl Kombinationen:** \(3^3 = 27\) verschiedene Kombinationen m√∂glich
- **Lieblingsfarbmischung:** Z. B. Rot + Gr√ºn + Blau = Wei√ü (alle Farben gemischt)

## üéØ Aufgabe 4: Interaktiver Kreis-Rechner ‚Äì L√∂sung

### Vollst√§ndige L√∂sung

```
from ipywidgets import interact
import math

def berechne_kreis(radius):
    """Berechnet Fl√§che und Umfang eines Kreises."""
    if radius <= 0:
        print("‚ö†Ô∏è Der Radius muss eine positive Zahl sein!")
        return
    
    fl√§che = math.pi * (radius ** 2)
    umfang = 2 * math.pi * radius
    
    print(f"Radius:  {radius:.2f} Einheiten")
    print(f"Fl√§che:  {fl√§che:.4f} Einheiten¬≤")
    print(f"Umfang:  {umfang:.4f} Einheiten")
    print()
    
    # Zusatzinformation
    if fl√§che > umfang:
        print(f"‚ûú Die Fl√§che ({fl√§che:.2f}) ist gr√∂√üer als der Umfang ({umfang:.2f})")
    elif fl√§che < umfang:
        print(f"‚ûú Der Umfang ({umfang:.2f}) ist gr√∂√üer als die Fl√§che ({fl√§che:.2f})")
    else:
        print("‚ûú Fl√§che und Umfang sind gleich!")

interact(berechne_kreis, radius=(0.5, 10, 0.5))
```

### Erkl√§rung

- **Kreisfl√§che:** \(A = \pi r^2\)
- **Kreisumfang:** \(U = 2\pi r\)
- **Fehlerbehandlung:** Mit `if radius <= 0` verhindern wir ung√ºltige Eingaben
- **`math.pi`:** Der mathematische Wert von œÄ mit hoher Pr√§zision

### Erweiterte Version (mit Visualisierung)

```
from ipywidgets import interact
import math
import matplotlib.pyplot as plt
import numpy as np

def berechne_kreis_visual(radius):
    """Berechnet Kreis-Parameter und zeigt eine Visualisierung."""
    if radius <= 0:
        print("‚ö†Ô∏è Der Radius muss eine positive Zahl sein!")
        return
    
    fl√§che = math.pi * (radius ** 2)
    umfang = 2 * math.pi * radius
    
    print(f"Radius:  {radius:.2f} Einheiten")
    print(f"Fl√§che:  {fl√§che:.4f} Einheiten¬≤")
    print(f"Umfang:  {umfang:.4f} Einheiten")
    
    # Zeichne den Kreis
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # Parametrische Gleichung des Kreises
    theta = np.linspace(0, 2*np.pi, 100)
    x = radius * np.cos(theta)
    y = radius * np.sin(theta)
    
    ax.plot(x, y, 'b-', linewidth=2, label=f'Kreis (r={radius})')
    ax.fill(x, y, alpha=0.2)
    ax.plot([0, radius], [0, 0], 'r-', linewidth=2, label='Radius')
    
    ax.set_aspect('equal')
    ax.grid(True, alpha=0.3)
    ax.legend()
    ax.set_title(f'Kreis: r={radius}, A={fl√§che:.2f}, U={umfang:.2f}')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    
    plt.show()

interact(berechne_kreis_visual, radius=(0.5, 10, 0.5))
```

### Antworten auf Reflexionsfragen

- **Wann ist Fl√§che > Umfang?** Bei \(r > 2\): \(\pi r^2 > 2\pi r \Rightarrow r > 2\)
- **Verdopplung des Radius:** Fl√§che wird \((2r)^2 = 4r^2\), also 4-mal so gro√ü

## üéØ Aufgabe 5: Grafik mit Slider ‚Äì Sinuskurve ‚Äì L√∂sung

### Vollst√§ndige L√∂sung

```
from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt

def plot_sinus(amplitude, frequenz):
    """Zeichnet eine Sinuskurve mit variabler Amplitude und Frequenz."""
    x = np.linspace(0, 2*np.pi, 100)
    y = amplitude * np.sin(frequenz * x)
    
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, 'b-', linewidth=2)
    plt.xlabel('x')
    plt.ylabel('y = A¬∑sin(f¬∑x)')
    plt.title(f'Sinuskurve: A={amplitude}, f={frequenz}')
    plt.grid(True, alpha=0.3)
    plt.ylim(-5, 5)
    plt.show()

interact(plot_sinus, amplitude=(1, 5, 0.5), frequenz=(1, 3, 0.5))
```

### Erkl√§rung

- **`np.linspace(0, 2œÄ, 100)`:** Erstellt 100 gleichm√§√üig verteilte x-Werte von 0 bis 2œÄ
- **Amplitude (A):** Bestimmt die "H√∂he" der Welle (maximaler y-Wert)
- **Frequenz (f):** Bestimmt, wie viele "Schwingungen" in den Bereich passen
- **`plt.ylim(-5, 5)`:** Fixiert die y-Achse, damit die Skalierung nicht springt

### Erweiterte Version (mit mehreren Funktionen)

```
from ipywidgets import interact, widgets, VBox
import numpy as np
import matplotlib.pyplot as plt

def plot_trigonometrisch(amplitude, frequenz, funktion):
    """Zeichnet verschiedene trigonometrische Funktionen."""
    x = np.linspace(0, 2*np.pi, 100)
    
    if funktion == 'sin':
        y = amplitude * np.sin(frequenz * x)
        label = 'sin'
    elif funktion == 'cos':
        y = amplitude * np.cos(frequenz * x)
        label = 'cos'
    else:  # tan
        y = amplitude * np.tan(frequenz * x)
        label = 'tan'
    
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, 'b-', linewidth=2, label=label)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'{label.upper()}-Kurve: A={amplitude}, f={frequenz}')
    plt.grid(True, alpha=0.3)
    plt.ylim(-5, 5)
    plt.legend()
    plt.show()

interact(
    plot_trigonometrisch,
    amplitude=(1, 5, 0.5),
    frequenz=(1, 3, 0.5),
    funktion=widgets.Dropdown(options=['sin', 'cos', 'tan'], value='sin')
)
```

### Antworten auf Reflexionsfragen

- **Wenn Amplitude erh√∂ht:** Die Kurve wird "h√∂her" (gr√∂√üere maximale und minimale Werte)
- **Wenn Frequenz erh√∂ht:** Die Kurve oszilliert schneller (mehr Schwingungen in der gleichen Spanne)
- **Interessanteste Werte:** Z. B. A=2.5, f=2 oder A=3.5, f=1.5

## üéØ Aufgabe 6: Fibonacci-Generator ‚Äì L√∂sung

### Vollst√§ndige L√∂sung

```
from ipywidgets import interact

def fibonacci(n):
    """Berechnet und zeigt die ersten n Fibonacci-Zahlen."""
    if n <= 0:
        print("‚ö†Ô∏è n muss mindestens 1 sein!")
        return
    
    # Fibonacci-Folge berechnen
    fib_list = []
    a, b = 1, 1
    for i in range(n):
        fib_list.append(a)
        a, b = b, a + b
    
    # Ausgabe
    print(f"Die ersten {n} Fibonacci-Zahlen:")
    print(fib_list)
    print()
    
    # Summe
    summe = sum(fib_list)
    print(f"Summe: {summe}")
    print()
    
    # Verh√§ltnis zum Goldenen Schnitt
    if n > 1:
        verh√§ltnis = fib_list[-1] / fib_list[-2]
        goldener_schnitt = (1 + np.sqrt(5)) / 2
        print(f"Verh√§ltnis F({n}) / F({n-1}): {verh√§ltnis:.6f}")
        print(f"Goldener Schnitt œÜ:         {goldener_schnitt:.6f}")
        print(f"Differenz:                  {abs(verh√§ltnis - goldener_schnitt):.6f}")

import numpy as np
interact(fibonacci, n=(1, 30, 1))
```

### Erkl√§rung

- **Fibonacci-Rekurrenz:** \(F_n = F_{n-1} + F_{n-2}\) mit \(F_1 = 1, F_2 = 1\)
- **Implementierung:** Mit `a, b = b, a + b` schieben wir die Werte weiter (Speicher-effizient)
- **Goldener Schnitt:** Das Verh√§ltnis \(\frac{F_n}{F_{n-1}}\) n√§hert sich asymptotisch \(\phi = \frac{1+\sqrt{5}}{2} \approx 1.618\)

### Erweiterte Version (mit Visualisierung)

```
from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt

def fibonacci_visualisiert(n):
    """Fibonacci mit Balkendiagramm und Verh√§ltnis-Plot."""
    if n <= 0:
        print("‚ö†Ô∏è n muss mindestens 1 sein!")
        return
    
    # Fibonacci-Folge berechnen
    fib_list = []
    a, b = 1, 1
    for i in range(n):
        fib_list.append(a)
        a, b = b, a + b
    
    # Ausgabe
    print(f"Die ersten {n} Fibonacci-Zahlen: {fib_list}")
    print(f"Summe: {sum(fib_list)}")
    
    # Visualisierung
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    
    # Balkendiagramm
    ax1.bar(range(1, n+1), fib_list, color='steelblue')
    ax1.set_xlabel('Index')
    ax1.set_ylabel('Fibonacci-Zahl')
    ax1.set_title('Fibonacci-Folge')
    ax1.grid(True, alpha=0.3, axis='y')
    
    # Verh√§ltnis zum Goldenen Schnitt
    if n > 1:
        verh√§ltnisse = [fib_list[i+1] / fib_list[i] for i in range(n-1)]
        goldener_schnitt = (1 + np.sqrt(5)) / 2
        
        ax2.plot(range(2, n+1), verh√§ltnisse, 'b-o', label='F(n)/F(n-1)')
        ax2.axhline(y=goldener_schnitt, color='r', linestyle='--', label=f'œÜ ‚âà {goldener_schnitt:.4f}')
        ax2.set_xlabel('Index')
        ax2.set_ylabel('Verh√§ltnis')
        ax2.set_title('Konvergenz zum Goldenen Schnitt')
        ax2.legend()
        ax2.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()

interact(fibonacci_visualisiert, n=(1, 30, 1))
```

### Antworten auf Reflexionsfragen

- **F(10) vs F(11):** F(10) = 55, F(11) = 89, also F(11) > F(10)
- **Wie oft verdoppeln:** F(1) = 1, verdoppelt: 2, 4, 8, 16, 32, 64, 128, 256, 512, 1024, 2048, ... F(30) = 832040, braucht etwa 20 Verdoppelungen

## üß© Zusatzaufgabe: BMI-Rechner ‚Äì Beispiell√∂sung

### Vollst√§ndige L√∂sung

```
import ipywidgets as widgets
from ipywidgets import VBox, Output, interactive

def berechne_bmi(gewicht_kg, gr√∂√üe_cm):
    """Berechnet BMI und gibt Bewertung aus."""
    if gewicht_kg <= 0 or gr√∂√üe_cm <= 0:
        print("‚ö†Ô∏è Gewicht und Gr√∂√üe m√ºssen positiv sein!")
        return
    
    gr√∂√üe_m = gr√∂√üe_cm / 100
    bmi = gewicht_kg / (gr√∂√üe_m ** 2)
    
    # BMI-Kategorien
    if bmi < 18.5:
        kategorie = "Untergewicht"
        emoji = "üìâ"
    elif bmi < 25:
        kategorie = "Normalgewicht"
        emoji = "‚úÖ"
    elif bmi < 30:
        kategorie = "√úbergewicht"
        emoji = "‚ö†Ô∏è"
    else:
        kategorie = "Adipositas"
        emoji = "üö®"
    
    print(f"Gewicht: {gewicht_kg} kg")
    print(f"Gr√∂√üe:   {gr√∂√üe_cm} cm")
    print(f"BMI:     {bmi:.2f}")
    print(f"Status:  {emoji} {kategorie}")

# Widgets erstellen
weight_slider = widgets.FloatSlider(
    value=70,
    min=30,
    max=200,
    step=1,
    description='Gewicht (kg):'
)

height_slider = widgets.FloatSlider(
    value=170,
    min=100,
    max=230,
    step=1,
    description='Gr√∂√üe (cm):'
)

# Interactive verwenden
w = interactive(berechne_bmi, gewicht_kg=weight_slider, gr√∂√üe_cm=height_slider)

display(VBox([weight_slider, height_slider, w.children[-1]]))
```

## üß© Zusatzaufgabe: Einheitenumrechner ‚Äì Beispiell√∂sung

### Vollst√§ndige L√∂sung

```
import ipywidgets as widgets
from ipywidgets import VBox, HBox, Output

# Widgets
value_input = widgets.FloatText(
    value=1.0,
    description='Wert:',
    style={'description_width': '100px'}
)

from_unit = widgets.Dropdown(
    options=['km', 'm', 'cm', 'mm'],
    value='km',
    description='Von:',
    style={'description_width': '100px'}
)

to_unit = widgets.Dropdown(
    options=['km', 'm', 'cm', 'mm'],
    value='m',
    description='Zu:',
    style={'description_width': '100px'}
)

output = Output()

# Umrechnungsfaktoren zu Millimetern
UMRECHNUNGSFAKTOREN = {
    'km': 1_000_000,
    'm': 1_000,
    'cm': 10,
    'mm': 1
}

def umrechnen_einheiten(value, from_u, to_u):
    """Rechnet zwischen verschiedenen L√§ngeneinheiten um."""
    with output:
        output.clear_output()
        
        if value < 0:
            print("‚ö†Ô∏è Der Wert muss positiv sein!")
            return
        
        # Konvertiere zu Millimetern
        mm = value * UMRECHNUNGSFAKTOREN[from_u]
        
        # Konvertiere zu Zieleinheit
        ergebnis = mm / UMRECHNUNGSFAKTOREN[to_u]
        
        print(f"{value} {from_u} = {ergebnis:.6f} {to_u}")

def on_change(change):
    umrechnen_einheiten(value_input.value, from_unit.value, to_unit.value)

value_input.observe(on_change, names='value')
from_unit.observe(on_change, names='value')
to_unit.observe(on_change, names='value')

# Initial anzeigen
on_change(None)

# Layout
display(VBox([
    value_input,
    HBox([from_unit, to_unit]),
    output
]))
```

## üìã Zusammenfassung der L√∂sungsmuster

### Muster 1: Einfache Observer mit `observe()`

### Muster 2: Automatische Widgets mit `interact()`

### Muster 4: Reaktive Ausgabe mit `Output()`

```
output = Output()

def update_output(change):
    with output:
        output.clear_output()
        print("Neue Ausgabe")

widget.observe(update_output, names='value')
display(output)
```

### 1. **Bidirektionales Binding**

```
import ipywidgets as widgets

slider1 = widgets.IntSlider(value=5, min=0, max=10)
slider2 = widgets.IntSlider(value=5, min=0, max=10)

widgets.jslink((slider1, 'value'), (slider2, 'value'))
display(slider1, slider2)
```

### 2. **Bedingte Widget-Anzeige**

```
option = widgets.Dropdown(options=['Option A', 'Option B'])
text_field = widgets.Text(placeholder='Nur sichtbar bei Option A')
int_slider = widgets.IntSlider(description='Option B')

def toggle_visibility(change):
    if change['new'] == 'Option A':
        text_field.layout = widgets.Layout(display='block')
        int_slider.layout = widgets.Layout(display='none')
    else:
        text_field.layout = widgets.Layout(display='none')
        int_slider.layout = widgets.Layout(display='block')

option.observe(toggle_visibility, names='value')
toggle_visibility({'new': 'Option A'})

display(option, text_field, int_slider)
```
