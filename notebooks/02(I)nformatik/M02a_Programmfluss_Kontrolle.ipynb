## ðŸ“š Teil 1: Bedingte Anweisungen (`if`, `elif`, `else`)

### Was sind bedingte Anweisungen?

Stell dir vor, du hast einen Geldautomaten. Wenn genug Geld auf deinem Konto ist, gibt er Geld aus. Sonst bricht er ab. Das ist eine **Bedingung**.

In Python funktioniert das so:

````python
alter = 18

if alter >= 18:
    print("Du darfst wÃ¤hlen!")
else:
    print("Du musst noch ein paar Jahre warten.")
````

**Ausgabe:** `Du darfst wÃ¤hlen!`

### Syntax erklÃ¤rt

```
if <Bedingung>:
    # Code hier wird ausgefÃ¼hrt, wenn Bedingung WAHR ist
else:
    # Code hier wird ausgefÃ¼hrt, wenn Bedingung FALSCH ist
```

**Wichtig:** Der Code unter `if` muss **eingerÃ¼ckt** sein (4 Leerzeichen oder 1 Tab). Das ist keine optionale Formatierung â€“ Python benÃ¶tigt das!

### Beispiel: Temperaturkontrolle

````python
temperatur = 25

if temperatur > 30:
    print("Es ist sehr heiÃŸ!")
elif temperatur > 20:
    print("Angenehm warm")
elif temperatur > 10:
    print("Es ist kÃ¼hl")
else:
    print("Es ist kalt")
````

**Ausgabe:** `Angenehm warm`

### Was ist `elif`?

`elif` bedeutet â€želse if" (sonst wenn). Mit `elif` kannst du mehrere Bedingungen hintereinander prÃ¼fen. Das Programm prÃ¼ft sie der Reihe nach und stoppt, sobald es eine wahre Bedingung findet.

### Vergleichsoperatoren

Um Bedingungen zu schreiben, brauchst du diese Operatoren:

| Operator | Bedeutung | Beispiel |
|----------|-----------|----------|
| `==` | gleich | `5 == 5` â†’ True |
| `!=` | nicht gleich | `5 != 3` â†’ True |
| `>` | grÃ¶ÃŸer als | `10 > 5` â†’ True |
| `<` | kleiner als | `3 < 5` â†’ True |
| `>=` | grÃ¶ÃŸer oder gleich | `5 >= 5` â†’ True |
| `<=` | kleiner oder gleich | `3 <= 5` â†’ True |

### Vorsicht: `=` vs. `==`

- **`=`** ist ein Zuweisungsoperator (Variable setzen)
- **`==`** ist ein Vergleichsoperator (prÃ¼fen, ob zwei Werte gleich sind)

````python
x = 5      # x wird auf 5 gesetzt
x == 5     # prÃ¼ft, ob x gleich 5 ist (True)
````

### Mehrere Bedingungen kombinieren

Mit `and` und `or` kannst du mehrere Bedingungen zusammenfassen:

````python
alter = 20
hat_flugschein = True

if alter >= 18 and hat_flugschein:
    print("Du darfst allein fliegen!")
else:
    print("Das ist leider nicht mÃ¶glich.")
````

**Ausgabe:** `Du darfst allein fliegen!`

Oder mit `or`:

````python
tag = "Samstag"

if tag == "Samstag" or tag == "Sonntag":
    print("Wochenende!")
else:
    print("Arbeitstag")
````

**Ausgabe:** `Wochenende!`

## ðŸ”„ Teil 2: Schleifen

### Was sind Schleifen?

Schleifen wiederholen Code automatisch, ohne dass du ihn mehrmals tippen musst.

**Beispiel:** Du willst die Zahlen 1 bis 5 ausdrucken.

Ohne Schleife:

````python
print(1)
print(2)
print(3)
print(4)
print(5)
````

Mit Schleife:

````python
for i in range(1, 6):
    print(i)
````

Viel kÃ¼rzer, oder?

### `for`-Schleife â€“ Wiederholung mit bekannter Anzahl

Die `for`-Schleife ist perfekt, wenn du weiÃŸt, wie oft du wiederholen mÃ¶chtest.

````python
for i in range(5):
    print(f"Das ist Wiederholung Nummer {i}")
````

**Ausgabe:**
```
Das ist Wiederholung Nummer 0
Das ist Wiederholung Nummer 1
Das ist Wiederholung Nummer 2
Das ist Wiederholung Nummer 3
Das ist Wiederholung Nummer 4
```

### Verstehen: `range()`

`range()` erstellt eine Liste von Zahlen:

- `range(5)` â†’ 0, 1, 2, 3, 4 (von 0 bis 4)
- `range(1, 6)` â†’ 1, 2, 3, 4, 5 (von 1 bis 5)
- `range(0, 10, 2)` â†’ 0, 2, 4, 6, 8 (von 0 bis 9, in Schritten von 2)

````python
# Alle geraden Zahlen von 0 bis 8
for zahl in range(0, 10, 2):
    print(zahl)
````

**Ausgabe:** `0 2 4 6 8`

### `for`-Schleife mit Listen

Du kannst auch direkt Ã¼ber eine Liste iterieren:

````python
frÃ¼chte = ["Apfel", "Banane", "Kirsche"]

for frucht in frÃ¼chte:
    print(f"Ich mag {frucht}")
````

**Ausgabe:**
```
Ich mag Apfel
Ich mag Banane
Ich mag Kirsche
```

### `while`-Schleife â€“ Wiederholung mit Bedingung

Die `while`-Schleife wiederholt Code, **solange** eine Bedingung wahr ist. Sie ist perfekt, wenn du nicht weiÃŸt, wie oft wiederholt werden soll.

````python
zaehler = 0

while zaehler < 5:
    print(f"Zaehler: {zaehler}")
    zaehler = zaehler + 1  # oder: zaehler += 1
````

**Ausgabe:**
```
Zaehler: 0
Zaehler: 1
Zaehler: 2
Zaehler: 3
Zaehler: 4
```

### âš ï¸ Vorsicht: Endlosschleife!

Wenn die Bedingung in einer `while`-Schleife nie falsch wird, lÃ¤uft das Programm fÃ¼r immer. Das nennt man eine **Endlosschleife**:

````python
while True:  # Diese Bedingung ist IMMER wahr!
    print("Das geht fÃ¼r immer weiter...")
````

Diesen Code solltest du **nicht** ausfÃ¼hren, ohne einen Abbruch-Mechanismus einzubauen!

### `break` â€“ Schleife beenden

Mit `break` kannst du eine Schleife vorzeitig beenden:

````python
for i in range(10):
    if i == 5:
        break  # Schleife stoppt hier
    print(i)
````

**Ausgabe:** `0 1 2 3 4`

### `continue` â€“ Einen Durchlauf Ã¼berspringen

Mit `continue` springst du zum nÃ¤chsten Durchlauf:

````python
for i in range(5):
    if i == 2:
        continue  # Dieser Durchlauf wird Ã¼bersprungen
    print(i)
````

**Ausgabe:** `0 1 3 4`

## ðŸ’¡ Praxisbeispiel: Multiplikationstabelle

Hier kombinieren wir bedingte Anweisungen und Schleifen:

````python
# Multiplikationstabelle fÃ¼r das kleine 1x1
for i in range(1, 11):
    for j in range(1, 11):
        ergebnis = i * j
        print(f"{i} Ã— {j} = {ergebnis}", end="  ")
    print()  # Neue Zeile nach jeder Reihe
````

Dieses Programm:
1. Nutzt zwei verschachtelte Schleifen
2. Berechnet Produkte
3. Gibt sie formatiert aus
