## âœ… LÃ¶sung 1 â€“ NumPy Arrays erstellen

### Aufgabe
Erstelle verschiedene Arrays und untersuche deren Eigenschaften:

1. Erstelle ein Array mit den Zahlen von 0 bis 9.
2. Erzeuge ein Array mit 5 Werten zwischen 1 und 100 (gleichmÃ¤ÃŸig verteilt).
3. Erzeuge ein Array voller Zufallszahlen zwischen 0 und 1 (LÃ¤nge: 8).
4. PrÃ¼fe fÃ¼r jedes Array deren `shape` und `dtype`.

### LÃ¶sung

````python
import numpy as np

# Array 1: Zahlen von 0 bis 9
array1 = np.arange(0, 10)
print("Array 1:")
print(array1)
print(f"Shape: {array1.shape}, Dtype: {array1.dtype}\n")

# Array 2: 5 Werte zwischen 1 und 100 (gleichmÃ¤ÃŸig verteilt)
array2 = np.linspace(1, 100, 5)
print("Array 2:")
print(array2)
print(f"Shape: {array2.shape}, Dtype: {array2.dtype}\n")

# Array 3: 8 Zufallszahlen zwischen 0 und 1
array3 = np.random.random(8)
print("Array 3:")
print(array3)
print(f"Shape: {array3.shape}, Dtype: {array3.dtype}\n")
````

## âœ… LÃ¶sung 2 â€“ Vektorisierte Berechnungen

### Aufgabe
Nutze vektorisierte Operationen, um verschiedene Berechnungen mit ganzen Arrays durchzufÃ¼hren.

1. Multipliziere alle Elemente eines Arrays mit 2.
2. Addiere 10 zu jedem Element eines Arrays.
3. Wandle alle Zahlen eines Arrays in ihre Quadratwerte um.
4. Wende die Sinus-Funktion auf ein Array von Winkeln an (Array: Gradwerte von 0 bis 360 in 30er-Schritten).

### LÃ¶sung

````python
import numpy as np

# Definiere ein Base-Array
array = np.array([1, 2, 3, 4, 5])

# 1. Multipliziere mit 2
verdoppelt = array * 2
print("Verdoppelt:", verdoppelt)  # [2 4 6 8 10]

# 2. Addiere 10
plus_zehn = array + 10
print("Plus 10:", plus_zehn)  # [11 12 13 14 15]

# 3. Quadrate
quadrate = array ** 2
print("Quadrate:", quadrate)  # [1 4 9 16 25]

# 4. Sinus von Winkeln (Grad zu Radiant)
grad_array = np.arange(0, 360, 30)  # [0, 30, 60, ..., 330]
rad_array = np.radians(grad_array)   # Konvertiere in Radiant
sinus = np.sin(rad_array)
print("\nGradwerte:", grad_array)
print("Sinus-Werte:")
print(sinus)

# SchÃ¶ne Ausgabe mit Zuordnung
for grad, sin_wert in zip(grad_array, sinus):
    print(f"sin({grad:3d}Â°) = {sin_wert:7.4f}")
````

### ErklÃ¤rung

- **Vektorisierung**: `array * 2` multipliziert **alle** Elemente gleichzeitig, nicht einzeln in einer Schleife.
- **Potenz**: `array ** 2` quadriert jedes Element.
- **`np.radians()`**: Konvertiert Grad in Radiant, da `np.sin()` mit Radiant rechnet.
  - 0Â° = 0 rad
  - 90Â° = Ï€/2 rad
  - 180Â° = Ï€ rad
  - 270Â° = 3Ï€/2 rad
  - 360Â° = 2Ï€ rad
- **`zip()`**: Kombiniert zwei Arrays Element fÃ¼r Element zum Ausgeben.

## âœ… LÃ¶sung 4 â€“ Numerische Integration

### Aufgabe
Berechne numerisch den FlÃ¤cheninhalt unter der Sinuskurve im Intervall [0, Ï€].

1. Teile das Intervall in 20 Abschnitte.
2. Berechne die y-Werte als `np.sin(x)`.
3. SchÃ¤tze die FlÃ¤che mit der Summe der Rechtecke.

### LÃ¶sung

````python
import numpy as np

# Parameter
n = 20
x = np.linspace(0, np.pi, n)
y = np.sin(x)
dx = np.pi / (n - 1)

# Numerische FlÃ¤che
flaeche = np.sum(y) * dx

# Echte mathematische LÃ¶sung
echte_flaeche = 2.0  # âˆ«â‚€^Ï€ sin(x)dx = 2

print(f"Numerisch berechnete FlÃ¤che: {flaeche:.4f}")
print(f"Echte FlÃ¤che: {echte_flaeche:.4f}")
print(f"Fehler: {abs(flaeche - echte_flaeche):.4f}")
````

### Erweiterte LÃ¶sung mit Genauigkeitsanalyse

````python
import numpy as np
import matplotlib.pyplot as plt

# Berechne fÃ¼r verschiedene n
n_werte = [5, 10, 20, 50, 100, 500, 1000]
fehler = []

for n in n_werte:
    x = np.linspace(0, np.pi, n)
    y = np.sin(x)
    dx = np.pi / (n - 1)
    flaeche = np.sum(y) * dx
    fehler.append(abs(flaeche - 2.0))

# Visualisierung
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Fehler Ã¼ber n
axes[0].plot(n_werte, fehler, 'ro-', linewidth=2, markersize=8)
axes[0].set_xlabel('Anzahl der Punkte (n)')
axes[0].set_ylabel('Fehler')
axes[0].set_title('Numerischer Fehler vs. Anzahl der Punkte')
axes[0].grid(True)
axes[0].set_xscale('log')
axes[0].set_yscale('log')

# Beispiel mit n=20
n = 20
x = np.linspace(0, np.pi, n)
y = np.sin(x)
dx = np.pi / (n - 1)

x_fein = np.linspace(0, np.pi, 200)
y_fein = np.sin(x_fein)

axes[1].plot(x_fein, y_fein, 'b-', linewidth=2, label='sin(x)')
axes[1].bar(x, y, width=dx, alpha=0.3, edgecolor='black', label=f'Rechtecke (n={n})')
axes[1].fill_between(x, 0, y, alpha=0.2)
axes[1].set_xlabel('x')
axes[1].set_ylabel('sin(x)')
axes[1].set_title(f'Numerische Integration (n={n})')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()

# Tabelle mit Ergebnissen
print("\n" + "="*50)
print("Genauigkeit der numerischen Integration")
print("="*50)
print(f"{'n':<10} {'Berechnet':<15} {'Fehler':<15}")
print("-"*50)
for n, err in zip(n_werte, fehler):
    x = np.linspace(0, np.pi, n)
    y = np.sin(x)
    dx = np.pi / (n - 1)
    flaeche = np.sum(y) * dx
    print(f"{n:<10} {flaeche:<15.6f} {err:<15.6f}")
print("="*50)
````

### ErklÃ¤rung

**Die Methode:**
- Der Bereich [0, Ï€] wird in `n` Punkte unterteilt.
- Die Breite eines Rechtecks ist `dx = Ï€ / (n - 1)`.
- Die HÃ¶he jedes Rechtecks ist `sin(x)` an diesem Punkt.
- FlÃ¤che â‰ˆ Summe aller HÃ¶hen Ã— Breite.

**Warum ist das nÃ¼tzlich?**
- Manche Funktionen sind schwer mathematisch zu integrieren.
- Mit Numerik kÃ¶nnen wir jede beliebige Funktion nÃ¤herungsweise integrieren.
- Mit mehr Punkten wird die NÃ¤herung besser.

## âœ… LÃ¶sung 5 â€“ Bewegungssimulation

### Aufgabe
Simuliere den Flug eines Balls unter Einfluss der Schwerkraft (vereinfacht, ohne Sprung):

1. Start: Geschwindigkeit `v0 = 15 m/s` nach oben,
2. Simuliere die HÃ¶he fÃ¼r `t` von 0 bis 3 Sekunden (Zeitschritt: 0.05 s),
3. Plot die HÃ¶he Ã¼ber der Zeit.

### Einfache LÃ¶sung (analytisch)

````python
import numpy as np
import matplotlib.pyplot as plt

g = 9.81
v0 = 15

# Zeitarray
t = np.arange(0, 3, 0.05)

# Formel der gleichmÃ¤ÃŸig beschleunigten Bewegung
# y(t) = v0*t - 0.5*g*tÂ²
y = v0 * t - 0.5 * g * t**2

# Plot
plt.figure(figsize=(10, 6))
plt.plot(t, y, 'b-', linewidth=2, label='HÃ¶he')
plt.axhline(y=0, color='k', linestyle='-', linewidth=0.5)
plt.xlabel('Zeit (s)')
plt.ylabel('HÃ¶he (m)')
plt.title('Flug eines Balls (v0 = 15 m/s)')
plt.grid(True)
plt.legend()
plt.show()

# Statistik
print(f"Maximale HÃ¶he: {y.max():.2f} m")
print(f"Zeit bis zum Aufprall: {t[np.argmin(np.abs(y))]:.2f} s")
````

### Erweiterte LÃ¶sung (Simulation mit Zeitschritte)

````python
import numpy as np
import matplotlib.pyplot as plt

g = 9.81
v0 = 15
dt = 0.05
t_max = 3

# Zeitpunkte
t = np.arange(0, t_max, dt)
n = len(t)

# Arrays initialisieren
y = np.zeros(n)
v = np.zeros(n)

# Anfangsbedingungen
y[0] = 0
v[0] = v0

# Simulation Schritt fÃ¼r Schritt
for i in range(1, n):
    # Beschleunigung wirkt sich auf die Geschwindigkeit aus
    v[i] = v[i-1] - g * dt
    
    # Geschwindigkeit wirkt sich auf die Position aus
    y[i] = y[i-1] + v[i] * dt

# Plot
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# HÃ¶he Ã¼ber Zeit
axes[0, 0].plot(t, y, 'b-', linewidth=2)
axes[0, 0].axhline(y=0, color='k', linestyle='--', alpha=0.3)
axes[0, 0].fill_between(t, 0, y, alpha=0.2)
axes[0, 0].set_xlabel('Zeit (s)')
axes[0, 0].set_ylabel('HÃ¶he (m)')
axes[0, 0].set_title('HÃ¶he des Balls Ã¼ber Zeit')
axes[0, 0].grid(True)

# Geschwindigkeit Ã¼ber Zeit
axes[0, 1].plot(t, v, 'r-', linewidth=2)
axes[0, 1].axhline(y=0, color='k', linestyle='--', alpha=0.3)
axes[0, 1].set_xlabel('Zeit (s)')
axes[0, 1].set_ylabel('Geschwindigkeit (m/s)')
axes[0, 1].set_title('Geschwindigkeit des Balls')
axes[0, 1].grid(True)

# Phase-Raum: v(y)
axes[1, 0].plot(y, v, 'g-', linewidth=2)
axes[1, 0].scatter(y[0], v[0], color='blue', s=100, label='Start', zorder=5)
axes[1, 0].scatter(y[-1], v[-1], color='red', s=100, label='Ende', zorder=5)
axes[1, 0].set_xlabel('HÃ¶he (m)')
axes[1, 0].set_ylabel('Geschwindigkeit (m/s)')
axes[1, 0].set_title('Phase-Raum (Geschwindigkeit vs. HÃ¶he)')
axes[1, 0].legend()
axes[1, 0].grid(True)

# Beschleunigung (konstant)
a = np.full_like(t, -g)
axes[1, 1].plot(t, a, 'purple', linewidth=2)
axes[1, 1].set_xlabel('Zeit (s)')
axes[1, 1].set_ylabel('Beschleunigung (m/sÂ²)')
axes[1, 1].set_title('Beschleunigung (konstant = -g)')
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()

# Ausgabe
print("="*50)
print("Bewegungssimulation: Wurf nach oben")
print("="*50)
print(f"Anfangsgeschwindigkeit: {v0} m/s")
print(f"Maximale HÃ¶he: {y.max():.2f} m")
print(f"Zeit zur max. HÃ¶he: {t[np.argmax(y)]:.2f} s")
print(f"Zeit bis Aufprall: {t[np.argmin(np.abs(y[y < 0.1]))]:.2f} s")
print(f"Endgeschwindigkeit (bei t=3s): {v[-1]:.2f} m/s")
````

### LÃ¶sung mit BodenberÃ¼hrung (erweitert)

````python
import numpy as np
import matplotlib.pyplot as plt

g = 9.81
v0 = 20   # Etwas hÃ¶her fÃ¼r sichtbare SprÃ¼nge
dt = 0.01
t_max = 4

t = np.arange(0, t_max, dt)
n = len(t)

y = np.zeros(n)
v = np.zeros(n)

y[0] = 0
v[0] = v0

# Simulation mit BodenberÃ¼hrung
for i in range(1, n):
    v[i] = v[i-1] - g * dt
    y[i] = y[i-1] + v[i] * dt
    
    # Wenn Ball den Boden berÃ¼hrt, springt er (mit Energieverlust)
    if y[i] < 0:
        y[i] = -y[i]          # HÃ¶he spiegeln
        v[i] = -v[i] * 0.85   # Geschwindigkeit umkehren, 85% Energie behalten

# Plot nur HÃ¶he
plt.figure(figsize=(12, 6))
plt.plot(t, y, 'b-', linewidth=2)
plt.axhline(y=0, color='k', linestyle='-', linewidth=1)
plt.fill_between(t, 0, y, alpha=0.2)
plt.xlabel('Zeit (s)')
plt.ylabel('HÃ¶he (m)')
plt.title('Springender Ball (mit Energieverlust)')
plt.grid(True)
plt.show()

print(f"Maximale HÃ¶he beim 1. Sprung: {y[y > 0.1].max():.2f} m")
print(f"Anzahl der SprÃ¼nge (Aufpralls): {np.sum(np.diff(y) < -0.5)}")
````

### ErklÃ¤rung

**Physikalische Gleichungen:**
- **Geschwindigkeit**: v(t) = vâ‚€ - gÂ·t
- **HÃ¶he**: y(t) = vâ‚€Â·t - Â½Â·gÂ·tÂ²

**Die Simulation:**
1. FÃ¼r jeden Zeitschritt wird die neue Geschwindigkeit berechnet.
2. Mit der neuen Geschwindigkeit wird die neue Position berechnet.
3. Der Prozess wiederholt sich fÃ¼r alle Zeitpunkte.

**BodenberÃ¼hrung:**
- Wenn die HÃ¶he negativ wird (unter Boden), wird sie gespiegelt.
- Die Geschwindigkeit wird negativ (Richtungswechsel).
- Mit Faktor 0.85 verliert der Ball bei jedem Aufprall 15% seiner Energie.

### Fehler 1: Falsche Array-Form

````python
# FALSCH:
x = np.linspace(0, 2*np.pi, 10)
y = np.sin(x)
result = x + 1  # Funktioniert
# ABER:
z = np.array([[1, 2], [3, 4]])
result = x + z  # ERROR: Formen passen nicht!

# RICHTIG: Forms mÃ¼ssen kompatibel sein oder Broadcasting-Regeln erfÃ¼llen
````

### Fehler 2: Grad statt Radiant

````python
# FALSCH:
angle = 90
result = np.sin(angle)  # â‰ˆ 0.893, nicht 1!

# RICHTIG:
angle_deg = 90
angle_rad = np.radians(angle_deg)
result = np.sin(angle_rad)  # 1.0
````

### Fehler 3: Alte Variable verwenden

````python
# FALSCH:
array = [1, 2, 3]
array = np.array(array)
print(array.mean())  # Funktioniert, aber unbewusst

# BESSER: Klar, dass array jetzt ein NumPy-Array ist
````

### Fehler 4: Vergessene Dimensionen

````python
# FALSCH:
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
result = a.reshape(3, 1) + b  # Shape (3,1) + (3,) = (3,3)

# RICHTIG: Bewusst mit Shapes arbeiten
print(a.reshape(3, 1).shape)  # (3, 1)
print(result.shape)  # (3, 3)
````
