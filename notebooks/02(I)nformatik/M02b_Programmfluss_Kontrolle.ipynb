## ðŸ”¥ Aufgabe 10: `while`-Schleife â€“ Benutzereingabe

### Aufgabenstellung
Schreibe ein Programm, das die Benutzer*in auffordert, eine Zahl einzugeben. Das Programm soll wiederholt fragen, bis die richtige Zahl (z.B. 42) eingegeben wird.

````python
richtige_zahl = 42
geraten = 0

while geraten != richtige_zahl:
    geraten = int(input("Rate eine Zahl zwischen 1 und 100: "))
    
    if geraten < richtige_zahl:
        print("Zu klein!")
    elif geraten > richtige_zahl:
        print("Zu groÃŸ!")
    else:
        print("Richtig! Du hast gewonnen!")
````

### Aufgabe fÃ¼r dich
FÃ¼hre das Programm aus und teste es. Was passiert, wenn du die falsche Zahl eingibst?

### LÃ¶sungshinweis
> Das Programm lÃ¤uft in einer `while`-Schleife, solange `geraten != richtige_zahl` ist. Wenn du die richtige Zahl eingibst, wird die Bedingung falsch und die Schleife endet.
