## **L√∂sung zu Aufgabe 2: Rechteck berechnen**

### Aufgabe
Schreibe eine Funktion `rechteck_berechne(breite, hoehe)`, die **Fl√§che und Umfang** eines Rechtecks zur√ºckgibt (beide Werte!).

### L√∂sung

````python
def rechteck_berechne(breite, hoehe):
    """
    Berechnet Fl√§che und Umfang eines Rechtecks.
    R√ºckgabe: (flaehe, umfang)
    """
    flaehe = breite * hoehe
    umfang = 2 * (breite + hoehe)
    return flaehe, umfang

# Test
b, h = 5, 3
f, u = rechteck_berechne(b, h)
print(f"Rechteck {b}x{h}: Fl√§che = {f}, Umfang = {u}")

# Weitere Tests
breite, hoehe = 10, 7
flaehe, umfang = rechteck_berechne(breite, hoehe)
print(f"Rechteck {breite}x{hoehe}: Fl√§che = {flaehe}, Umfang = {umfang}")
````

**Ausgabe:**
```
Rechteck 5x3: Fl√§che = 15, Umfang = 16
Rechteck 10x7: Fl√§che = 70, Umfang = 34
```

### Erkl√§rung
- Eine Funktion kann **mehrere Werte** als Tupel zur√ºckgeben: `return wert1, wert2`.
- Beim Aufruf k√∂nnen wir die R√ºckgabewerte **auspacken** in separate Variablen: `f, u = rechteck_berechne(...)`.
- Dies ist ein gro√üer Vorteil von Python gegen√ºber vielen anderen Programmiersprachen.

## **L√∂sung zu Aufgabe 3: Liste quadrieren**

### Aufgabe
Erstelle eine Funktion `quadriere_liste(liste)`, die alle Elemente einer Zahl-Liste quadriert und als neue Liste zur√ºckgibt.

### L√∂sung (Variante 1: Mit for-Schleife)

````python
def quadriere_liste(liste):
    """Quadriert alle Elemente einer Liste."""
    neue_liste = []
    for zahl in liste:
        neue_liste.append(zahl ** 2)
    return neue_liste

# Test
zahlen = [1, 2, 3, 4, 5]
quadriert = quadriere_liste(zahlen)
print(f"Original: {zahlen}")
print(f"Quadriert: {quadriert}")
````

**Ausgabe:**
```
Original: [1, 2, 3, 4, 5]
Quadriert: [1, 4, 9, 16, 25]
```

### L√∂sung (Variante 2: Mit List-Comprehension - eleganter!)

````python
def quadriere_liste(liste):
    """Quadriert alle Elemente einer Liste (elegante Variante)."""
    return [zahl ** 2 for zahl in liste]

# Test (identisch wie oben)
zahlen = [1, 2, 3, 4, 5]
quadriert = quadriere_liste(zahlen)
print(f"Original: {zahlen}")
print(f"Quadriert: {quadriert}")
````

### Erkl√§rung
- **Variante 1** ist sehr verst√§ndlich: Wir erstellen eine leere Liste und f√ºgen quadrierte Werte hinzu.
- **Variante 2** ist ein Python-Idiom namens **List-Comprehension**: Sie ist k√ºrzer und oft schneller.
- Beide Varianten produzieren das gleiche Ergebnis.
- Beachte: Der `**`-Operator bedeutet "Potenzieren".

## **L√∂sung zu Aufgabe 4: Lokaler vs. globaler Scope**

### Aufgabe
Gegeben ist folgender Code ‚Äî was wird ausgegeben?

````python
x = 10

def scope_test():
    x = 20
    print(f"x in Funktion: {x}")

scope_test()
print(f"x au√üerhalb: {x}")
````

### L√∂sung

````python
x = 10

def scope_test():
    x = 20  # Dies ist eine NEUE, lokale Variable namens x
    print(f"x in Funktion: {x}")

scope_test()
print(f"x au√üerhalb: {x}")
````

**Ausgabe:**
```
x in Funktion: 20
x au√üerhalb: 10
```

### Erkl√§rung
- Die **globale Variable `x`** hat den Wert `10`.
- Innerhalb der Funktion wird eine **neue, lokale Variable `x`** mit Wert `20` erstellt.
- Diese lokale Variable **√ºberlagert** die globale nur innerhalb der Funktion.
- Nach der Funktion ist die globale Variable unver√§ndert.

### Was passiert mit `global`?

````python
x = 10

def scope_test_global():
    global x  # Sage: "Ich m√∂chte die GLOBALE Variable x √§ndern"
    x = 20
    print(f"x in Funktion: {x}")

scope_test_global()
print(f"x au√üerhalb: {x}")
````

**Ausgabe:**
```
x in Funktion: 20
x au√üerhalb: 20
```

### Erkl√§rung
- Mit dem Schl√ºsselwort `global` teilen wir Python mit, dass wir die **globale Variable** √§ndern m√∂chten.
- Jetzt modifizieren wir nicht eine lokale, sondern die **tats√§chliche globale Variable**.
- Nach der Funktion hat `x` den neuen Wert `20`.

‚ö†Ô∏è **Best Practice:** Vermeide `global` wo m√∂glich! Nutze stattdessen Parametern und R√ºckgabewerte.

## **L√∂sung zu Aufgabe 5: Fibonacci als Funktion**

### Aufgabe
Erstelle eine Funktion `fibonacci(n)`, die die n-te Fibonacci-Zahl berechnet. Nutze eine Schleife, keine Rekursion.

### L√∂sung

````python
def fibonacci(n):
    """
    Berechnet die n-te Fibonacci-Zahl iterativ.
    F(0)=0, F(1)=1, F(n)=F(n-1)+F(n-2)
    """
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        a, b = 0, 1  # Starte mit F(0) und F(1)
        for _ in range(2, n + 1):
            a, b = b, a + b  # Verschiebe: b wird neue a, a+b wird neue b
        return b

# Test
print("Fibonacci-Reihe:")
for i in range(10):
    print(f"F({i}) = {fibonacci(i)}")

# Gr√∂√üere Zahlen testen
print(f"\nF(20) = {fibonacci(20)}")
print(f"F(30) = {fibonacci(30)}")
````

**Ausgabe:**
```
Fibonacci-Reihe:
F(0) = 0
F(1) = 1
F(2) = 1
F(3) = 2
F(4) = 3
F(5) = 5
F(6) = 8
F(7) = 13
F(8) = 21
F(9) = 34

F(20) = 6765
F(30) = 832040
```

### Erkl√§rung
- **Basisfall:** F(0) = 0, F(1) = 1 (m√ºssen speziell behandelt werden).
- **Schleife:** Mit `for _ in range(2, n+1)` iterieren wir von 2 bis n.
- **Tupel-Zuweisung:** `a, b = b, a + b` ist Pythons Weg, zwei Variablen gleichzeitig zu aktualisieren.
  - Dies ist equivalent zu:
    ```python
    temp = a + b
    a = b
    b = temp
    ```
- **Effizienz:** Diese iterative L√∂sung ist viel schneller als Rekursion!

### Vergleich mit Rekursion

````python
def fibonacci_rekursiv(n):
    """Fibonacci rekursiv - WARNUNG: Sehr langsam!"""
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci_rekursiv(n - 1) + fibonacci_rekursiv(n - 2)

# Zeitvergleich
import time

# Iterativ: sehr schnell
start = time.time()
result_iter = fibonacci(35)
time_iter = time.time() - start

# Rekursiv: sehr langsam f√ºr n=35!
# (Kommentiert, um die Zeit zu sparen)
# start = time.time()
# result_rek = fibonacci_rekursiv(35)
# time_rek = time.time() - start

print(f"Iterativ F(35) = {result_iter}, Zeit: {time_iter:.6f}s")
# print(f"Rekursiv F(35) = {result_rek}, Zeit: {time_rek:.6f}s")
````

## **L√∂sung zu Aufgabe 6: Analyse von Messdaten**

### Aufgabe
Schreibe drei Funktionen: `minimum`, `maximum`, `mittelwert` und kombiniere sie.

### L√∂sung

````python
def minimum(liste):
    """Gibt das kleinste Element einer Liste zur√ºck."""
    return min(liste)

def maximum(liste):
    """Gibt das gr√∂√üte Element einer Liste zur√ºck."""
    return max(liste)

def mittelwert(liste):
    """Gibt den Durchschnitt einer Liste zur√ºck."""
    return sum(liste) / len(liste)

def analyse(daten):
    """F√ºhrt eine umfassende Analyse durch."""
    ergebnis = {
        "Minimum": minimum(daten),
        "Maximum": maximum(daten),
        "Mittelwert": mittelwert(daten),
        "Anzahl": len(daten)
    }
    return ergebnis

# Test mit Messdaten
messwerte = [10, 15, 12, 18, 16, 14, 19, 13, 17, 15]

print("Messdaten:", messwerte)
print()

ergebnisse = analyse(messwerte)
for key, value in ergebnisse.items():
    print(f"{key}: {value}")
````

**Ausgabe:**
```
Messdaten: [10, 15, 12, 18, 16, 14, 19, 13, 17, 15]

Minimum: 10
Maximum: 19
Mittelwert: 15.0
Anzahl: 10
```

### Verbesserte L√∂sung: Mit Standardabweichung

````python
import math

def minimum(liste):
    return min(liste)

def maximum(liste):
    return max(liste)

def mittelwert(liste):
    return sum(liste) / len(liste)

def standardabweichung(liste):
    """Berechnet die Standardabweichung."""
    mw = mittelwert(liste)
    summe_abweichungen = sum((x - mw) ** 2 for x in liste)
    varianz = summe_abweichungen / len(liste)
    return math.sqrt(varianz)

def analyse(daten):
    """Umfassende statistische Analyse."""
    return {
        "Minimum": minimum(daten),
        "Maximum": maximum(daten),
        "Mittelwert": mittelwert(daten),
        "Standardabweichung": standardabweichung(daten),
        "Anzahl": len(daten),
        "Summe": sum(daten)
    }

# Test
messwerte = [10, 15, 12, 18, 16, 14, 19, 13, 17, 15]
ergebnisse = analyse(messwerte)

for key, value in ergebnisse.items():
    if isinstance(value, float):
        print(f"{key}: {value:.2f}")
    else:
        print(f"{key}: {value}")
````

**Ausgabe:**
```
Minimum: 10
Maximum: 19
Mittelwert: 15.00
Standardabweichung: 2.55
Anzahl: 10
Summe: 150
```

### Erkl√§rung
- Wir **modulieren** die Analyse in kleine, spezialisierte Funktionen.
- Jede Funktion hat eine **klare Aufgabe**.
- Diese k√∂nnen leicht **getestet und wiederverwendet** werden.
- Die `analyse()`-Funktion **kombiniert alles** in ein Dictionary.
- Mit `isinstance()` pr√ºfen wir den Datentyp, um Floats sch√∂ner zu formatieren.

## **L√∂sung zu Aufgabe 7: Variablen-Scope visualisieren**

### Aufgabe
Schreibe eine Funktion mit lokaler und globaler Variable. √Ñndere die globale Variable innerhalb der Funktion.

### L√∂sung

````python
# Globale Variable
zaehler = 0

def erhoehe_zaehler(menge=1):
    """Erh√∂ht den globalen Z√§hler."""
    global zaehler  # Wichtig: global deklarieren!
    zaehler += menge
    print(f"Z√§hler erh√∂ht um {menge}. Neuer Wert: {zaehler}")

def zuruecksetzen():
    """Setzt den globalen Z√§hler zur√ºck."""
    global zaehler
    zaehler = 0
    print("Z√§hler zur√ºckgesetzt auf 0")

def zeige_zaehler():
    """Zeigt den aktuellen Z√§hlerwert."""
    print(f"Aktueller Z√§hlerwert: {zaehler}")

# Demo
print("=== Demo: Globale Variable mit Funktionen ===\n")

zeige_zaehler()  # 0

erhoehe_zaehler(5)   # 5
erhoehe_zaehler(3)   # 8
erhoehe_zaehler(2)   # 10

zeige_zaehler()  # 10

zuruecksetzen()  # 0
zeige_zaehler()  # 0
````

**Ausgabe:**
```
=== Demo: Globale Variable mit Funktionen ===

Aktueller Z√§hlerwert: 0
Z√§hler erh√∂ht um 5. Neuer Wert: 5
Z√§hler erh√∂ht um 3. Neuer Wert: 8
Z√§hler erh√∂ht um 2. Neuer Wert: 10
Aktueller Z√§hlerwert: 10
Z√§hler zur√ºckgesetzt auf 0
Aktueller Z√§hlerwert: 0
```

### Was passiert, wenn wir `global` weglassen?

````python
zaehler = 0

def erhoehe_zaehler_FALSCH(menge=1):
    zaehler += menge  # ‚ùå FEHLER: UnboundLocalError!
    print(f"Z√§hler: {zaehler}")

erhoehe_zaehler_FALSCH(5)
````

**Fehler:**
```
UnboundLocalError: local variable 'zaehler' referenced before assignment
```

### Erkl√§rung des Fehlers
- Ohne `global` versucht Python, `zaehler` als **lokale Variable** zu behandeln.
- Aber `zaehler` wurde lokal noch nicht definiert (vor der Zuweisung gelesen).
- Python findet einen Widerspruch und wirft einen Fehler.

### Best Practice: Vermeide `global`!

Statt `global` zu verwenden, ist es besser, **Parameter und R√ºckgabewerte** zu nutzen:

````python
def erhoehe_zaehler_BESSER(zaehler, menge=1):
    """Erh√∂ht den Z√§hler und gibt den neuen Wert zur√ºck."""
    zaehler += menge
    return zaehler

# Verwendung
z = 0
z = erhoehe_zaehler_BESSER(z, 5)  # z = 5
z = erhoehe_zaehler_BESSER(z, 3)  # z = 8
z = erhoehe_zaehler_BESSER(z, 2)  # z = 10
print(z)  # 10
````

**Vorteile:**
- ‚úÖ Keine versteckten Effekte (keine globalen Modifikationen)
- ‚úÖ Funktionen sind **unabh√§ngiger und testbarer**
- ‚úÖ Code ist **leichter zu verstehen**

## üìö Zusammenfassung

| Konzept | Bedeutung | Beispiel |
|---------|-----------|---------|
| **Parameter** | Eingaben einer Funktion | `def f(x, y):` |
| **R√ºckgabewert** | Ausgabe einer Funktion | `return x + y` |
| **Lokale Variable** | Existiert nur in der Funktion | Definiert in der Funktion |
| **Globale Variable** | Existiert √ºberall | Definiert au√üerhalb aller Funktionen |
| **`global`** | √Ñndert globale Variablen | Nutze sparsam! |
| **Modularisierung** | Code in wiederverwendbare Teile teilen | Mehrere kleine Funktionen statt einer gro√üen |
| **Rekursion** | Funktion ruft sich selbst auf | Langsam, aber elegant |
| **Iteration** | Schleife statt Rekursion | Schneller f√ºr Fibonacci |
