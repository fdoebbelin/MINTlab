## üíª Teil 2: Datentypen und erste Operationen

Python kennt verschiedene **Datentypen** ‚Äì das sind sozusagen die ‚ÄûRohstoffe", mit denen wir arbeiten:

### Ganze Zahlen (Integer)

````python
# Integer: Ganze Zahlen ohne Dezimalstellen
x = 42
print(x)
print(type(x))  # type() zeigt den Datentyp
````

**Erkl√§rung:**
- `x = 42` speichert die Zahl 42 in der **Variable** `x`
- `print(x)` gibt den Wert aus
- `type(x)` verr√§t uns, dass `x` vom Typ ‚Äûint" (integer) ist

### Dezimalzahlen (Float)

````python
# Float: Zahlen mit Dezimalstellen
pi = 3.14159
durchmesser = 2.5
print(pi)
print(type(pi))
````

### Zeichenketten (String)

````python
# String: Text in Anf√ºhrungszeichen
name = "Alice"
stadt = 'Berlin'
print(name)
print(type(name))
````

**Tipp:** Du kannst einfache (`'...'`) oder doppelte (`"..."`) Anf√ºhrungszeichen verwenden ‚Äì das Ergebnis ist identisch.

### Wahrheitswerte (Boolean)

````python
# Boolean: True oder False
ist_aktiviert = True
ist_feuer = False
print(ist_aktiviert)
print(type(ist_aktiviert))
````

## üî¢ Teil 3: Operationen und Berechnungen

Mit diesen Datentypen kannst du **Operationen** durchf√ºhren:

### Arithmetische Operationen

````python
# Addition, Subtraktion, Multiplikation, Division
a = 10
b = 3

addition = a + b
subtraktion = a - b
multiplikation = a * b
division = a / b  # Ergebnis ist immer ein Float
ganzzahlige_division = a // b  # Ganzzahlige Division
rest = a % b  # Modulo: Der Rest nach Division

print(f"10 + 3 = {addition}")
print(f"10 - 3 = {subtraktion}")
print(f"10 * 3 = {multiplikation}")
print(f"10 / 3 = {division}")
print(f"10 // 3 = {ganzzahlige_division}")
print(f"10 % 3 = {rest}")
````

**Was ist `f-String`?**  
Die `f` vor den Anf√ºhrungszeichen bedeutet ‚Äûformatierter String". Mit `{variable}` kannst du Variablen direkt in den Text einbauen. Sehr praktisch!

### Potenzen

````python
# Potenzieren mit **
quadrat = 5 ** 2
kubik = 2 ** 3

print(f"5 hoch 2 = {quadrat}")
print(f"2 hoch 3 = {kubik}")
````

### Operationen mit Strings

````python
# Strings kann man verbinden (concatenieren)
vorname = "Max"
nachname = "Mustermann"
vollname = vorname + " " + nachname
print(vollname)

# Oder mit f-Strings (noch eleganter):
print(f"Guten Tag, {vorname} {nachname}!")

# Strings kann man auch ‚Äûmultiplizieren"
wiederholung = "Ha" * 3
print(wiederholung)  # Ergebnis: HaHaHa
````

## üìä Teil 4: Variablen verstehen

Eine **Variable** ist wie eine beschriftete Box:
- Die **Beschriftung** ist der Name (z. B. `temperatur`)
- Der **Inhalt** ist der Wert (z. B. `25.5`)

````python
# Variablen definieren und nutzen
temperatur = 22.5
feuchte = 65
stadt = "M√ºnchen"

print(f"Temperatur in {stadt}: {temperatur}¬∞C")
print(f"Luftfeuchte: {feuchte}%")

# Variablen k√∂nnen ver√§ndert werden
temperatur = 23.1
print(f"Neue Temperatur: {temperatur}¬∞C")

# Eine Variable kann den Wert einer anderen annehmen
kopie = temperatur
print(f"Kopie der Temperatur: {kopie}¬∞C")
````

**Wichtig:** Python ist eine **dynamisch typisierte** Sprache. Das hei√üt: Eine Variable kann zuerst ein Integer sein, sp√§ter aber einen String enthalten ‚Äì Python passt sich an!

````python
wert = 42
print(type(wert))  # int

wert = "Jetzt bin ich ein Text"
print(type(wert))  # str
````

## üß™ Teil 5: Eingaben von Benutzern

Mit `input()` kannst du den Benutzer Daten eingeben lassen:

````python
# Der Computer stellt eine Frage, der Benutzer antwortet
name = input("Wie hei√üt du? ")
print(f"Sch√∂n, dich kennenzulernen, {name}!")

# Benutzereingaben sind immer Strings ‚Äì auch Zahlen!
alter_als_string = input("Wie alt bist du? ")
print(f"Du hast eingegeben: {alter_als_string}")
print(f"Das ist vom Typ: {type(alter_als_string)}")

# Um mit Zahlen zu rechnen, m√ºssen wir sie umwandeln (konvertieren)
alter_als_zahl = int(alter_als_string)
print(f"N√§chstes Jahr bist du: {alter_als_zahl + 1} Jahre alt")
````

**Achtung:** `input()` gibt immer einen String zur√ºck ‚Äì auch wenn der Benutzer Zahlen eingibt! Wenn du rechnen m√∂chtest, musst du `int()` oder `float()` verwenden, um die Typen zu konvertieren.

## üí° Teil 7: Kleine interaktive Demonstration

````python
# Berechnung des Umfangs und der Fl√§che eines Kreises

import math

# Eingabe vom Benutzer
radius = float(input("Gib den Radius eines Kreises ein (in cm): "))

# Berechnungen
umfang = 2 * math.pi * radius
flaeche = math.pi * radius ** 2

# Ausgabe
print(f"\n--- Kreis-Berechnung ---")
print(f"Radius: {radius} cm")
print(f"Umfang: {umfang:.2f} cm")
print(f"Fl√§che: {flaeche:.2f} cm¬≤")
````

**Neue Konzepte:**
- `import math`: Die Bibliothek `math` wird geladen ‚Äì sie enth√§lt Konstanten wie `math.pi` und Funktionen
- `:.2f`: Formatierung ‚Äì zeigt die Zahl mit 2 Dezimalstellen an
