## ‚úèÔ∏è Aufgabe 3: Arithmetische Operationen

### Ziel
Du rechnest mit Zahlen und verstehst die verschiedenen Divisionsoperatoren.

### Aufgabenstellung

F√ºhre diese Code-Zelle aus:

````
# Zwei Zahlen definieren
a = 17
b = 5

# Verschiedene Operationen
print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
print(f"{a} // {b} = {a // b}")  # Ganzzahlige Division
print(f"{a} % {b} = {a % b}")    # Modulo (Rest)
print(f"{a} ** {b} = {a ** b}")  # Potenz
````

### L√∂sungshinweis
> **Ganzzahlige Division (`//`):** Das Ergebnis wird auf die n√§chst kleinere ganze Zahl abgerundet.  
> **Modulo (`%`):** Das Ergebnis ist der Rest nach einer Division.  
> Beispiel: 17 √∑ 5 = 3 Rest 2, also 17 // 5 = 3 und 17 % 5 = 2
