## ðŸ“š Teil 1: Grundlagen â€“ Arrays statt Listen

### Warum NumPy Arrays besser sind als Listen

**Normale Python-Listen:**

```python
# Eine Liste mit Zahlen
meine_liste = [1, 2, 3, 4, 5]
meine_liste[0]  # Zugriff auf ein Element: 1
```

**NumPy Arrays:**

```python
import numpy as np

# Ein NumPy Array mit den gleichen Zahlen
mein_array = np.array([1, 2, 3, 4, 5])
mein_array[0]  # Zugriff auf ein Element: 1
```

Das sieht Ã¤hnlich aus â€“ aber der Unterschied liegt in der Geschwindigkeit und den MÃ¶glichkeiten!

### Demonstration: Listen vs. Arrays

```python
import numpy as np
import time

# GroÃŸe Liste
groÃŸe_liste = list(range(1000000))

# GroÃŸes Array
groÃŸes_array = np.array(range(1000000))

# Alle Elemente der Liste verdoppeln (langsam)
start = time.time()
ergebnis_liste = [x * 2 for x in groÃŸe_liste]
zeit_liste = time.time() - start

# Alle Elemente des Arrays verdoppeln (schnell)
start = time.time()
ergebnis_array = groÃŸes_array * 2
zeit_array = time.time() - start

print(f"Listen-Methode: {zeit_liste:.4f} Sekunden")
print(f"Array-Methode:  {zeit_array:.4f} Sekunden")
print(f"NumPy ist etwa {zeit_liste / zeit_array:.0f}x schneller!")
```

**Warum ist NumPy schneller?** NumPy Arrays werden in C (einer schnellen Programmiersprache) verarbeitet, wÃ¤hrend Python-Listen in reinem Python arbeiten.

## ðŸ”¨ Teil 2: Arrays erstellen

### Verschiedene Wege, Arrays zu erstellen

```python
import numpy as np

# 1. Aus einer Liste
array1 = np.array([1, 2, 3, 4, 5])
print("Aus Liste:", array1)

# 2. Mit np.arange() - wie range(), aber fÃ¼r Arrays
array2 = np.arange(0, 10, 2)  # Von 0 bis 10 in 2er-Schritten
print("Mit arange():", array2)

# 3. Mit np.linspace() - gleichmÃ¤ÃŸig verteilte Punkte
array3 = np.linspace(0, 10, 5)  # 5 Punkte zwischen 0 und 10
print("Mit linspace():", array3)

# 4. Mit np.zeros() - Array voll mit Nullen
array4 = np.zeros(5)
print("Nullen:", array4)

# 5. Mit np.ones() - Array voll mit Einsen
array5 = np.ones(5)
print("Einsen:", array5)

# 6. Mit np.random.random() - Zufallszahlen zwischen 0 und 1
array6 = np.random.random(5)
print("Zufallszahlen:", array6)
```

### Eigenschaften eines Arrays erforschen

```python
import numpy as np

# Ein Array erstellen
daten = np.array([1, 2, 3, 4, 5])

# GrÃ¶ÃŸe (LÃ¤nge)
print("Anzahl der Elemente:", len(daten))
print("Form (Shape):", daten.shape)  # (5,) bedeutet ein 1D-Array mit 5 Elementen
print("Datentyp:", daten.dtype)  # int64 oder Ã¤hnliches

# Statistische Eigenschaften
print("Summe:", daten.sum())
print("Durchschnitt:", daten.mean())
print("Minimum:", daten.min())
print("Maximum:", daten.max())
```

**Was ist ein 1D-, 2D- oder 3D-Array?**

```python
import numpy as np

# 1D-Array: eine Reihe von Zahlen
array_1d = np.array([1, 2, 3, 4, 5])
print("1D Shape:", array_1d.shape)  # (5,)

# 2D-Array: eine Matrix (Tabelle)
array_2d = np.array([[1, 2, 3],
                     [4, 5, 6]])
print("2D Shape:", array_2d.shape)  # (2, 3) - 2 Reihen, 3 Spalten

# 3D-Array: ein WÃ¼rfel von Zahlen
array_3d = np.array([[[1, 2], [3, 4]],
                     [[5, 6], [7, 8]]])
print("3D Shape:", array_3d.shape)  # (2, 2, 2)
```

## âš¡ Teil 3: Vektorisierung â€“ Die Kraft von NumPy

**Vektorisierung** bedeutet: Operationen auf ganze Arrays auf einmal anwenden, statt Element fÃ¼r Element mit Schleifen.

### Beispiel 1: Einfache Rechenoperationen

```python
import numpy as np

# Array erstellen
zahlen = np.array([1, 2, 3, 4, 5])

# Mit Python-Listen mÃ¼sste man eine Schleife schreiben:
# ergebnis = []
# for z in zahlen:
#     ergebnis.append(z * 2)

# Mit NumPy: einfach das Array multiplizieren!
verdoppelt = zahlen * 2
print("Original:", zahlen)
print("Verdoppelt:", verdoppelt)

# Das funktioniert mit allen Grundrechenarten
print("Plus 10:", zahlen + 10)
print("Minus 5:", zahlen - 5)
print("Geteilt durch 2:", zahlen / 2)
print("Potenz 2:", zahlen ** 2)
```

### Beispiel 2: Mathematische Funktionen

```python
import numpy as np
import matplotlib.pyplot as plt

# Array von x-Werten
x = np.linspace(0, 2*np.pi, 100)  # 100 Punkte zwischen 0 und 2Ï€

# Die Sinus-Funktion auf alle x-Werte anwenden
y = np.sin(x)

# Plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sinuskurve mit NumPy')
plt.grid(True)
plt.show()

# Andere mathematische Funktionen
print("Cosinus:", np.cos(x[:5]))  # Die ersten 5 Werte
print("Exponential:", np.exp(x[:3]))  # e^x fÃ¼r die ersten 3 Werte
print("Logarithmus:", np.log([1, 10, 100, 1000]))
```

### Beispiel 3: Logische Operationen und Filterung

```python
import numpy as np

# Array mit Messwerten
temperaturen = np.array([18, 22, 25, 30, 28, 15, 19])

# Welche Temperaturen sind Ã¼ber 25Â°C?
hoch = temperaturen > 25
print("Ãœber 25Â°C:", hoch)  # [False False True True True False False]

# Die tatsÃ¤chlichen Werte extrahieren
hohe_temperaturen = temperaturen[temperaturen > 25]
print("Hohe Temperaturen:", hohe_temperaturen)  # [30 28]

# Anzahl der hohen Temperaturen
anzahl = (temperaturen > 25).sum()
print("Anzahl Ã¼ber 25Â°C:", anzahl)
```

## ðŸŽ² Teil 4: Broadcasting â€“ Arrays verschiedener GrÃ¶ÃŸen kombinieren

**Broadcasting** ist eine magische NumPy-FÃ¤higkeit: Sie kÃ¶nnen Arrays unterschiedlicher GrÃ¶ÃŸe zusammen verwenden!

### Beispiel 1: Zahl mit Array

```python
import numpy as np

# Array von Messwerten (in Metern)
lÃ¤ngen = np.array([1.5, 2.0, 2.5, 3.0])

# Alle in Zentimeter umrechnen: einfach mit 100 multiplizieren
lÃ¤ngen_cm = lÃ¤ngen * 100
print("Original (m):", lÃ¤ngen)
print("Umgerechnet (cm):", lÃ¤ngen_cm)

# Das ist Broadcasting! Die Zahl 100 wird "ausgedehnt" auf die GrÃ¶ÃŸe des Arrays
```

### Beispiel 2: Verschiedene Array-GrÃ¶ÃŸen

```python
import numpy as np

# Ein Array mit einer Spalte
spalte = np.array([[1],
                   [2],
                   [3]])
print("Spalte Shape:", spalte.shape)  # (3, 1)

# Ein Array mit einer Reihe
reihe = np.array([10, 20, 30])
print("Reihe Shape:", reihe.shape)  # (3,)

# Addieren! NumPy dehnt beide Arrays automatisch auf (3, 3) aus
ergebnis = spalte + reihe
print("Ergebnis:")
print(ergebnis)
# [[11 21 31]
#  [12 22 32]
#  [13 23 33]]
```

### Beispiel 3: Praktisches Beispiel â€“ Normalisierung

```python
import numpy as np

# Messwerte von 3 Sensoren (3 Zeilen, 4 Messungen pro Sensor)
messwerte = np.array([[10, 15, 20, 25],
                      [100, 120, 140, 160],
                      [1000, 1100, 1200, 1300]])

print("Original:")
print(messwerte)

# Durchschnittswerte pro Sensor
durchschnitte = messwerte.mean(axis=1, keepdims=True)
print("\nDurchschnitte pro Sensor:")
print(durchschnitte)

# Alle Werte um den Durchschnitt zentrieren (Normalisierung)
normalisiert = messwerte - durchschnitte
print("\nNormalisiert (um Durchschnitt zentriert):")
print(normalisiert)
```

## ðŸ”¬ Teil 5: Praktische Anwendung â€“ Numerische Integration

**Numerische Integration** bedeutet: Den FlÃ¤cheninhalt unter einer Kurve berechnen.

### Die Idee

Statt die mathematische Formel zu nutzen, kÃ¶nnen wir:
1. Die Kurve in kleine Rechtecke aufteilen
2. Die FlÃ¤chen der Rechtecke addieren
3. Das ist ein sehr guter NÃ¤herungswert fÃ¼r das echte Integral

### Beispiel: FlÃ¤che unter einer Sinuskurve

```python
import numpy as np
import matplotlib.pyplot as plt

# Aufteilen des Intervals [0, Ï€] in n Punkte
n = 50
x = np.linspace(0, np.pi, n)
dx = np.pi / (n - 1)  # Breite eines Rechtecks

# HÃ¶hen der Rechtecke (Funktionswerte)
y = np.sin(x)

# FlÃ¤che = Summe aller Rechtecke
flaeche = np.sum(y) * dx
print(f"Numerisch berechnete FlÃ¤che: {flaeche:.4f}")

# Das echte Ergebnis (mathematisch)
echte_flaeche = 2.0  # âˆ«â‚€^Ï€ sin(x)dx = 2
print(f"Echte FlÃ¤che: {echte_flaeche:.4f}")
print(f"Fehler: {abs(flaeche - echte_flaeche):.4f}")

# Visualisierung
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
x_fein = np.linspace(0, np.pi, 200)
y_fein = np.sin(x_fein)
plt.plot(x_fein, y_fein, 'b-', linewidth=2, label='sin(x)')
plt.fill_between(x, y, alpha=0.3, label=f'NÃ¤herung (n={n})')
plt.bar(x, y, width=dx, alpha=0.3, edgecolor='black')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Numerische Integration mit Rechtecken')
plt.legend()
plt.grid(True)

# Genauigkeit mit unterschiedlichem n
plt.subplot(1, 2, 2)
n_werte = [10, 20, 50, 100, 200, 500]
fehler = []
for n_test in n_werte:
    x_test = np.linspace(0, np.pi, n_test)
    y_test = np.sin(x_test)
    dx_test = np.pi / (n_test - 1)
    flaeche_test = np.sum(y_test) * dx_test
    fehler.append(abs(flaeche_test - echte_flaeche))

plt.loglog(n_werte, fehler, 'ro-', linewidth=2, markersize=8)
plt.xlabel('Anzahl der Punkte (n)')
plt.ylabel('Fehler')
plt.title('Wie der Fehler mit mehr Punkten abnimmt')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## ðŸŽ¬ Teil 6: Praktische Anwendung â€“ Bewegungssimulation

Stellen Sie sich vor, wir werfen einen Ball. Mit NumPy kÃ¶nnen wir seine Position in jedem Moment berechnen.

### Die Physik

- **Geschwindigkeit**: v = Î”x / Î”t (wie schnell sich etwas bewegt)
- **Beschleunigung**: a = Î”v / Î”t (wie schnell die Geschwindigkeit sich Ã¤ndert)
- **Freier Fall**: a = -9.81 m/sÂ² (die Schwerkraft)

### Simulation: Springender Ball

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter
g = 9.81  # Gravitationsbeschleunigung in m/sÂ²
v0 = 20   # Anfangsgeschwindigkeit in m/s (nach oben)
dt = 0.01 # Zeitschritt in Sekunden
t_max = 4 # Gesamtdauer in Sekunden

# Zeitpunkte
t = np.arange(0, t_max, dt)
n = len(t)

# HÃ¶he und Geschwindigkeit initialisieren
y = np.zeros(n)
v = np.zeros(n)

y[0] = 0        # Starten bei HÃ¶he 0
v[0] = v0       # Mit Anfangsgeschwindigkeit

# Simulation: Schritt fÃ¼r Schritt
for i in range(1, n):
    # Neue Geschwindigkeit (Beschleunigung wirkt sich aus)
    v[i] = v[i-1] - g * dt
    
    # Neue HÃ¶he
    y[i] = y[i-1] + v[i] * dt
    
    # Wenn der Ball den Boden berÃ¼hrt, springt er zurÃ¼ck
    if y[i] < 0:
        y[i] = -y[i]           # HÃ¶he spiegeln
        v[i] = -v[i] * 0.9     # Geschwindigkeit umkehren (mit Energieverlust)

# Visualisierung
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# HÃ¶he Ã¼ber Zeit
axes[0, 0].plot(t, y, 'b-', linewidth=2)
axes[0, 0].axhline(y=0, color='k', linestyle='--', alpha=0.3)
axes[0, 0].set_xlabel('Zeit (s)')
axes[0, 0].set_ylabel('HÃ¶he (m)')
axes[0, 0].set_title('HÃ¶he des Balls Ã¼ber Zeit')
axes[0, 0].grid(True)

# Geschwindigkeit Ã¼ber Zeit
axes[0, 1].plot(t, v, 'r-', linewidth=2)
axes[0, 1].axhline(y=0, color='k', linestyle='--', alpha=0.3)
axes[0, 1].set_xlabel('Zeit (s)')
axes[0, 1].set_ylabel('Geschwindigkeit (m/s)')
axes[0, 1].set_title('Geschwindigkeit des Balls Ã¼ber Zeit')
axes[0, 1].grid(True)

# Phase-Raum: Geschwindigkeit vs. HÃ¶he
axes[1, 0].plot(y, v, 'g-', linewidth=2)
axes[1, 0].scatter(y[0], v[0], color='blue', s=100, label='Start', zorder=5)
axes[1, 0].scatter(y[-1], v[-1], color='red', s=100, label='Ende', zorder=5)
axes[1, 0].set_xlabel('HÃ¶he (m)')
axes[1, 0].set_ylabel('Geschwindigkeit (m/s)')
axes[1, 0].set_title('Phase-Raum')
axes[1, 0].legend()
axes[1, 0].grid(True)

# Animation-Ã¤hnliche Darstellung: einige Positionen
indices = np.arange(0, n, int(n/20))  # Jede 20. Position
axes[1, 1].scatter(indices * dt, y[indices], s=50, alpha=0.6)
for idx in indices:
    circle = plt.Circle((indices[0] * dt, y[indices[0]]), 0.5, 
                        fill=False, edgecolor='blue', alpha=0.3)
axes[1, 1].set_xlim(0, t_max)
axes[1, 1].set_ylim(-2, max(y) + 2)
axes[1, 1].set_xlabel('Zeit (s)')
axes[1, 1].set_ylabel('HÃ¶he (m)')
axes[1, 1].set_title('Ball-Positionen zu verschiedenen Zeiten')
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()

print(f"Maximale HÃ¶he: {y.max():.2f} m")
print(f"Erreichte Zeit: {t_max:.2f} s")
```
