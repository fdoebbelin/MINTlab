## ‚úÖ L√∂sung Aufgabe 1: Gerade oder ungerade?

### Vollst√§ndige L√∂sung

````
# Variable f√ºr die zu √ºberpr√ºfende Zahl
zahl = 7

# Bedingte Anweisung: Ist der Rest der Division durch 2 gleich 0?
if zahl % 2 == 0:
    print(f"{zahl} ist gerade")
else:
    print(f"{zahl} ist ungerade")
````

**Ausgabe:** `7 ist ungerade`

### Erkl√§rung

- Der **Modulo-Operator** `%` gibt den Rest der Division an
- `7 % 2 = 1` (Rest 1, also ungerade)
- `8 % 2 = 0` (Rest 0, also gerade)
- Mit einer `if/else`-Struktur unterscheiden wir die beiden F√§lle

### Variante: Mit Benutzereingabe

````
zahl = int(input("Gib eine Zahl ein: "))

if zahl % 2 == 0:
    print(f"{zahl} ist gerade")
else:
    print(f"{zahl} ist ungerade")
````

## ‚úÖ L√∂sung Aufgabe 5: Summe mit einer Schleife

### Vollst√§ndige L√∂sung

````
summe = 0  # Initialisierung: Starte mit 0

# Schleife: Addiere jede Zahl von 1 bis 100 zur summe
for i in range(1, 101):
    summe = summe + i  # oder: summe += i

print(f"Die Summe von 1 bis 100 ist: {summe}")
````

**Ausgabe:** `Die Summe von 1 bis 100 ist: 5050`

### Erkl√§rung

- **Akkumulator-Pattern:** `summe` startet bei 0 und wird in jeder Iteration erh√∂ht
- `summe = summe + i` ist gleichbedeutend mit `summe += i`
- Nach 100 Iterationen ist die Gesamtsumme berechnet
- Mathematisch: 1 + 2 + 3 + ... + 100 = 5050

### Mathematik: Gau√üsche Summenformel

Carl Friedrich Gau√ü entdeckte, dass die Summe der Zahlen 1 bis n mit dieser Formel berechnet werden kann:

$$\text{Summe} = \frac{n \times (n+1)}{2}$$

F√ºr n = 100:
$$\text{Summe} = \frac{100 \times 101}{2} = \frac{10100}{2} = 5050$$

Diese Formel ist deutlich schneller als eine Schleife mit 100 Iterationen!

### Verifikation in Python

````
# Mit Schleife
summe_schleife = 0
for i in range(1, 101):
    summe_schleife += i

# Mit Gau√üscher Formel
summe_formel = 100 * 101 // 2

print(f"Schleife: {summe_schleife}")
print(f"Formel: {summe_formel}")
print(f"Gleich? {summe_schleife == summe_formel}")
````

**Ausgabe:**
```
Schleife: 5050
Formel: 5050
Gleich? True
```

## ‚úÖ L√∂sung Aufgabe 6: Countdown

### Vollst√§ndige L√∂sung

````
# range(5, 0, -1) bedeutet: Von 5 bis 1, Schritte von -1
for i in range(5, 0, -1):
    print(i)

print("Z√ºnden!")
````

**Ausgabe:**
```

```

### Erkl√§rung

- `range(start, stop, step)`
  - **start:** 5 (Anfang)
  - **stop:** 0 (nicht erreicht! Stoppt bei 1)
  - **step:** -1 (r√ºckw√§rts z√§hlen)
- Die Zahl 0 wird nicht erreicht, deshalb ist die letzte Zahl 1

### Weitere Countdown-Varianten

````
# Mit while-Schleife
zaehler = 5
while zaehler > 0:
    print(zaehler)
    zaehler -= 1
print("Z√ºnden!")
````

````
# Mit input() f√ºr l√§ngere Countdowns
n = int(input("Von welcher Zahl soll der Countdown gehen? "))
for i in range(n, 0, -1):
    print(i)
print("Z√ºnden!")
````

## ‚úÖ L√∂sung Aufgabe 7: Schleife mit `break`

### Vollst√§ndige L√∂sung

````
# Schleife: Z√§hle von 1 bis 10
for i in range(1, 11):
    if i == 7:  # Wenn i gleich 7 ist
        break    # Beende die Schleife sofort
    print(i)
````

**Ausgabe:**

### Erkl√§rung

- `break` beendet die Schleife sofort
- Die Zahl 7 wird nicht mehr gedruckt
- Der Code nach der Schleife wird weiterhin ausgef√ºhrt (falls vorhanden)

### Alternative: Mit einer anderen Bedingung

````
# Breche ab, wenn die Zahl gr√∂√üer oder gleich 7 ist
for i in range(1, 11):
    if i >= 7:
        break
    print(i)
````

### Praktisches Beispiel: Suche in einer Liste

````
zahlen = [3, 5, 7, 9, 2, 6]
gesuchte_zahl = 7

for zahl in zahlen:
    print(f"Pr√ºfe: {zahl}")
    if zahl == gesuchte_zahl:
        print(f"Gefunden: {gesuchte_zahl}")
        break
````

**Ausgabe:**

## ‚úÖ L√∂sung Aufgabe 8: Schleife mit `continue`

### Vollst√§ndige L√∂sung

````
# Schleife: Z√§hle von 1 bis 10
for i in range(1, 11):
    if i % 2 == 0:      # Wenn i gerade ist
        continue        # √úberspringe diesen Durchlauf
    print(i)            # Drucke nur ungerade Zahlen
````

**Ausgabe:**
```
1

5
7
9
```

### Erkl√§rung

- `continue` springt zum n√§chsten Durchlauf der Schleife
- Alle Anweisungen nach `continue` werden √ºbersprungen
- Die geraden Zahlen (2, 4, 6, 8, 10) werden nicht gedruckt

### Vergleich: `break` vs. `continue`

| Befehl | Effekt | Beispiel |
|--------|--------|----------|
| `break` | Beendet die Schleife komplett | Stoppt nach dem ersten Fehler |
| `continue` | Springt zum n√§chsten Durchlauf | √úberspringt ung√ºltige Eintr√§ge |

## ‚úÖ L√∂sung Aufgabe 10: `while`-Schleife mit Benutzereingabe

### Vollst√§ndige L√∂sung

````
richtige_zahl = 42
geraten = 0

# Wiederhole, bis die richtige Zahl geraten wurde
while geraten != richtige_zahl:
    geraten = int(input("Rate eine Zahl zwischen 1 und 100: "))
    
    if geraten < richtige_zahl:
        print("Zu klein!")
    elif geraten > richtige_zahl:
        print("Zu gro√ü!")
    else:
        print("Richtig! Du hast gewonnen!")
````

### Beispiel-Durchlauf

```
Rate eine Zahl zwischen 1 und 100: 50
Zu klein!
Rate eine Zahl zwischen 1 und 100: 60
Zu gro√ü!
Rate eine Zahl zwischen 1 und 100: 42
Richtig! Du hast gewonnen!
```

### Erkl√§rung

- Die `while`-Schleife l√§uft solange, wie `geraten != richtige_zahl` ist
- Mit `int(input(...))` wird die Benutzereingabe in eine Ganzzahl umgewandelt
- Mit `if/elif/else` geben wir R√ºckmeldung
- Wenn die Bedingung falsch wird, endet die Schleife

### Erweiterung: Mit Versuchen z√§hlen

````
richtige_zahl = 42
geraten = 0
versuche = 0

while geraten != richtige_zahl:
    geraten = int(input("Rate eine Zahl zwischen 1 und 100: "))
    versuche += 1
    
    if geraten < richtige_zahl:
        print("Zu klein!")
    elif geraten > richtige_zahl:
        print("Zu gro√ü!")
    else:
        print(f"Richtig! Du hast {versuche} Versuche gebraucht!")
````

## ‚úÖ L√∂sung Aufgabe 11: Lieblingszahlen filtern

### Vollst√§ndige L√∂sung

````
zahlen = [5, 12, 3, 18, 7, 25, 9, 14]

# Iteriere √ºber alle Zahlen in der Liste
for zahl in zahlen:
    if zahl > 10:  # Wenn Zahl gr√∂√üer als 10 ist
        print(zahl)
````

**Ausgabe:**
```
12
18
25
14
```

### Erkl√§rung

- `for zahl in zahlen:` iteriert √ºber jedes Element der Liste
- Mit `if zahl > 10:` filtern wir die Zahlen
- Nur Zahlen > 10 werden gedruckt

### Alternative: Mit List Comprehension (fortgeschrittene Technik)

````
zahlen = [5, 12, 3, 18, 7, 25, 9, 14]

# Erstelle eine neue Liste mit nur Zahlen > 10
grosse_zahlen = [zahl for zahl in zahlen if zahl > 10]
print(grosse_zahlen)
````

**Ausgabe:** `[12, 18, 25, 14]`

### Weitere Filterbeispiele

````
# Nur gerade Zahlen
gerade = [zahl for zahl in zahlen if zahl % 2 == 0]

# Nur ungerade Zahlen
ungerade = [zahl for zahl in zahlen if zahl % 2 != 0]

# Nur Zahlen zwischen 10 und 20
mittel = [zahl for zahl in zahlen if 10 < zahl < 20]
````

## ‚úÖ L√∂sung Aufgabe 12: Sterne ausgeben

### Vollst√§ndige L√∂sung

````
anzahl = int(input("Wie viele Sterne? "))

# Schleife: Gebe anzahl-mal einen Stern ohne Zeilenumbruch aus
for i in range(anzahl):
    print("*", end="")

# Am Ende eine Neue Zeile
print()
````

### Beispiel-Durchl√§ufe

**Input:** `5` ‚Üí **Ausgabe:** `*****`

**Input:** `10` ‚Üí **Ausgabe:** `**********`

### Erkl√§rung

- `print("*", end="")` druckt einen Stern ohne Zeilenumbruch
- Der Parameter `end=""` setzt das Trennzeichen auf eine leere Zeichenkette
- Nach der Schleife rufen wir `print()` auf, um eine neue Zeile zu erzeugen

### Alternative 1: Mit String-Multiplikation

````
anzahl = int(input("Wie viele Sterne? "))

# In Python kann man Strings mit Zahlen multiplizieren
print("*" * anzahl)
````

Das ist viel eleganter und effizienter!

### Alternative 2: Mit `join()`

````
anzahl = int(input("Wie viele Sterne? "))

# Erstelle eine Liste mit Sternen und verbinde sie
print("".join(["*" for _ in range(anzahl)]))
````

### Erweiterung: Rechteck aus Sternen

````
breite = int(input("Breite: "))
hoehe = int(input("H√∂he: "))

# √Ñu√üere Schleife f√ºr die H√∂he
for i in range(hoehe):
    # Innere Schleife f√ºr die Breite
    for j in range(breite):
        print("*", end="")
    print()  # Neue Zeile nach jeder Reihe
````

**Beispiel-Output (3 √ó 3):**
```

***
***
```
